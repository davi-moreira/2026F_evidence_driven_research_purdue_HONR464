## 🎯 Definition of Done

You are done with this sprint when your submission contains the whole landscape,
built from **your own** curiosity:

1. **The hook.** Your curiosity in about three sentences, stated so a stranger
   would care, naming the big motivation and the small piece you will study.
2. **The funnel and its three terms.** Curiosity → topic → research problem, and
   the three terms your problem carries: **units of analysis**, **target
   population**, and **outcome**.
3. **The opportunity landscape.** At least two candidate problems, and the one you
   chose, defended on **importance**, **feasibility**, and **contribution**, with a
   one-line contribution case whose sources are real and retrievable.
4. **The answerable question and its self-score.** One sentence evidence can answer
   and be wrong about, scored on units, target population, outcome, and whether it
   can be wrong.
5. **The compass declaration and its boundary.** The M1-shape sentence ("my
   question is a [descriptive or causal] inquiry reaching [the data at hand / a
   population / unseen cases], because ..."), the claim it **permits**, the claim it
   **forbids**, and the one **giveaway word** with whether it tells the truth or
   traps.
6. **Uncertainty and limitations.** One honest sentence on what you are still
   unsure of.
7. **AI Research Ledger rows.** One row per AI-assisted step, each with a named
   verification method. A missing ledger returns the submission.

Plus the **dossier update line** naming where each part now lives in your Research
Project Dossier. You present this as a 2-minute pitch at the studio, then revise
from the red-team.

> wording will keep changing through M2 and M3. What is graded now is whether your

### 🧭 Lay Out Your Landscape

Edit the `candidates` list below to hold **your own** candidate problems, then run
the cell. It scores each one on the three worth-doing tests from nb01 and flags
which clear all three at once. The `candidates` list is the single source of truth,
so adding, renaming, or rescoring a problem cannot fall out of sync with the table.

The cell shuffles the display order with seed 464 so you judge each problem on its
scores, not on which one you happened to write first. A survivor is a *candidate*;
the choice is still yours to defend in prose.

```python
# Opportunity-landscape scorecard — edit the `candidates` list to hold YOUR problems.
# This list is the SINGLE SOURCE OF TRUTH: every column below is computed from it,
# so renaming, adding, or rescoring a candidate keeps the whole table consistent.

# EDIT ME — replace these with YOUR candidate problems (at least two).
# Each `problem` text should name its units, outcome, and target population.
# Score each 1-5 on the three worth-doing tests, using YOUR honest read.
candidates = [
    {"problem": "EDIT ME — candidate problem 1 (name its units, outcome, population)",
     "importance": 4, "feasibility": 3, "contribution": 4},
    {"problem": "EDIT ME — candidate problem 2 (name its units, outcome, population)",
     "importance": 3, "feasibility": 5, "contribution": 2},
]

BAR = 3  # a test is "cleared" at 3 of 5 or higher; raise it if your bar is stricter.
TESTS = ["importance", "feasibility", "contribution"]  # the three nb01 tests

land = pd.DataFrame(candidates)
# Shuffle the display order (seed 464) so you weigh the scores, not the writing order.
land = land.sample(frac=1, random_state=SEED).reset_index(drop=True)

# A candidate "survives" only if it clears the bar on ALL THREE tests at once.
# We read the test names from TESTS, never a hardcoded column, so the survivor
# rule stays in sync if you rename or reweight a test.
land["survives_all_three"] = (land[TESTS] >= BAR).all(axis=1)

print("Your opportunity landscape (display order shuffled, seed 464):")
print(land.to_string(index=False))
print()

survivors = land.loc[land["survives_all_three"], "problem"].tolist()
if survivors:
    print(f"✓ {len(survivors)} candidate(s) clear all three tests.")
    print("  A survivor is a CANDIDATE, not a verdict. The choice is yours to defend.")
else:
    print("⚠️ No candidate clears all three yet. That is a finding, not a dead end:")
    print("  sharpen one problem, or lower a score you cannot honestly defend.")
```

### 🗡️ Red-Team Exchange

The studio's 29–41 minute block. Pair with one partner and run two rounds of
attack. Bring in an AI reviewer as a third voice using the **Research Question
Diagnostician** role card (`ai_resources/agent_role_cards.md`; full briefing in
`genai_studio/roles/research_question_diagnostician.md`). It reads a question
against the compass and names the nearest wrong classification, so you can defend
the right one.

**Round 1: attack the compass declaration (about 6 minutes).**
Your partner reads your declaration and tries to catch a **silent upgrade**: a
permitted claim that quietly reaches a population your data cannot speak for, or
drifts toward cause. They also paste your question to the Diagnostician and ask for
its compass reading and the single word driving it. You defend the classification
against the compass definitions, or reclassify, then rewrite the boundary.

**Round 2: attack the question and the landscape choice (about 6 minutes).**
Your partner checks that your question truly names its three terms and can be
wrong, and questions whether your chosen problem really beats its alternatives on
all three tests. They challenge one term or one test you cannot defend, and you
sharpen it or concede.

**What to log.** Write down the single strongest hit from each round and how you
changed the landscape in response. An attack that changed nothing still gets a line
saying why it did not land. That record is part of what you submit.
