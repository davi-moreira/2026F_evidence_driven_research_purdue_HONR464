### 🧭 Check Your Ledger Is Really Built

Edit the `sources` list below to hold **your own** retrieved sources, then run the
cell. It checks that each row names where you retrieved it, states a claim in your
words, and carries a real evidence type, then counts how many read as verified
against the six M2 asks for. The `sources` list is the single source of truth, so
adding, editing, or removing a source cannot fall out of sync with the counts.

The cell shuffles the display order with seed 464 so you judge each row on its
retrieval, not on where it happened to land in your list. A row that passes is one
you can *defend*, not one you pasted.

```python
# Verified-source ledger scorecard — edit the `sources` list to hold YOUR sources.
# This list is the SINGLE SOURCE OF TRUTH: every check below is computed from it, so
# adding, editing, or removing a source keeps the table and the counts in sync.

# EDIT ME — one dict per source you PERSONALLY retrieved and read.
#   where_retrieved: the exact database/journal/repository page, never "an AI told me".
#   claim:           the source's central claim, in YOUR OWN words.
#   evidence_type:   what the evidence actually supports — one of the four below.
#   relation:        how it sits with your other sources, for the evidence map you
#                    draw in nb02 — "agrees", "contradicts", "qualifies", or "standalone".
sources = [
    {"cite": "EDIT ME — author, year, title, venue",
     "where_retrieved": "EDIT ME — e.g. Purdue Libraries database, PDF read in full",
     "claim": "EDIT ME — the source's central claim, in your own words",
     "evidence_type": "descriptive", "relation": "standalone"},
    {"cite": "EDIT ME — author, year, title, venue",
     "where_retrieved": "EDIT ME — e.g. Google Scholar, publisher PDF read in full",
     "claim": "EDIT ME — the source's central claim, in your own words",
     "evidence_type": "generalization", "relation": "contradicts"},
    # ... keep adding rows until you have at least six RETRIEVED sources.
]

TARGET = 6  # M2 requires at least this many personally retrieved sources
EVIDENCE_TYPES = {"descriptive", "generalization", "predictive", "causal"}

led = pd.DataFrame(sources)
# Shuffle display order (seed 464) so you weigh each row's retrieval, not its position.
led = led.sample(frac=1, random_state=SEED).reset_index(drop=True)

# A row reads as "verified-shaped" only if it names a real retrieval location, a claim
# in your words, and a valid evidence type. Anything still saying "EDIT ME" or sourced
# to an AI fails, because a source you cannot locate is not evidence yet. We read the
# valid types from EVIDENCE_TYPES, never a hardcoded string, so the rule stays in sync.
def needs_work(row):
    blob = f"{row['where_retrieved']} {row['claim']} {row['cite']}".lower()
    unretrieved = ("edit me" in blob) or ("ai told me" in blob)
    bad_type = row["evidence_type"] not in EVIDENCE_TYPES
    return unretrieved or bad_type

led["needs_work"] = led.apply(needs_work, axis=1)

print(f"Your verified-source ledger ({len(led)} rows, display order shuffled, seed 464):")
print(led[["cite", "where_retrieved", "evidence_type", "relation", "needs_work"]]
      .to_string(index=False))
print()

verified = int((~led["needs_work"]).sum())
print(f"Rows that read as verified: {verified} of a required {TARGET}.")
if verified >= TARGET:
    print("✓ You have at least six retrieved sources. Each row is a claim you can")
    print("  defend, not a citation you pasted. The map and the gap are still yours to reason.")
else:
    print(f"⚠️ {TARGET - verified} more retrieved source(s) needed. That is a finding,")
    print("  not a dead end: run the retrieval-verification loop on your next lead and log it.")
```

### 🗡️ Red-Team Exchange

The studio's 29–41 minute block. Pair with one partner and run two rounds of attack.
Bring in an AI reviewer as a third voice using the **Evidence & Citation Verifier**
role card (full briefing in `genai_studio/roles/evidence_citation_verifier.md`). It
checks whether a citation is *checkable* and flags what looks fabricated, then sends
you to the library to confirm it yourself. It never certifies that a source is real,
so a citation it does not flag is still not confirmed until you retrieve it.

**Round 1: attack the ledger (about 6 minutes).**
Your partner picks two rows from your ledger and asks the question that decides this
milestone: "show me where you retrieved this, and the claim in the source's own
words." You paste each citation to the Evidence & Citation Verifier and read back its
checkability review and fabrication-risk flags. Any row you cannot defend by naming
its retrieval, or that the Verifier flags and you have not re-checked, comes off the
ledger until you retrieve it again.

**Round 2: attack the gap (about 6 minutes).**
Your partner tries to close your gap: they name a source, yours or one of their own,
that might already cover the spot where you placed it. If one of your own verified
sources fills the gap, that is the most important find of the block, and you move the
gap to where the map is genuinely quiet. You defend the gap against the honesty test,
or you concede and re-place it.

**What to log.** Write down the single strongest hit from each round and how you
changed the ledger or the gap in response. An attack that changed nothing still gets
a line saying why it did not land. That record is part of what you submit.

> **A question that often comes up here:** *"What if the Verifier flags a source I
> know is real?"* Retrieve it and keep it. A flag is a prompt to check, never a
> verdict, and real sources sometimes look odd. The reverse also holds: a clean-looking
> citation the Verifier did not flag is not confirmed either. Only your own retrieval
> of the actual source settles whether it belongs on your ledger.
