## 1. Poster Logic vs Paper Logic

**Guiding question:** *what does a research note have to argue in writing that
your poster was allowed to imply while you stood next to it?*

> *"Your poster let you point at a box and talk. Your note has no you standing
> next to it. Every leap the box let you imply, the prose has to earn, in order,
> for a reader who will never hear you explain it."*
> — a **thesis advisor**, reading the first draft of your research note

Two words name the shift, and they sound alike until you write one.

- **Poster logic** is communication built for a reader standing in front of you,
  where six boxes and your live voice carry the argument together. Example: a
  "Gap" box that just says *"little work on rural clinics"* while you explain the
  rest out loud.
- **Paper logic** is communication that has to stand alone on the page, where
  every claim is argued in prose because you are not there to fill the silences.
  Example: that same gap, now a paragraph naming what prior work covered, what it
  left out, and why the missing piece matters.

The move from one to the other is not "add more words." It is **make the
reasoning visible.** A box implies the connection between problem and question; a
paragraph has to state it, in order, so a stranger reaches your conclusion by the
same steps you did.

> **A question that often comes up here:** *"If my poster already made the point,
> why rewrite it as prose at all?"* Because the poster made the point to a reader
> you could coach in real time. The note makes it to a reader who only has the
> page. Anything you carried in your voice, the tone that said *"this gap really
> matters,"* the gesture that linked two boxes, is gone. Prose is where you put
> back everything your presence used to supply. The note is not longer for its
> own sake. It is longer because it has to do the work your voice was doing.

## 2. The Problem–Gap–Question Spine and Evidence Density

**Guiding question:** *how do you build an argument a reader can follow from the
trouble that started it to the question you can actually answer?*

A research note has a backbone, and readers feel it even when they cannot name
it. This course calls it the **problem–gap–question spine**, the same name your
M14 note uses. Three linked pieces, each defined plainly:

- **Problem** is the real-world or scholarly trouble your work responds to.
  Example: rural clinics report supply shortages that no one has measured.
- **Gap** is the specific thing prior, retrievable work has not yet established.
  Example: existing surveys cover urban clinics only, so rural stock-outs stay
  undocumented.
- **Question** is the answerable question that closes exactly that gap. Example:
  in the rural clinics my frame reaches, what share reported a stock-out last
  quarter?

The spine only holds if the gap is real, and a gap is real only when you can show
what is already known. That is **literature synthesis**: a few real, retrievable
sources woven into an argument about what is settled and what is missing, never a
name-drop. Example: three studies that each measured part of the problem, each
cited so a reader can open it.

The enemy of synthesis has a measurable signature. **Evidence density** is how
much of a paragraph rests on sources a reader could retrieve, versus bare
assertion. Example: *"as everyone knows"* carries zero density; a sentence
resting on a source a reader can open carries some. A paragraph of confident
claims with nothing to retrieve is not a synthesis. It is an opinion in a lab
coat, and Lecture 1 exists to catch it.

```python
# Draw the problem->gap->question spine as the backbone a note argues in order.
# (networkx is imported once in the setup cell.)
G = nx.DiGraph()
nodes = ["Problem", "Gap", "Question", "Design", "Evidence"]
pos = {n: (i * 1.6, 0.0) for i, n in enumerate(nodes)}
G.add_nodes_from(nodes)
edge_labels = {
    ("Problem", "Gap"): "motivates",
    ("Gap", "Question"): "sharpens into",
    ("Question", "Design"): "answered by",
    ("Design", "Evidence"): "produces",
}
G.add_edges_from(edge_labels.keys())

fig, ax = plt.subplots(figsize=(11, 3.2))
nx.draw_networkx_nodes(G, pos, node_color="#DCE6F1", edgecolors="#4C72B0",
                       linewidths=1.5, node_size=4600, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=10, ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=18, edge_color="#555555",
                       width=1.6, node_size=4600, ax=ax)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8,
                             label_pos=0.5, ax=ax)
ax.set_title("The problem-to-gap-to-question spine a research note argues in order")
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ Spine drawn: each box becomes a paragraph, and each arrow is a sentence")
print("  the note has to earn. A poster shows the boxes; a note argues the arrows.")
```

**Reading the diagram.** Read it left to right and you are reading the order a
note must move in. The problem motivates the gap. The gap sharpens into one
answerable question. The question is answered by a design, and the design
produces evidence. On a poster each box can stand alone. In a note the arrows are
the argument, and every arrow is a sentence a reader can refuse to believe until
you earn it.

*In plain terms, the picture says a research note is not five boxes. It is four
connections, and the connections are what your prose has to argue.*

```python
# A tiny evidence-density scorer. A sentence "rests on a source" when it carries
# a retrievable marker [verified: ...]; density = sourced sentences / all sentences.
def evidence_density(paragraph):
    parts = paragraph.replace("!", ".").replace("?", ".").split(".")
    sentences = [s.strip() for s in parts if s.strip()]
    sourced = [s for s in sentences if "[verified" in s.lower()]
    n = len(sentences)
    return n, len(sourced), n - len(sourced), (len(sourced) / n if n else 0.0)

assertion = (
    "As everyone knows, people no longer trust their public institutions. "
    "Trust has obviously collapsed everywhere you look. "
    "It is common sense that nobody believes official numbers anymore. "
    "Clearly the situation is worse now than it has ever been."
)
synthesis = (
    "Reported trust in national institutions has fallen across several "
    "long-running surveys [verified: retrieved]. "
    "One widely cited 2000 study of civic life documented a decades-long decline "
    "in community-group membership [verified: retrieved]. "
    "A 2015 reproducibility project found that many published effects did not "
    "hold up under independent re-analysis [verified: retrieved]. "
    "Together these strands do not settle a single cause, but each rests on a "
    "source a reader can open."
)

for name, para in [("Assertion", assertion), ("Synthesis", synthesis)]:
    n, sourced, bare, dens = evidence_density(para)
    print(f"{name:10s}: {n} sentences, {sourced} sourced, {bare} bare "
          f"-> evidence density = {dens:.2f}")
```

```python
# Self-check: the assertion paragraph rests on nothing; the synthesis mostly does.
n_a, s_a, b_a, d_a = evidence_density(assertion)
n_s, s_s, b_s, d_s = evidence_density(synthesis)
assert d_a == 0.0, "the assertion paragraph should score 0.0 density"
assert d_s >= 0.7, "the synthesis paragraph should score at least 0.70 density"
assert b_a == 4 and s_s == 3, "expected 4 bare claims in assertion, 3 sourced in synthesis"
print(f"✓ Self-check passed: assertion density {d_a:.2f} ({b_a} bare claims), "
      f"synthesis density {d_s:.2f} ({s_s} sourced of {n_s}).")
```

**Reading the evidence.** The assertion paragraph scores 0.0. Every sentence
sounds confident, and not one rests on anything a reader could open. The synthesis
paragraph scores 0.75: three of its four sentences carry a source, and the fourth
is an honest summary that claims no more than the three support. The scorer is
crude on purpose. It cannot judge whether a source actually says what you claim.
It only asks the first question a reviewer asks: is there anything here to
retrieve at all? A paragraph that scores zero is not a synthesis. It is the *"as
everyone knows"* move the claim boundary forbids.

## 3. The Results Hierarchy and Limitations Sized to Evidence

**Guiding question:** *which of your findings has earned the right to lead, and
how big are your limitations allowed to be?*

> *"Show me your results in the order you trust them. If your flashiest number
> comes from forty people and leads the paper, I already know how this review
> ends. I do not want your most surprising result first. I want your most
> defensible one."*
> — a **journal reviewer**, on the first page they actually weigh

Not all findings are equal, and pretending they are is how good work overclaims.
Three terms keep the back half of a note honest.

- **Results hierarchy** is the order you present findings in, strongest and
  best-supported first, exploratory and fragile last. Example: a whole-sample
  relationship with a tight interval leads; a surprising result from a tiny
  subgroup comes last, clearly labeled exploratory.
- **Inflated contribution** is a claim written to sound like a bigger finding
  than the evidence supports. Example: reporting a fragile forty-person result as
  your main discovery because it is the most surprising thing you found.
- **Limitations sized to evidence** are limitations stated in proportion to what
  actually threatens your claim, neither hidden nor performed. Example: naming the
  resample caveat and the coverage gap that most bend your number, instead of a
  generic *"more research is needed."*

The ranking is not a matter of taste. It tracks how much evidence backs each
finding, and you can often measure that directly. A finding from your whole
sample, with a tight interval around it, sits at the top. A finding from a tiny
slice, with an interval so wide it barely rules anything out, sits at the bottom.
The next cell shows both, computed from the same data.

> **A question that often comes up here:** *"Isn't my most surprising result the
> one I should lead with?"* Surprise and evidence are different axes, and the
> whole discipline is not confusing them. Your most surprising result is often
> your least supported one, precisely because it came from a small or unusual
> slice where noise runs high. Lead with what you can defend, and the surprising
> finding earns a careful, clearly-labeled paragraph near the end. A reviewer who
> sees a fragile result leading the note stops trusting the whole thing.

```python
# Load a real survey and confirm its shape before trusting anything.
lapop = load_course_data("lapop_brazil.csv")
assert lapop.shape == (10000, 10), "unexpected shape — flag this!"
print("✓ Loaded lapop_brazil.csv —", lapop.shape[0], "rows,", lapop.shape[1], "columns")

# The headline candidate: are reported trust in police and in the military related?
r_full = lapop["trust_police"].corr(lapop["trust_military"])

def boot_corr_ci(frame, a, b, gen, n_boot=2000):
    """95% bootstrap interval for the correlation of two columns (seeded gen)."""
    va, vb = frame[a].to_numpy(), frame[b].to_numpy()
    m = len(frame)
    out = np.empty(n_boot)
    for i in range(n_boot):
        pick = gen.integers(0, m, m)          # resample rows with replacement
        out[i] = np.corrcoef(va[pick], vb[pick])[0, 1]
    return np.nanpercentile(out, [2.5, 97.5])

rng_boot = np.random.default_rng(SEED)        # isolated, reproducible on its own
lo_full, hi_full = boot_corr_ci(lapop, "trust_police", "trust_military", rng_boot)

# An exploratory candidate: the SAME relationship in a tiny 40-row slice.
small = lapop.sample(n=40, random_state=SEED)
r_small = small["trust_police"].corr(small["trust_military"])
lo_small, hi_small = boot_corr_ci(small, "trust_police", "trust_military", rng_boot)

print("\nRESULTS HIERARCHY (ranked by how much evidence backs each candidate):")
print(f"  Tier 1 headline    : r = {r_full:.2f}   95% CI [{lo_full:.2f}, {hi_full:.2f}]"
      f"   width {hi_full - lo_full:.2f}   (n = {len(lapop):,})")
print(f"  Tier 3 exploratory : r = {r_small:.2f}   95% CI [{lo_small:.2f}, {hi_small:.2f}]"
      f"   width {hi_small - lo_small:.2f}   (n = 40)")
```

```python
# Figure: the two intervals on one axis. Each tier gets its OWN hue; neither
# reuses the other's, so the wide interval never blends into the tight one.
fig, ax = plt.subplots(figsize=(9, 3.6))
ax.plot([lo_full, hi_full], [1, 1], color="#4C72B0", linewidth=3, solid_capstyle="round")
ax.scatter([r_full], [1], color="#4C72B0", s=110, zorder=3)
ax.plot([lo_small, hi_small], [0, 0], color="#8172B3", linewidth=3, solid_capstyle="round")
ax.scatter([r_small], [0], color="#8172B3", s=110, zorder=3)
ax.axvline(0, color="gray", linestyle=":", linewidth=1.2, label="no relationship (r = 0)")
ax.set_yticks([0, 1])
ax.set_yticklabels(["Exploratory\n(n = 40)", "Headline\n(n = 10,000)"])
ax.set_ylim(-0.6, 1.6)
ax.set_xlim(-0.2, 1.0)
ax.set_xlabel("Correlation r (police trust, military trust), with 95% bootstrap interval")
ax.set_title("A results hierarchy: a tight interval leads, a wide one waits at the back")
ax.legend(loc="lower right")
plt.tight_layout()
plt.show()

print(f"✓ Figure drawn. The headline interval is narrow; the 40-row interval is")
print(f"  much wider, which is exactly why the small-slice finding cannot lead.")
```

```python
# Self-check: the whole sample gives a tight interval; the tiny slice a wide one.
assert 0.40 < r_full < 0.50, "headline correlation drifted — check the data"
assert (hi_full - lo_full) < 0.10, "headline interval should be tight at n = 10,000"
assert (hi_small - lo_small) > (hi_full - lo_full), "the n=40 interval should be wider"
ratio = (hi_small - lo_small) / (hi_full - lo_full)
print(f"✓ Self-check passed: headline r = {r_full:.2f} on a tight interval "
      f"[{lo_full:.2f}, {hi_full:.2f}];\n  the 40-row interval is about {ratio:.0f}x wider, "
      f"so it belongs at the BACK of the note, labeled exploratory.")
```

## 4. The Reproducibility Capsule

**Guiding question:** *what does a stranger need to rerun your work and get your
numbers, without you in the room to help?*

A note that a reader trusts and cannot rerun is still half a claim. The other half
is the **reproducibility capsule**: everything a stranger needs to rebuild your
numbers, and nothing they would have to guess. Five parts:

- A **runnable notebook** that passes **restart-and-run-all**, which means
  clearing the kernel and running every cell top to bottom, proving the notebook
  works without you clicking cells in a remembered order. Example: Runtime,
  Restart and run all, and your headline number comes back.
- A **data-provenance note**: where each dataset came from, its version, and how
  it may be used. Example: the lapop resample caveat and its MIT license.
- A **fixed seed**, so every random step returns the same numbers, not just some.
- A **decision log** of the by-hand choices that shaped the result, each with its
  reason. Example: every exclusion, recode, and cut point, and why.
- Your finalized **AI-use ledger**: every tool, its task, and how you verified it.

Capsules break in boringly consistent ways, which is good news, because boring
failures are catchable. Formalized, they are **the five package sins**: a
**hard-coded path** that exists only on your laptop, a **missing seed** that moves
every run, a **by-hand edit** no clean run reproduces, an **undocumented
exclusion** with no logged reason, and **stale or unversioned data** a reader
cannot reobtain. The next cell hunts all five automatically.

> **A question that often comes up here:** *"My project is small, a few hundred
> rows. Isn't a whole capsule overkill?"* The size is why it is teachable, not why
> it is exempt. A hard-coded path breaks a two-cell notebook as surely as a
> two-thousand-cell one, and the discipline you build on something small is the
> discipline you will actually have when the stakes are large. Small is where you
> learn to be trustworthy cheaply.

### 🛠️ Hands-On: audit a reproducibility capsule

Run the cell below. It defines `audit_capsule`, a small, transparent checker that
scans a list of source lines for the five package sins. It does not judge your
ideas. It only asks whether a stranger could *run* your work. Watch it score a
sinful capsule and a clean one, then later paste your own notebook's key lines in.

**🏠 Homework depth: run this on your own.**

> by check, what each of the five tests is looking for in the source lines, and why
> each sin would break a cold rerun. Then give me one reproducibility problem this
> auditor cannot catch, no matter how many checks I add, that only a human running

>
> **After running, verify (counters *confident fabrication*: a fluent walkthrough
> can describe checks the function does not actually run):**
> - [ ] Match Gemini's list of flagged sins against the two lists your cell actually
>   printed. If it names a sin the auditor did not flag, trust the printout.
> - [ ] Confirm its "problem the auditor cannot catch" is real by re-reading the

> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# A tiny capsule auditor. Each check is one thing a stranger's eyes would catch.
# It scans a list of source LINES (strings) — paste your own capsule's key lines in.
def audit_capsule(lines):
    text = "\n".join(lines).lower()
    sins = []
    # Sin 1 — a hard-coded absolute path: dies on every machine but its author's.
    if any(("/users/" in ln.lower()) or ("/home/" in ln.lower()) for ln in lines):
        sins.append("hard-coded path")
    # Sin 2 — no fixed seed anywhere: any sample/split/shuffle moves every run.
    if not any(k in text for k in ("seed", "random_state", "default_rng")):
        sins.append("missing seed")
    # Sin 3 — a value patched by hand: no clean run will reproduce it.
    if any((".loc[" in ln and "=" in ln) or ("by hand" in ln.lower())
           or ("manually" in ln.lower()) for ln in lines):
        sins.append("by-hand edit")
    # Sin 4 — an exclusion/filter with no logged reason nearby.
    filtered = [ln for ln in lines
                if (".drop(" in ln) or (".dropna(" in ln)
                or ("df[df[" in ln.replace(" ", ""))]
    if filtered and not any("reason" in ln.lower() for ln in lines):
        sins.append("undocumented exclusion")
    # Sin 5 — data with no shareable provenance: a reader can't reobtain the inputs.
    loads_csv = any("read_csv" in ln for ln in lines)
    shareable = any(("http" in ln.lower()) or ("load_course_data" in ln) for ln in lines)
    versioned = any(("version" in ln.lower()) or ("license" in ln.lower()) for ln in lines)
    if loads_csv and not shareable and not versioned:
        sins.append("stale/unversioned data")
    return sins
```

```python
# A deliberately sinful capsule — every line breaks reproducibility on purpose.
sinful = [
    'df = pd.read_csv("/Users/rivera/Desktop/thesis/trust_FINAL_v3.csv")',
    'draw = df.sample(200)                    # nothing fixes the shuffle',
    'df = df[df["party"] == "NPP"]            # kept one group only; no note on why',
    'df.loc[41, "age"] = 33                   # patched this row by hand post-survey',
    '# data pulled sometime last spring; nobody logged the download date',
]

# A clean capsule — same study, nothing a stranger would have to guess.
clean = [
    "SEED = 464",
    "rng = np.random.default_rng(SEED)",
    'df = load_course_data("lapop_brazil.csv")   # provenance + MIT license in the ledger',
    'df = df[df["trust_police"].notna()]   # reason: pre-registered complete cases; in the decision log',
    "draw = df.sample(200, random_state=SEED)",
]

print("Sinful capsule flags:", audit_capsule(sinful))
print("Clean  capsule flags:", audit_capsule(clean))
```

```python
# Self-check: the sinful capsule trips all five checks, the clean one none.
assert audit_capsule(sinful) == ["hard-coded path", "missing seed", "by-hand edit",
                                 "undocumented exclusion", "stale/unversioned data"], \
    "auditor drifted — the sinful capsule should trip all five checks"
assert audit_capsule(clean) == [], "clean capsule should trip no checks"
print("✓ Self-check passed: 5 sins in the sinful capsule, 0 in the clean one.")
print("  Reminder: a clean pass proves the capsule is RUNNABLE, never that it is CORRECT.")
```

**Reading the evidence.** The sinful capsule trips all five checks; the clean one
trips none. But read the last line of the self-check again, because it carries the
whole lesson. A capsule with zero flags is not a capsule that is **correct**. It
is a capsule that is **runnable**. The auditor cannot see that your seed is fixed
on the wrong sample, that a logged reason for an exclusion is a bad reason, or that
your analysis answers a different question than the one you declared. Those are the
failures a real person finds when they run your work cold, which is why your M14
capsule is exercised by a partner, not just by a script. The auditor gets you to
the starting line. The partner is the race.
