## 1. What Is Left for the Researcher?

**Guiding question:** *when an AI can generate the code and the prose, which part of
research becomes more important, not less?*

> *"I don't care how confident you sound, or how polished the writing is. I care
> what your evidence is, where it came from, and what it can and cannot show. Bring
> me that, and we can talk."*
> — a **thesis advisor**, eventually, to every honors researcher

Welcome to the course. Your professor, **Davi Moreira**, has spent years measuring
whether things that sound true actually are, from academic research to development
projects on three continents. That gap, between sounding right and being right, is
exactly why this course exists.

Three ideas frame the whole semester. Meet them once, plainly.

- **Evidence-driven research** is turning a curiosity into a claim you can defend
  with traceable, verified evidence. Example: not "napping helps grades," but "in
  the 200 first-years I surveyed, nappers averaged 0.2 grade points higher, and here
  is the survey, the numbers, and the limits."
- **AI as your arm and research assistant, not your brain** is the course's iron
  rule. The tool can reach, fetch, draft, and compute for you, like an arm or an
  assistant. It does not get to decide what is true or what you claim. Example: let
  it draft a summary paragraph, then you check every number in it yourself.
- **Assistance versus intellectual substitution** is the line you patrol. **Assistance**
  speeds up work you could check and own. **Intellectual substitution** is letting the
  tool make a judgment that was yours to make. Example of the bad kind: pasting an
  AI's conclusion into your poster without deciding, yourself, that the evidence
  supports it.

So the answer to the puzzle. When the machine writes the code and the prose, the
human job that grows is **judgment**: deciding what to ask, what counts as evidence,
what the evidence licenses, and where to stop. That judgment is the course.

> **A question that often comes up here:** *"I'm not a numbers person. Will this
> course bury me in statistics?"* No. It assumes no quantitative or computing
> background. Every piece of code arrives already built. Your job is to run it, read
> what it says, and reason about what it does and does not establish. The thinking is
> the hard part, and the thinking is in plain English.

---

The rule turns into a three-step habit you will use in every notebook:
**Ask → Verify → Document**. **Ask** means use the tool, ambitiously. **Verify**
means retrieve and confirm every fact, source, and number yourself. **Document**
means log the tool, the task, and how you checked, in a record that travels with your
work. One more term before the diagram:

- A **claim** is a statement about the world that could be true or false. Example:
  "this app can tell which users are about to cancel." A claim is not yet knowledge.
  It becomes knowledge only when evidence you can point to supports it.

The loop below is the shape of the whole course. Skip the Verify and Document steps,
and you are not doing research. You are repeating a machine's guess in your own
voice.

```python
# The habit that runs the course, drawn as a loop: Ask -> Verify -> Document.
# (networkx is imported once in the setup cell.)
G = nx.DiGraph()
pos = {"Ask": (0.0, 0.0), "Verify": (2.0, 0.0), "Document": (4.0, 0.0)}
G.add_edges_from([("Ask", "Verify"), ("Verify", "Document")])

fig, ax = plt.subplots(figsize=(9, 3.2))
nx.draw_networkx_nodes(G, pos, node_color="#DCE6F1", edgecolors="#4C72B0",
                       linewidths=1.5, node_size=6000, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=12, ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=22, edge_color="#555555",
                       width=1.8, node_size=6000, ax=ax)
labels = {("Ask", "Verify"): "use the tool", ("Verify", "Document"): "confirm it yourself"}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=9, ax=ax)
# The return arrow: Document feeds the next Ask.
ax.annotate("", xy=(0.0, 0.35), xytext=(4.0, 0.35),
            arrowprops=dict(arrowstyle="-|>", color="#4C72B0",
                            connectionstyle="arc3,rad=-0.25", lw=1.6))
ax.text(2.0, 0.62, "log it, then ask the next question", ha="center",
        fontsize=9, color="#4C72B0")
ax.set_ylim(-0.4, 0.8)
ax.set_title("The habit that runs the course: Ask → Verify → Document")
ax.axis("off")
plt.tight_layout()
plt.show()
print("✓ Ask → Verify → Document loop drawn.")
```

*In plain terms, the diagram says: use the tool, then confirm what it gave you

## 2. Why Sounding Right Is Not Evidence

**Guiding question:** *can you tell a real source from an invented one by how
confident and polished it sounds?*

AI tools **hallucinate**, and you need the word. A **hallucination** is confident,
polished output that is simply made up: a study that was never run, a statistic with
no basis, a journal article that does not exist. Example: ask for scholarly sources
and you may get a mix of real papers and convincing inventions, with nothing marking
which is which. The tone is identical for both.

That is why one skill matters more than any prompt trick: **verification** by
**retrievability**. A source is **retrievable** when you can actually open it: find
the real paper, and confirm the authors, year, and venue match. Example: you paste
the exact title into Google Scholar or the Purdue Libraries catalog, and the real
article opens. If nothing opens, treat the source as invented until it does.

### 🛠️ Run the Study: the retrieval test

Run the cell below. It simulates 12 candidate sources, sets each one's polish and its
real-or-not label **independently**, and then checks whether sorting by polish would
catch the fakes. Read the printout before the next markdown cell.

```python
# Simulate the retrieval test. An AI returns 12 candidate sources for one query.
# 'polish' = how authoritative each SOUNDS (0-1). 'retrievable' = whether it actually
# opens when you search for it. The two are drawn INDEPENDENTLY, which is the point:
# sounding good and being real are unrelated.
n_sources = 12
polish = rng.uniform(0.55, 0.99, size=n_sources)     # everything sounds authoritative
retrievable = rng.random(n_sources) < 0.60           # ~60% actually exist
corr = np.corrcoef(polish, retrievable.astype(float))[0, 1]

n_fake = int((~retrievable).sum())
order = np.argsort(-polish)                           # most confident-sounding first
top_half = order[: n_sources // 2]                    # the 6 you'd 'trust' on tone alone
fake_in_top_half = int((~retrievable[top_half]).sum())

print(f"Sources returned              : {n_sources}")
print(f"Actually retrievable          : {int(retrievable.sum())}")
print(f"Fabricated (do not open)      : {n_fake}")
print(f"Correlation(polish, real)     : {corr:+.2f}")
print(f"Fabricated among the 6 most confident-sounding: {fake_in_top_half}")

# Plot: polish on the x-axis, real-or-not on the y-axis. Shape AND color both encode
# the label, so the picture never depends on color alone.
fig, ax = plt.subplots()
real = retrievable
ax.scatter(polish[real], np.ones(real.sum()), s=130, marker="o",
           color="#4C72B0", label="retrievable (you can open it)")
ax.scatter(polish[~real], np.zeros((~real).sum()), s=170, marker="X",
           color="#DD8452", label="fabricated (nothing opens)")
ax.set_yticks([0, 1])
ax.set_yticklabels(["fabricated", "retrievable"])
ax.set_xlim(0.5, 1.02)
ax.set_ylim(-0.5, 1.5)
ax.set_xlabel("How authoritative the source SOUNDS (polish)")
ax.set_title("Polish does not sort real from fabricated: both spread across the range")
ax.legend(loc="center left")
plt.tight_layout()
plt.show()
```

```python
# Self-check: polish and existence are roughly unrelated, so tone cannot filter fakes.
assert n_sources == 12
assert n_fake >= 1, "the demo needs at least one fabricated source to catch"
assert abs(corr) < 0.6, "polish and existence were drawn independently — corr should be modest"
print(f"✓ Self-check passed: with r = {corr:+.2f} between polish and existence, sorting "
      f"the 12 by how good they sound leaves {fake_in_top_half} fabricated source(s) "
      f"sitting in the 'confident' half.\n  Tone is not a filter. Retrieval is.")
```

## 3. Why Smart People Trust a Wrong Machine

**Guiding question:** *what makes a fluent, confident answer so hard to doubt, and
what discipline keeps you doubting anyway?*

> *"The most dangerous output isn't the one that's obviously wrong. It's the one
> that's confident, detailed, well-organized, and wrong. That one gets used."*
> — a **skeptical peer**, reviewing a draft that leaned too hard on a tool

Three named habits explain the trap. Learn the names, because you cannot guard against
a danger you cannot name.

- **Automation bias** is the tendency to over-trust an automated system and stop
  checking, precisely because it is automated. Example: drivers who follow a GPS
  straight into a closed road because "the machine said turn."
- **False confidence** is when fluent, detailed output makes you feel sure of
  something you never actually verified. Example: a polished paragraph makes a wrong
  statistic feel solid, so you quote it.
- **Illusion of understanding** is mistaking a smooth explanation for real
  comprehension. Example: you nod along to an AI's line-by-line walkthrough of a cell,
  then find you could not defend it to a skeptic.

None of this is new to AI. Long before these tools, Tversky and Kahneman (1974)
showed people lean on confident mental shortcuts that misfire in predictable ways. And
Open Science Collaboration (2015) re-ran 100 published psychology studies and found
that many did not hold up, a reminder that even peer-reviewed, confident-sounding
results need checking. A fluent AI answer simply makes the old trap faster and
smoother.

> **A question that often comes up here:** *"If these tools invent things and mislead,
> why use them at all?"* Because they are genuinely powerful, and you will use them
> for the rest of your career no matter what any course says. A search that once took
> an afternoon now takes minutes. The skill worth building is not avoidance. It is
> using the tool ambitiously and then verifying like a professional, every time.

---

The discipline that beats the trap is a loop with a human at both ends. You met its
shorthand, Ask → Verify → Document. Here is the full version the course trains,
**SDIIVDD**: **Specify, Delegate, Interrogate, Inspect, Verify, Document, Defend**.

You **specify** the task and write down your own expected answer first. You
**delegate** the well-specified work to the tool. You **interrogate** the draft for
assumptions and errors. You **inspect** what it actually produced, not its summary of
itself. You **verify** with a real check before trusting anything. You **document** the
use in your Ledger. You **defend** the result in your own words, with the tool out of
the room.

The two ends are human on purpose. That connects to one more term. **Never-delegate
decisions** are the calls that stay yours no matter how good the tool gets: the
question, the design and population, the measures, the ethics, the claim boundary, the
uncertainty, and the public defense. Example: you may ask Gemini to list candidate
measures for your concept, but deciding which one truly captures it is your call.

```python
# The full loop, drawn as a chain: Specify -> Delegate -> Interrogate -> Inspect ->
# Verify -> Document -> Defend. Specify and Defend are the human bookends.
steps = ["Specify", "Delegate", "Interrogate", "Inspect", "Verify", "Document", "Defend"]
G2 = nx.DiGraph()
G2.add_edges_from([(steps[i], steps[i + 1]) for i in range(len(steps) - 1)])
pos2 = {s: (i, 0) for i, s in enumerate(steps)}

# Human-owned bookends get a darker outline AND are named in the caption (not color
# alone). Verify, where trust is earned, is filled a distinct shade.
fill = ["#B7C9E2", "#DCE6F1", "#DCE6F1", "#DCE6F1", "#55A868", "#DCE6F1", "#B7C9E2"]
edgecol = ["#1F3B6E", "#4C72B0", "#4C72B0", "#4C72B0", "#2F6B45", "#4C72B0", "#1F3B6E"]

fig, ax = plt.subplots(figsize=(12, 3))
nx.draw_networkx_nodes(G2, pos2, node_color=fill, edgecolors=edgecol,
                       linewidths=2.0, node_size=3400, ax=ax)
nx.draw_networkx_labels(G2, pos2, font_size=8.5, ax=ax)
nx.draw_networkx_edges(G2, pos2, arrowstyle="-|>", arrowsize=16, edge_color="#555555",
                       width=1.5, node_size=3400, ax=ax)
ax.set_title("SDIIVDD: the AI-collaboration loop, human at both ends (Specify, Defend)")
ax.set_xlim(-0.6, 6.6)
ax.set_ylim(-0.6, 0.6)
ax.axis("off")
plt.tight_layout()
plt.show()
print("✓ SDIIVDD loop drawn. Shorthand: Ask (Specify + Delegate) → Verify (Interrogate")
print("  + Inspect + Verify) → Document, and Defend closes it. You own the first and last step.")
```

*In plain terms, the chain says: you set the task, hand out the work, argue with the
draft, look at what it really did, confirm it with a real check, log it, and defend it
yourself. The two darker end circles are the steps you never hand to a tool.*

## 4. Verify the Number, Not the Paragraph

### 🛠️ Run the Study: check the claimed number

Run the cell below. It builds a small randomized study with a real built-in effect,

> printed number is the actual effect of the workshop in this toy world, and say

```python
# A small randomized study: does a study-skills workshop raise a quiz score?
# Re-seed a fresh generator so this result is reproducible on its own (seed 464).
rng_study = np.random.default_rng(SEED)

n = 120
TRUE_EFFECT = 3.0                                        # the real effect in this toy world
workshop = rng_study.permutation(np.repeat([1, 0], n // 2))  # random assignment, 60 vs 60
score = rng_study.normal(72, 8, n) + TRUE_EFFECT * workshop
study = pd.DataFrame({"workshop": workshop, "score": score.round(1)})

means = study.groupby("workshop")["score"].mean()
observed_gap = means[1] - means[0]
CLAIMED_GAP = 9.0                                        # the confident one-line summary

print(f"Mean score, no workshop : {means[0]:.2f}")
print(f"Mean score, workshop    : {means[1]:.2f}")
print(f"Verified gap (from data): {observed_gap:+.2f} points")
print(f"Claimed gap (summary)   : {CLAIMED_GAP:+.2f} points")
print(f"The summary overstated the gap by {CLAIMED_GAP - observed_gap:.2f} points.")

# Plot the claimed number next to the verified one. Hue AND the value label both carry
# the meaning, so the picture never depends on color alone. The verified bar is blue;
# the unverified claim gets its own orange so it never reads as the trusted value.
fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(["Verified gap\n(from the data)", "Claimed gap\n(confident summary)"],
              [observed_gap, CLAIMED_GAP],
              color=["#4C72B0", "#DD8452"], edgecolor="white")
for b, v in zip(bars, [observed_gap, CLAIMED_GAP]):
    ax.text(b.get_x() + b.get_width() / 2, v + 0.15, f"{v:+.1f}", ha="center", fontsize=11)
ax.set_ylim(0, 10)
ax.set_ylabel("Effect on quiz score (points)")
ax.set_title("Verify the number: the confident summary is not what the data show")
plt.tight_layout()
plt.show()
```

```python
# Self-check: random assignment licenses a causal gap here, but it is far below 9.
assert observed_gap > 0, "the built-in effect is positive; the observed gap should be too"
assert observed_gap < CLAIMED_GAP - 2, "the claim should clearly overstate the data"
print(f"✓ Self-check passed: the verified gap is {observed_gap:+.2f} points, well below the "
      f"confident '{CLAIMED_GAP:+.0f} points'.\n  Random assignment lets this gap estimate "
      f"the workshop's effect in the toy world, but the claimed number was never in the data.")
```
