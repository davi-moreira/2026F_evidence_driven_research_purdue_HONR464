## 🎯 Definition of Done

You are done with this sprint when your submission has all three pieces built from

**The final research chapter:**

1. **The whole arc, finalized.** Your question and its **inquiry declaration** (the

**The AI-management portfolio (the piece this week built):**

**The Evidence Defense (live at the final-defenses studio):**

4. **Ten minutes in three movements** (claims, choices, verification), one graceful

Plus the **AI Research Ledger** rows and the **dossier update line**. At Friday's

> **A question that often comes up here:** *"My research note already argued the
> finding. Isn't the chapter just the note with a portfolio stapled on?"* No. The note
> argued one finding to a reader who had the page. The chapter argues the whole arc,
> and the portfolio proves you stayed in charge of the tools that helped you write it.
> A reader cannot tell a verified claim from a lucky guess by reading the claim. The
> portfolio and ledger are where that check lives.

```python
# Setup + portfolio self-check — run this cell first, then edit the four inputs.
import numpy as np

SEED = 464  # course number — keeps this notebook's Monte Carlo reproducible
rng = np.random.default_rng(SEED)

# --- EDIT ME (1) — your AI review team, as you ACTUALLY ran it this semester.
n_roles = 4            # how many AI roles read your chapter or a key claim
share_one_framing = True   # True if they share one base model AND one prompt framing
p_catch = 0.5          # honest guess: chance ONE role catches a real flaw on its own
rho_blind = 0.40       # chance the team shares a blind spot (set 0 only if truly independent)

# --- EDIT ME (2) — your full-semester AI Research Ledger, in two counts.
ledger_rows_total = 12   # total AI-assisted rows across the whole semester
escalation_rows = 1      # rows where YOU overrode an AI output (a human override)

# ---- everything below is DERIVED from the four edits above ----
# One source of truth per number: change an input once, the whole readout follows.

# CHECK 1 — the independence check. Simulate 5000 chapters, each with one hidden flaw.
# A "false consensus" = all your roles say "fine" while a real flaw is present.
n_ch = 5000
caught = (rng.random((n_ch, n_roles)) < p_catch)
if share_one_framing:
    blind = rng.random(n_ch) < rho_blind        # shared blind spot: all roles miss together
    caught = caught & (~blind[:, None])
false_consensus_rate = (~caught.any(axis=1)).mean()

print("CHECK 1  independence of your AI team")
print(f"         roles = {n_roles}, share one framing = {share_one_framing}")
print(f"         chance a unanimous 'looks fine' is actually a MISSED flaw: "
      f"{false_consensus_rate:.1%}")
if share_one_framing and false_consensus_rate > 0.15:
    print("         -> high false-consensus risk. Agreement here is close to ONE voice,")
    print("            not several. Your portfolio needs an independent check you ran,")
    print("            or two roles you made genuinely independent, not the head-count.")
else:
    print("         -> lower risk, but 'lower' is not 'zero'. Name the independent")
    print("            check behind any agreement you lean on in the chapter.")

# CHECK 2 — the escalation floor. A semester of AI use with zero overrides is a flag.
print("\nCHECK 2  escalation floor")
print(f"         ledger rows = {ledger_rows_total}, overrides logged = {escalation_rows}")
if escalation_rows == 0:
    print("         -> ZERO overrides. Per the brief this is a warning sign, not a badge:")
    print("            either AI touched nothing that mattered, or a boundary slipped by.")
    print("            Find the escalation your ledger is missing and log it before you defend.")
else:
    print(f"         -> {escalation_rows} override(s) logged. Make sure at least one carries")
    print("            a NON-AI verification in its row. That is what the portfolio is read for.")

print("\nBoth are FLOORS, never proof: a clean readout means your portfolio is")
print("well-formed, not that your claim is right. The Friday cross-examination is the proof.")
```

### 🤝 AI Research Partner

AI is your arm and your research assistant in this final sprint, not your brain. This
week the skill is different from every week before it: you are not tasking one
assistant, you are **managing a small team** and deciding for yourself when its roles
disagree. More roles do not mean more truth. Three roles that quietly share one wrong
assumption will agree with great confidence and walk you off a cliff.

**Good to delegate here:**
- Proposing a **decomposition** of your chapter into scoped roles, so you can decide
  which are worth the hand-off and which you can supervise.
- Red-teaming a paragraph **you** already wrote, or rehearsing the hardest
  cross-examination questions, so you can prepare an answer from a real ledger row.
- Ranking your limitations or listing the inputs a cold replicator would need, so you
  can check the ranking against your own printout.

**Never delegate (these are yours to decide, no matter how good the roles get):**
- Which claims your chapter can defend and where each claim must stop.
- Which sources are real and the words you put your name on.
- Every **human override**: when an AI output would *settle* rather than *inform* a
  decision that carries your name, you take the pen back. The public defense is yours.

**Commit first, then ask.** Write your own answer before you open any role, and commit
your chapter to git before a long multi-role session, so you can compare what you wrote
against what the roles changed. Verify what comes back, and log it in the **AI Research
Ledger** below. The full never-delegate list is in

### 🗡️ Red-Team Exchange

The final Friday studio's **29–41 minute block**. Pair with one partner and run two
rounds, with an AI reviewer as a third voice. The role for this milestone is the **AI
Research Team Orchestrator**, which helps you plan how you split your chapter across
roles and how you logged their conflicts (quick card in
[`ai_resources/agent_role_cards.md`](https://github.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/blob/main/ai_resources/agent_role_cards.md);
full briefing in
[`genai_studio/roles/ai_research_team_orchestrator.md`](https://github.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/blob/main/genai_studio/roles/ai_research_team_orchestrator.md)).
It is a planning aid: it proposes a decomposition and a way to log conflicts, and it
hands every consequential decision back to you. It cannot resolve a disagreement, and
it never runs your capsule.

**First, the Orchestrator pass (before the rounds).** Paste your portfolio's
decomposition and one conflict-and-override entry. It returns whether your roles were
wired so an early error could flow downstream unchecked, and whether your logged
agreement could have been correlated by construction. Every flag is a **suspect you
test yourself** against your ledger, not a verdict.

**Round 1: cross-examine the defense (about 15 minutes).** Your partner plays a
hostile examiner. They take your headline claim and press its boundary: "show me the
printout that says this, not a bigger version of it." You run your three movements,
claims then choices then verification, and they force one **graceful concession**. You
concede precisely and name the smaller true claim, or you defend from a ledger row.
Watch for the three forbidden moves in your own answers: inventing a new claim,
un-conceding by insistence, and blaming the AI.

**Round 2: rerun the capsule cold (about 15 minutes).** Your partner opens your
chapter's notebook on a fresh kernel and runs **restart-and-run-all**, with no help
from you. They log whether your headline number comes back and where, if anywhere, it
broke. A number that reruns for you and not for them is a finding you fix before Friday
night, because M15 is terminal and there is no Monday.

**Mind the correlated-error trap.** If Gemini rehearsed your questions and the
Orchestrator calls your wiring sound, that agreement is not proof your claim holds.
Neither tool ran your capsule. Only your partner's cold rerun and your own verified
ledger settle whether the number reproduces and the claim stands.

**What to log.** Write down the concession the cross-examination forced, the strongest
question you could not answer from a ledger row (and the row you will build), and
whether the cold rerun reproduced your headline number. A challenge that changed
nothing still earns a line saying why it did not land. That record feeds your portfolio.

> **A question that often comes up here:** *"What if my partner and the Orchestrator
> disagree about whether my claim is bounded?"* Your evidence settles it, not either of
> them. The Orchestrator read only your plan and can miss a naked claim; your partner
> read your page but not your full ledger. Say in one sentence which read your own
> printout and claim ledger support, and why. That adjudication is stronger proof you
> understand your chapter than agreeing with either would be.
