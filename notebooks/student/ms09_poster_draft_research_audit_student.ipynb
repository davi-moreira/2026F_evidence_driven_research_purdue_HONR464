```python
# Pre-listed robustness grid — edit the EDIT ME block to point at YOUR M8 estimate.
# It computes ONE headline gap under a SMALL grid of specifications you commit to
# BEFORE looking, then prints the direction and the range. Every label flows from a
# single source of truth, so renaming a column in ONE place updates the whole grid;
# there is no second hardcoded name to fall out of sync.

# --- Single source of truth: name your columns and labels ONCE. -----------
GROUP_COL   = "trust_group"   # EDIT ME — column that splits high vs low on your key variable
OUTCOME_COL = "support"       # EDIT ME — the number your headline is about
HIGH        = "high"          # EDIT ME — label for the high group
LOW         = "low"           # EDIT ME — label for the low group

# --- PLACEHOLDER data: replace this whole block with YOUR M8 data. ---------
# A seeded stand-in so the grid runs before your real data are wired in. Swap it
# for load_course_data(...) or your own CSV, keeping the names above. This
# placeholder is NOT evidence about anything; it only proves the pipeline runs.
n_per = 400                                   # EDIT ME — your per-group size
df = pd.DataFrame({
    GROUP_COL:   [HIGH] * n_per + [LOW] * n_per,
    OUTCOME_COL: np.concatenate([rng.normal( 0.35, 1.0, n_per),
                                 rng.normal(-0.35, 1.0, n_per)]),
    "ideology_extreme": rng.random(2 * n_per) < 0.15,   # a subgroup you may drop
})
# --------------------------------------------------------------------------

# --- The headline: a gap in mean OUTCOME between the two groups. -----------
def headline_gap(data):
    hi = data.loc[data[GROUP_COL] == HIGH, OUTCOME_COL]
    lo = data.loc[data[GROUP_COL] == LOW,  OUTCOME_COL]
    return hi.mean() - lo.mean()

# --- Your PRE-LISTED grid: commit this list BEFORE reading any estimate. ---
# Reporting only the flattering corner is a specification search. Reporting the
# whole grid is robustness. Same code, opposite integrity.
q_lo, q_hi = df[OUTCOME_COL].quantile(0.02), df[OUTCOME_COL].quantile(0.98)
specs = {
    "all rows":               df,
    "drop ideology extremes": df.loc[~df["ideology_extreme"]],
    "trim outer 2%":          df.loc[df[OUTCOME_COL].between(q_lo, q_hi)],
}
gaps = {name: headline_gap(data) for name, data in specs.items()}

# --- One placebo: shuffle the group label; a real gap should collapse. -----
placebo_df = df.copy()
placebo_df[GROUP_COL] = rng.permutation(placebo_df[GROUP_COL].to_numpy())
placebo_gap = headline_gap(placebo_df)

for name, g in gaps.items():
    print(f"  {name:24s}: gap = {g:+.3f}")
lo_g, hi_g = min(gaps.values()), max(gaps.values())
print(f"\nDirection + range across the grid: {lo_g:+.3f} to {hi_g:+.3f}")
print(f"Placebo (shuffled label) gap:      {placebo_gap:+.3f}   (should sit near 0)")
print("\n✓ Ran on PLACEHOLDER data — swap in your M8 estimate before you report anything.")
```
