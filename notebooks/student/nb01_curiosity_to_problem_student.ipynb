## 1. Why a Question Comes First

**Guiding question:** *how does a curiosity become a research problem someone
would fund a term of your life to study?*

> *"I can tell within one paragraph whether a researcher knows what question they
> are asking. If they do not, nothing downstream saves the project. Not the data,
> not the method, not the writing."*
> — a **skeptical peer** you will one day be grateful for

Everything in research sits downstream of the question. A vague question makes
every later choice arbitrary. Any data will do, any method will do, and any answer
will do, because nothing was pinned down enough to be wrong. A sharp question does
the opposite. It tells you what evidence you need and what would count as being
mistaken. That is why this notebook comes before any tool does.

The path has four stops, and you will walk it with your own project:

**curiosity** *(I wonder about ...)* → **topic** *(the studyable territory)* →
**research problem** *(the specific unknown worth solving)* → **question** *(the
sentence evidence can answer)*.

> **A question that often comes up here:** *"My curiosity feels too big for this.
> I care about something huge, like inequality or mental health. Do I have to
> shrink it?"* You narrow the QUESTION, not the caring. The huge thing stays as
> your motivation. The question is the part small enough to answer this term.
> Answering it is how you earn the right to say something about the huge thing.

---

## 2. The Funnel, One Stop at a Time

Each stop on the funnel adds something concrete. Here are the four, each as a term
you will reuse all term: a **bold name**, a one-sentence meaning, and a concrete
example.

- **Curiosity**: a felt interest with no edges yet. Example: "why do some study
  groups click and others fall apart?" You cannot collect evidence on it, because
  it does not say who or what to look at.
- **Topic**: a curiosity with edges, a territory someone could gather evidence
  inside. Example: "coordination in undergraduate project teams." Now you can name
  who it is about.
- **Research problem**: a specific unknown inside that territory whose absence
  matters. Example: "we do not know whether team size changes how evenly work gets
  shared." A problem points at one missing piece, not a whole subject.
- **Question**: the single sentence evidence can answer, and be wrong about.
  Example: "in four-person versus seven-person course teams, is peer-rated
  workload more evenly distributed?"

The picture below shows the funnel narrowing from top to bottom. Read it as a
filter: each stop throws away vagueness the stop above it still allowed.

```python
# The funnel: a curiosity narrows, stop by stop, into an answerable question.
stops = ["CURIOSITY\n(I wonder about ...)",
         "TOPIC\n(the studyable territory)",
         "RESEARCH PROBLEM\n(the specific unknown that matters)",
         "QUESTION\n(the sentence evidence can answer)"]
widths = [1.00, 0.76, 0.52, 0.30]
fills  = ["#EAF1F8", "#D4E2F0", "#BFD3E8", "#A9C4E0"]  # light enough for black text

fig, ax = plt.subplots(figsize=(8, 5))
for i, (label, w, c) in enumerate(zip(stops, widths, fills)):
    y = 3 - i
    ax.barh(y, w, height=0.82, left=(1 - w) / 2, color=c, edgecolor="#33506E", linewidth=1.4)
    ax.text(0.5, y, label, ha="center", va="center", fontsize=9)
ax.set_xlim(0, 1)
ax.set_ylim(-0.6, 3.6)
ax.axis("off")
ax.set_title("The funnel: a curiosity narrows into a question evidence can answer")
plt.tight_layout()
plt.show()

print("✓ Funnel drawn. Top = widest and vaguest; bottom = narrowest and answerable.")
print("  Each step down throws away vagueness the step above it still allowed.")
```

*In plain terms, the picture says the same idea gets narrower and more concrete
as you move down. The top band is a feeling; the bottom band is a sentence you
could be wrong about.*

Three more terms name what a good question must quietly carry. A question that
skips any of them cannot be answered, because you would not know what to record.

- **Units of analysis**: the who or what each row of your evidence stands for.
  Example: one course team, one respondent, one game, one river-sampling station.
  If you cannot say what one unit is, you cannot count anything.
- **Target population**: the full group your question is really about, beyond the
  handful you can reach. Example: all first-year engineering majors at Purdue, not
  only the twenty in your section.
- **Outcome** (also called the **target**): the observable thing that varies, the
  quantity you will actually record. Example: a quiz score, a turnout rate, an
  evenness-of-workload rating. It must be measurable, even if imperfectly.

> **A question that often comes up here:** *"Is naming units really that important?
> It feels obvious."* It is the most common silent failure. "Does social media
> hurt teens?" hides its units: teens? posts? platforms? days? Each choice is a
> different study with a different answer. Naming the unit out loud is how you
> notice you were about to run four studies at once.

A colleague hands you what they call one research question: **"How common is
burnout among nurses, and does working night shifts cause it?"** They are proud of
it. It sounds thorough.

Here is the puzzle. This single sentence is secretly asking for two completely
different kinds of answer, and no one design can deliver both at once. Find the
seam. Where does the sentence split, and what is each half asking for? One half
wants a headcount of something that already exists. The other wants to know what
would change if you intervened. Commit your split on paper before we go further.
Telling those two kinds apart, and refusing to let them ride in one sentence, is
the whole job of Lecture 2.

## 3. The Inquiry Compass: Kind and Reach

**Guiding question:** *what KIND of answer does your question ask for, and how far
is it trying to reach?*

> *"Before I approve a single method, I make my advisees answer two questions
> about their question. What kind of answer does it want, and for whom must that
> answer hold? Get those two wrong and every method you pick after is aimed at the
> wrong target."*
> — a **thesis advisor** at your first committee meeting

The **inquiry compass** is the course's map for classifying a research question. It
asks two things, and only two. Together they place any question at one of four
named positions, and that position fixes what its answer may claim.

**Axis 1, KIND. Is the question descriptive or causal?**

- **Descriptive question**: asks what the world is or was, no intervention needed.
  Example: "what share of first-years skip breakfast?" You just look and count.
- **Causal question**: asks what would differ if someone intervened, which needs a
  comparison to a world that did not happen. Example: "does a free breakfast raise
  attendance?" You have to imagine the same people without the breakfast.

**Axis 2, REACH. For which units must the answer hold?**

- **Data at hand**: only the units you actually observed. Example: "in these 10,000
  interviews, the military rates highest."
- **A population**: a larger group beyond your data that your sampling reaches.
  Example: "across all adults in the country, trust in the military is highest."
- **Unseen cases**: units not yet observed, usually future ones. Example: "the next
  applicant we have not met yet will re-enroll."

---

Cross the two axes and you get the four positions. The picture below is the whole
compass on one grid: **kind** down the side, **reach** across the top.

- **Description** (descriptive, data at hand): reports what is there in the units
  you observed. Example: "in our sample, 30% skipped breakfast."
- **Generalization** (descriptive, a population): estimates a population quantity
  with stated uncertainty. Example: "about 30% of all first-years skip breakfast,
  give or take a margin." It reaches past the data, so it must pay for that reach
  with a sampling design.
- **Prediction** (descriptive, unseen cases): forecasts new cases against a stated
  baseline. Example: "this incoming class member will likely skip breakfast."
  **Prediction is still a descriptive question**, just aimed at cases you have not
  seen, not a causal one. Guessing who will drop out is not the same as knowing
  what makes them drop out.
- **Causal reasoning** (causal kind): asks what an intervention would change.
  Example: "would free breakfast raise attendance?" This is the only position that
  earns the word "because," and only with a design that isolates the cause.

> **A question that often comes up here:** *"Prediction feels causal to me. If a
> model predicts dropout from attendance, doesn't attendance cause dropout?"* No,
> and this is the trap the compass exists to catch. A predictor only needs a
> pattern that travels to new cases. It can predict beautifully using a feature
> that causes nothing, as long as that feature rides along with the outcome.
> Performance is not understanding. Prediction sits in the descriptive half for
> exactly this reason.

```python
# The inquiry compass on one grid: KIND (rows) x REACH (columns).
fig, ax = plt.subplots(figsize=(9.5, 4.7))
reaches = ["Reach:\ndata at hand", "Reach:\na population", "Reach:\nunseen cases"]
desc_cells = {0: "Description", 1: "Generalization", 2: "Prediction"}

for cx in range(3):
    ax.add_patch(mpatches.Rectangle((cx, 1), 1, 1, facecolor="#EAF1F8",
                                    edgecolor="#33506E", linewidth=1.3))
    ax.text(cx + 0.5, 1.5, desc_cells[cx], ha="center", va="center",
            fontsize=11, fontweight="bold")

# Causal kind: one position, drawn across the reaches it can hold.
ax.add_patch(mpatches.Rectangle((0, 0), 2, 1, facecolor="#F2E4D3",
                                edgecolor="#33506E", linewidth=1.3))
ax.text(1.0, 0.5, "Causal reasoning", ha="center", va="center",
        fontsize=11, fontweight="bold")
ax.add_patch(mpatches.Rectangle((2, 0), 1, 1, facecolor="#E8E8E8",
                                edgecolor="#BBBBBB", linewidth=1.0))
ax.text(2.5, 0.5, "(rare)", ha="center", va="center", fontsize=9, color="#777777")

ax.text(-0.12, 1.5, "Descriptive", ha="right", va="center", fontsize=11, rotation=90)
ax.text(-0.12, 0.5, "Causal", ha="right", va="center", fontsize=11, rotation=90)
for cx in range(3):
    ax.text(cx + 0.5, 2.14, reaches[cx], ha="center", va="center", fontsize=9)

ax.set_xlim(-0.65, 3.05)
ax.set_ylim(-0.15, 2.4)
ax.axis("off")
ax.set_title("The inquiry compass: KIND (rows) x REACH (columns)")
plt.tight_layout()
plt.show()

print("✓ Compass drawn. Read any question by asking two things:")
print("  KIND  -> descriptive or causal?   REACH -> data at hand, a population, or unseen?")
```

*In plain terms, the grid says every research question lands in exactly one box.
The top row is the descriptive half, split by how far it reaches. The bottom row
is the causal half, which is defined by the kind of answer it wants, not by reach.*

Two more tools finish the skill.

- **Giveaway words**: the words that hint at a position, useful but not to be
  trusted. Example: "cause," "effect," "raise" lean causal; "predict," "will,"
  "next" lean prediction; "all," "population" lean generalization. They are hints,
  not proof, because a question can wear a word it does not mean.
- **Double-barreled question**: one sentence that secretly asks two questions of
  different kinds. Example: "how common is burnout, and does night work cause it?"
  It cannot be answered as one, and the fix is always to split it into two clean
  questions, each with its own compass position.

The trap you will watch happen next is that sorting by giveaway words alone, which
is exactly what a hurried tool does, gets fooled. The honest way to classify is to
ask what answer the question actually wants, not which words it happens to carry.

### 🛠️ Run the Study: the compass on a deck of questions

Run the cell below. It deals the eight-question deck in a reproducible shuffled
order (seed 464), sorts each question by giveaway words, and compares that guess to
the true compass position. Read the mismatches before the next markdown cell.

**🔴 Live in class: we run this one together.**
**Before you ask:** commit your own position for question 5 and question 7 on paper
first, so you can tell whether Gemini reasons or just pattern-matches like the
keyword sorter.

> 💡 **Gemini Prompt:** "Here are eight research questions [paste the eight
> question texts from the next cell]. For each, tell me its KIND (descriptive or
> causal) and its REACH (data at hand, a population, or unseen cases), and name the
> single word that most influenced you. Put it in a table. Then, for any question
> whose answer would only need a pattern and not an intervention, say so
> explicitly."
>
> **After running, verify (counters *silent scope change*: a partner that quietly
> upgrades a descriptive question to a causal one has answered a different question
> than you asked):**
> - [ ] Compare Gemini's table to the "true position" column your cell printed.
>   Every disagreement is a question to reread against the two definitions, because
>   one of you is wrong.
> - [ ] Flag any question Gemini calls causal that never mentions an intervention.
>   Check it against your printout, and if the design only looks at what already
>   is, it is descriptive.
> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# A deck of eight sample questions with their reasoned-out true positions.
deck = pd.DataFrame([
    (1, "In our file of 10,000 interviews, what share of respondents rate trust in the police above 4?", "Description"),
    (2, "What percentage of all registered voters in the state would support ranked-choice voting?",      "Generalization"),
    (3, "Which of this year's applicants will still be enrolled three years from now?",                   "Prediction"),
    (4, "Does giving first-year majors a free breakfast raise their exam scores?",                        "Causal reasoning"),
    (5, "What share of campus buildings have effective ventilation during class hours?",                  "Description"),
    (6, "How many first-years skipped breakfast, and does skipping it lower their exam scores?",          "Double-barreled"),
    (7, "Do most adults across the country trust scientists more than elected officials?",                "Generalization"),
    (8, "Which of next year's incoming first-years will leave before their second year?",                 "Prediction"),
], columns=["id", "question", "true_position"])

# A naive keyword sorter: it only scans for giveaway words, in a fixed priority.
CAUSAL_WORDS  = ["cause", "effect", "affect", "impact", "raise", "lower", "reduce", "increase", "improve", "because"]
PREDICT_WORDS = ["will ", "predict", "forecast", " next ", "from now", "going to"]
POP_WORDS     = ["all registered", "percentage of all", "every citizen"]

def keyword_sort(text):
    t = text.lower()
    if any(w in t for w in CAUSAL_WORDS):
        return "Causal reasoning"
    if any(w in t for w in PREDICT_WORDS):
        return "Prediction"
    if any(w in t for w in POP_WORDS):
        return "Generalization"
    return "Description"

# Deal the deck in a reproducible shuffled order (seed 464), then classify.
order = rng.permutation(deck.index.to_numpy())
deck = deck.loc[order].reset_index(drop=True)
deck["keyword_guess"] = deck["question"].apply(keyword_sort)
deck["match"] = np.where(deck["keyword_guess"] == deck["true_position"], "ok", "MISS")

print("Dealt order (seed 464). Keyword sorter vs the true position:")
print(deck[["id", "true_position", "keyword_guess", "match"]].to_string(index=False))

n_ok = (deck["match"] == "ok").sum()
missed = sorted(deck.loc[deck["match"] == "MISS", "id"].tolist())
print(f"\nKeyword sorter got {n_ok} of {len(deck)} right ({n_ok / len(deck):.1%}).")
print(f"It missed question ids: {missed}")
```

```python
# Self-check: the keyword sorter's failures are the lesson, so pin them down.
assert n_ok == 5, "keyword sorter should get exactly 5 of 8"
assert missed == [5, 6, 7], "the misses should be questions 5, 6, and 7"
print("✓ Self-check passed: 5 of 8 correct; the three misses are 5, 6, and 7.")
print("  Q5 carries 'effective' (the sorter sees 'effect' and cries causal).")
print("  Q6 is double-barreled (the sorter grabs one half and drops the other).")
print("  Q7 is a population question with no keyword the brittle list knows.")
```

```python
# Plot each single-position question on the compass, marking the sorter's hits and misses.
xy = {"Description": (0.5, 1.5), "Generalization": (1.5, 1.5),
      "Prediction": (2.5, 1.5), "Causal reasoning": (1.0, 0.5)}

fig, ax = plt.subplots(figsize=(9.5, 4.7))
for cx in range(3):
    ax.add_patch(mpatches.Rectangle((cx, 1), 1, 1, facecolor="#F4F8FC",
                                    edgecolor="#9BB6D2", linewidth=1.0))
ax.add_patch(mpatches.Rectangle((0, 0), 2, 1, facecolor="#FBF3E9",
                                edgecolor="#9BB6D2", linewidth=1.0))
for name, (x, y) in xy.items():
    ax.text(x, y + 0.34, name, ha="center", va="center", fontsize=9, color="#33506E")

single = deck[deck["true_position"] != "Double-barreled"].copy()
for pos, grp in single.groupby("true_position"):
    x0, y0 = xy[pos]
    xs = np.linspace(-0.22, 0.22, len(grp)) if len(grp) > 1 else [0.0]
    for dx, (_, r) in zip(xs, grp.iterrows()):
        hit = r["match"] == "ok"
        ax.scatter(x0 + dx, y0 - 0.05, s=230,
                   marker="o" if hit else "X",
                   color="#4C72B0" if hit else "#DD8452",
                   edgecolors="black", linewidths=0.8, zorder=3)
        ax.text(x0 + dx, y0 - 0.05, str(r["id"]), ha="center", va="center",
                fontsize=8, color="white", zorder=4, fontweight="bold")

ax.scatter([], [], marker="o", s=120, color="#4C72B0", edgecolors="black", label="sorter got it right")
ax.scatter([], [], marker="X", s=120, color="#DD8452", edgecolors="black", label="sorter missed")
ax.text(2.5, 0.5, "Q6 fuses two boxes\n(needs a split)", ha="center", va="center",
        fontsize=8.5, color="#8a5a2b", style="italic")
ax.set_xlim(-0.15, 3.05)
ax.set_ylim(-0.1, 2.15)
ax.axis("off")
ax.set_title("Where the keyword sorter lands each question (circle = right, X = wrong)")
ax.legend(loc="lower left", fontsize=8)
plt.tight_layout()
plt.show()

print("✓ Plotted. Shape carries the result: a circle is a hit, an X is a miss.")
```

*In plain terms, the picture pins each question to its true box on the compass.
A circle means the word-scanner landed it correctly; an X means it missed. The
two X marks (questions 5 and 7) plus the off-grid note for question 6 are the
whole lesson: surface words are not the position.*

The sorter choked on question 6 because it fused two kinds. The repair is always

into a descriptive half and a causal half.

descriptive half and one causal half, before Gemini offers one.

> things: [paste one double-barreled question, yours or the burnout one]. Split it
> into exactly two clean questions. For each half, name its KIND and REACH and the
> word that gives it away. Do not add any new variable that was not in my original
> sentence."
>
> **After running, verify (counters *silent scope change*: a split can quietly drop
> one half or smuggle in a variable you never asked about):**
> - [ ] Confirm both halves together still cover your whole original sentence, and
>   that neither half added a unit, outcome, or population you did not write.
> - [ ] Check each half's assigned position against the compass grid above. A half
>   with no intervention is descriptive, whatever word it carries.
> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# A double-barreled question comes apart into two clean, single-kind questions.
G = nx.DiGraph()
pos = {"fused": (0.0, 1.0), "descr": (2.0, 1.75), "causal": (2.0, 0.25)}
G.add_edges_from([("fused", "descr"), ("fused", "causal")])
labels = {
    "fused":  "How common is burnout,\nand does night work cause it?\n(two questions in one)",
    "descr":  "How common is burnout?\n(Description)",
    "causal": "Does night work\ncause burnout?\n(Causal reasoning)",
}
colors = {"fused": "#E8E8E8", "descr": "#EAF1F8", "causal": "#FBF3E9"}

fig, ax = plt.subplots(figsize=(9.5, 4.3))
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=20, edge_color="#555555",
                       width=1.6, node_size=15000, ax=ax)
nx.draw_networkx_nodes(G, pos, nodelist=list(labels),
                       node_color=[colors[n] for n in labels],
                       edgecolors="#33506E", linewidths=1.4, node_size=15000,
                       node_shape="s", ax=ax)
nx.draw_networkx_labels(G, pos, labels=labels, font_size=8.5, ax=ax)
ax.set_title("Split a double-barreled question into one descriptive and one causal question")
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ Split drawn. One fused sentence becomes two clean questions,")
print("  each with its own single compass position. Now each half can be answered.")
```

*In plain terms, the diagram shows one fused sentence branching into two arrows,
one to a descriptive question and one to a causal question. Neither box can hold
both, which is why the fused sentence had to be split before it could be answered.*
