## 3. How Do We Know? The Four Kinds of Answers

Every research question that evidence can answer is asking for one of four
kinds of answers. You will learn to tell them apart the way a musician tells
keys apart — instantly, and with consequences. Here is your first look:

| Approach | The question it asks | Example |
|---|---|---|
| **Description** | What is observed here? | *How many hours do first-years in this survey sleep?* |
| **Statistical inference** | What holds beyond this sample, with how much uncertainty? | *How many hours do Purdue first-years sleep?* |
| **Predictive modeling** | Can we forecast cases we haven't seen? | *Which admitted class will struggle in year one?* |
| **Causal reasoning** | What would change if we intervened? | *Would a later start time increase sleep?* |

Those four example questions sound nearly identical. They are not. They demand
different evidence, license different claims, and fail in different ways —
and mixing them up is the single most common way smart people fool themselves
and their readers.

> **A question that often comes up here:** *"Do I have to pick one approach and
> stay in it all semester?"* You will declare a primary approach for your
> project, because your question demands one — but the point is bigger than
> your project. The classification skill is a permanent lens: after this course
> you will hear the difference between "people who exercise sleep better" and
> "exercise improves sleep" for the rest of your life.

```python
# The tiny coffee "study" — run AFTER writing your prediction above.
n = 40
coffee = rng.permutation(np.repeat([1, 0], n // 2))     # who got coffee (random)
memory_score = rng.normal(70, 6, n) + 8 * coffee         # in this toy world, coffee adds 8 points

study = pd.DataFrame({"coffee": coffee, "score": memory_score.round(1)})
group_means = study.groupby("coffee")["score"].mean().round(2)
print("Mean score, no coffee :", group_means[0])
print("Mean score, coffee    :", group_means[1])
print("Difference            :", round(group_means[1] - group_means[0], 2))
```

## 4. The Rule That Runs the Course: Ask → Verify → Document

You will use AI tools constantly this semester — to search, draft, critique,
and explain. The course encourages it. The course also has one iron rule about it:

> **AI MAY PROPOSE. THE RESEARCHER MUST VERIFY. THE EVIDENCE MUST SUPPORT.**

AI assistants fabricate. Not occasionally — routinely, and most confidently
exactly where you know least. They invent studies, misstate findings, and cite
journals that have never existed. A researcher who repeats an AI's claim
without verifying it has not made an error; they have made someone else's
error, in their own name, with their credibility attached.

So every AI use in this course follows three steps: **Ask** (use the tool,
ambitiously), **Verify** (retrieve and confirm every fact and source yourself),
**Document** (log the tool, the task, and how you verified — your AI-use
ledger, which travels with every deliverable this semester).

Time to practice catching a fabrication red-handed.
