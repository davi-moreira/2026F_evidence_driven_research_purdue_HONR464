## 🧭 Inquiry & Claim Boundary

**Inquiry emphasis:** all positions (framing). This notebook opens the map you
will spend the semester learning to read: the **inquiry compass**, a simple
grid for sorting research questions by what they ask and how far their answers
must travel. One axis is the question's **kind**. A **descriptive** question
asks what the world looks like (*how much sleep do first-years get?*). A
**causal** question asks what would change if we acted (*would a later start
time increase sleep?*). The other axis is the question's **reach**: does the
answer cover only the data in front of you, a wider group the data came from,
or cases nobody has observed yet? Crossing kind with reach names the four
positions you will meet today — **description**, **generalization**,
**prediction**, and **causal reasoning**.

| | |
|---|---|
| **A claim this topic PERMITS** | "I can state what evidence WOULD be needed to support this claim." |
| **A claim this topic does NOT permit** | Any empirical claim without a stated evidence base, including claims an AI tool produced for you. |

## 4. How Do We Know? The Inquiry Compass

Start with a scenario. Your roommate announces, "people who nap crush their
exams." Before you can weigh that, you need to know what kind of statement it
is. Is it about the five friends she happens to know? About all undergraduates
everywhere? About what napping would do to *your* grade if you started
tomorrow? Each version needs different evidence.

The course's book sorts questions like these with the two compass moves you
just met. First, name the **kind**: descriptive (what does the world look
like?) or causal (what would change if we acted?). Second, name the **reach**:
the data at hand, a **population** beyond it (the larger group your data came
from), or cases not yet seen. Kind crossed with reach gives the four named
positions, your first look at the semester's spine:

| Compass position (kind · reach) | The question it asks | Example |
|---|---|---|
| **Description** (descriptive · data at hand) | What is observed here? | *How many hours do the first-years in this survey sleep?* |
| **Generalization** (descriptive · a population) | What holds beyond this sample, with how much uncertainty? | *How many hours do Purdue first-years sleep?* |
| **Prediction** (descriptive · unseen cases) | Can we forecast cases we haven't seen? | *Which admitted class will struggle in year one?* |
| **Causal reasoning** (causal) | What would change if we intervened? | *Would a later start time increase sleep?* |

Those four example questions sound nearly identical. They are not. They demand
different evidence, license different claims, and fail in different ways.
Mixing them up is the single most common way smart people fool themselves and
their readers.

> **A question that often comes up here:** *"Do I have to pick one compass
> position and stay in it all semester?"* You will declare a primary position
> for your project, because your question demands one. But the point is bigger
> than your project. The classification skill is a permanent lens. After this
> course you will hear the difference between "people who exercise sleep
> better" and "exercise improves sleep" for the rest of your life.

```python
# The tiny coffee "study" — run AFTER writing your prediction above.
n = 40
coffee = rng.permutation(np.repeat([1, 0], n // 2))     # who got coffee (random)
memory_score = rng.normal(70, 6, n) + 8 * coffee         # in this toy world, coffee adds 8 points

study = pd.DataFrame({"coffee": coffee, "score": memory_score.round(1)})
group_means = study.groupby("coffee")["score"].mean().round(2)
print("Mean score, no coffee :", group_means[0])
print("Mean score, coffee    :", group_means[1])
print("Difference            :", round(group_means[1] - group_means[0], 2))
```

## 5. The Rule That Runs the Course: Ask → Verify → Document

You will use AI tools constantly this semester: to search, draft, critique,
and explain. The course encourages it. The course also has one iron rule
about it:

> **AI MAY PROPOSE. THE RESEARCHER MUST VERIFY. THE EVIDENCE MUST SUPPORT.**

Why so strict? Because AI assistants **hallucinate**: they sometimes produce
confident, polished, entirely made-up material, such as a study that was never
run or a journal that has never existed. Ask one for scholarly sources and you
may receive a mix of real papers and convincing inventions, with nothing
marking which is which. A researcher who repeats an AI's claim without
checking it has not made an error. They have made someone else's error, in
their own name, with their credibility attached.

So every AI use in this course follows three steps. **Ask**: use the tool,
ambitiously. **Verify**: retrieve and confirm every fact and source yourself.
**Document**: log the tool, the task, and how you verified, in the AI-use
ledger that travels with every deliverable this semester.

> **A question that often comes up here:** *"If these tools invent things, why
> does the course encourage using them at all?"* Because they are genuinely
> powerful, and because you will use them for the rest of your career no
> matter what any course says. A search that once took an afternoon now takes
> minutes. The skill worth building is not avoidance. It is using the tool
> ambitiously and then verifying like a professional, every time.

Time to test the rule on a live case: your own curiosity.
