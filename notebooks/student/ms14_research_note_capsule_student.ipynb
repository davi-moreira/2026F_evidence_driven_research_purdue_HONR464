## 🎯 Definition of Done

You are done with this sprint when your submission has both halves built from
**your own** finished study, each part pointing at something a reader could check.

**The research note:**

1. **The problem–gap–question spine, in prose.** Three argued paragraphs, where the
   gap paragraph names what prior work covered, what it left out, and why the
   missing piece matters, and every claim rests on a real, retrieved source.
2. **The results hierarchy.** Your headline finding at the top, at its real
   boundary, with its uncertainty in the same breath, and any exploratory finding
   labeled and placed last, never leading.
3. **Limitations sized to the evidence,** plus your **AI-edit log** naming which AI
   edits you kept, changed, or rejected, and why for each.

**The reproducibility capsule:**

4. **A runnable notebook** that passes **restart-and-run-all**, a **data-provenance
   note**, a **fixed seed**, a **decision log**, and your finalized **AI-use
   ledger**, audited against the five package sins, with the one sin you found and
   fixed named.

Plus the **AI Research Ledger** rows and the **dossier update line**. At Friday's
studio a partner table-reads your note and reruns your capsule cold.

> **A question that often comes up here:** *"My poster already made the point.
> Isn't the note just the poster with more words?"* No. The poster made its point
> to a reader you could coach out loud; the note argues it to a stranger who has
> only the page. Every leap the box let you imply, the prose has to earn, in order.
> The note is longer because it does the work your voice was doing, not for its own
> sake.

### 🤝 AI Research Partner

AI is your arm and your research assistant in this sprint, not your brain. This
week it works as a **writing critic**, not a ghostwriter. A good use looks like
this: you draft a paragraph, then ask a tool to red-team it, list the citations it
proposes so you can retrieve each one yourself, or rank your limitations so you can
check the order against your own interval. A bad use is letting it write your
discussion, decide which finding leads, or hand you a source you never opened.

**Good to delegate here:**
- Red-teaming a discussion paragraph **you** already wrote, for any sentence that
  claims more than your evidence supports.
- Listing the inputs and steps a cold replicator would need from your capsule and
  might not find.
- Ranking your drafted limitations by how much each threatens the headline, so you
  can check the ranking against your printed interval.

**Never delegate (these are yours to decide):**
- Which claim is your headline and how big it is allowed to be.
- Which sources are real, and the words you put your name on.
- The wording of your limitations. A caveat you did not reason to is one you cannot
  defend at the table read.

**Commit first, then ask.** Draft your own sentence before you open a tool, so its
output has something of yours to disagree with. A cleaner sentence is not a bigger
finding: keep only edits that leave the claim its size. Verify what comes back, and
log it in the **AI Research Ledger** below. See
`ai_resources/human_responsibility_checklist.md` for the full never-delegate list.

### 🛠️ Hands-On: check the three floors your note and capsule must clear

Run the scaffold below on its default inputs first to see what it does. Then edit
the **three inputs** at the top with the pieces of *your own* project, and rerun.
Everything downstream is derived from those three edits, so there is one source of
truth for every number: change an input once and the whole readout stays
consistent. The scaffold checks three floors, not ceilings. Clearing them is
necessary, never sufficient, and the rest of the milestone is where the grade
lives.

```python
# --- Note & capsule self-check — edit ONLY the three inputs below, then rerun.
# Everything after them is DERIVED, so there is one source of truth per number.

# EDIT ME (1) — the per-unit numbers BEHIND your headline finding: one value per
# independent unit (a respondent, a site, a trial). The defaults let the cell run.
headline_values = [4.2, 3.1, 5.0, 2.8, 4.6, 3.7, 5.3, 2.2, 4.0, 3.9,
                   4.8, 2.5, 3.3, 5.1, 4.4, 3.0, 4.7, 2.9, 3.6, 4.1]

# EDIT ME (2) — your synthesis paragraph. Mark each sentence that rests on a real,
# retrieved source with [verified: ...]; leave bare, unsourced claims unmarked.
synthesis_paragraph = (
    "Prior surveys documented the problem in urban settings [verified: retrieved]. "
    "One retrieved study measured part of it directly [verified: retrieved]. "
    "No retrieved work has measured it in the units my frame reaches. "
    "That untouched case is the gap my question closes."
)

# EDIT ME (3) — a few key lines from your capsule, exactly as they appear in it.
capsule_lines = [
    "SEED = 464",
    "rng = np.random.default_rng(SEED)",
    'df = load_course_data("your_data.csv")   # provenance + license in the ledger',
    'df = df[df["outcome"].notna()]   # reason: pre-registered complete cases; in the decision log',
]

# ---- everything below is DERIVED from the three edits above ----

# FLOOR 1 — the headline's uncertainty: a seeded 95% bootstrap interval on the mean.
vals = np.asarray(headline_values, dtype=float)
n = len(vals)
boot = np.array([rng.choice(vals, size=n, replace=True).mean() for _ in range(2000)])
lo, hi = np.percentile(boot, [2.5, 97.5])
print(f"FLOOR 1  headline (n = {n}): mean = {vals.mean():.2f}   "
      f"95% interval [{lo:.2f}, {hi:.2f}]   width {hi - lo:.2f}")
print("         Report the interval in the SAME sentence as the estimate.")
print("         A wider interval is a smaller claim, not a failed one.")

# FLOOR 2 — the synthesis floor: evidence density = sourced sentences / all sentences.
parts = [s.strip() for s in synthesis_paragraph.replace("?", ".").replace("!", ".").split(".") if s.strip()]
sourced = [s for s in parts if "[verified" in s.lower()]
density = len(sourced) / len(parts) if parts else 0.0
print(f"\nFLOOR 2  synthesis: {len(parts)} sentences, {len(sourced)} sourced "
      f"-> evidence density = {density:.2f}")
print("         Density is a FLOOR, never proof of quality: it cannot tell")
print("         whether a source actually says what you claim.")

# FLOOR 3 — the capsule floor: scan the pasted lines for the five package sins.
low = [ln.lower() for ln in capsule_lines]
joined = "\n".join(low)
sins = []
if any(("/users/" in ln) or ("/home/" in ln) for ln in low):
    sins.append("hard-coded path")
if not any(k in joined for k in ("seed", "random_state", "default_rng")):
    sins.append("missing seed")
if any((".loc[" in ln and "=" in ln) or ("by hand" in ln) for ln in low):
    sins.append("by-hand edit")
filtered = [ln for ln in low if (".drop(" in ln) or ("df[df[" in ln.replace(" ", ""))]
if filtered and not any("reason" in ln for ln in low):
    sins.append("undocumented exclusion")
loads_csv = any("read_csv" in ln for ln in low)
shareable = any(("http" in ln) or ("load_course_data" in ln) for ln in low)
if loads_csv and not shareable and not any(("version" in ln) or ("license" in ln) for ln in low):
    sins.append("stale/unversioned data")
print(f"\nFLOOR 3  capsule sins flagged: {sins if sins else 'none (runnable, not yet proven correct)'}")
print("         A zero-flag capsule is RUNNABLE, never proven CORRECT.")
print("         The partner who reruns it cold on Friday is the real proof.")
```

### 🗡️ Red-Team Exchange

The Friday studio's 29–41 minute block. Pair with one partner and run two rounds,
with an AI reviewer as a third voice. The role for this milestone is the **Research
Note Reviewer**, which reads the *written genre*: whether every claim carries its
uncertainty and traces to your evidence (quick card in
`ai_resources/agent_role_cards.md`; full briefing in
`genai_studio/roles/research_note_reviewer.md`). It reviews the draft; it never
writes it, and it cannot run your capsule.

**First, the Research Note Reviewer pass (before the rounds).** Paste your full
draft with your compass classification and, for each key number, one line on where
it comes from. It returns a structure check, an uncertainty-and-limitations check,
and a traceability check. Every flag is a **suspect you test yourself** against the
M14 brief, not a verdict, and every missing claim or limitation it names is one
**you** must write.

**Round 1: attack the note's claims (about 15 minutes).** Your partner table-reads
your note and leaves two margin notes: your **sharpest sentence** and your
**weakest claim**. They challenge one claim you cannot trace to a number or a
retrieved source: "show me the printout that says this headline is 3.9, not 4." You
defend or concede, then revise so the weakest claim is disciplined by the evidence,
not cosmetically reworded.

**Round 2: rerun the capsule cold (about 15 minutes).** Your partner opens your
capsule on a fresh kernel and runs **restart-and-run-all**, with no help from you.
They log whether your headline number comes back and where, if anywhere, it broke.
A capsule that runs for you and not for them is a finding you fix before Friday
night, not a failure to hide.

**Mind the correlated-error trap.** If Gemini polished your prose and the Research
Note Reviewer calls it sound, that agreement is not confirmation your claims are
bounded. Neither tool ran your capsule. Only the partner rerun settles whether the
number reproduces.

**What to log.** Write down the strongest hit you gave and the strongest you got
back, the margin notes you received, and the one rerun that decided any
disagreement. A challenge that changed nothing still earns a line saying why it did
not land. That record is part of what you submit.

> **A question that often comes up here:** *"What if the Research Note Reviewer and
> my partner disagree about whether a claim is bounded?"* Your evidence settles it,
> not either of them. The Reviewer read only your text and can miss a naked claim or
> invent a structural worry; your partner read the page but not your ledger. Say in
> one sentence which read your own printout and claim ledger support, and why. That
> adjudication is stronger proof you understand your note than agreeing with a tool
> would be.
