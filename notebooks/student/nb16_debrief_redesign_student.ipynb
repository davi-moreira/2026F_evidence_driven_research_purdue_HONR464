## 2. Feedback Is Data — Code It Like Data

Any question you received can be coded on three dimensions. Together they turn a
blur of an afternoon into a table you could hand to a stranger:

| Code | What it captures | Values |
|---|---|---|
| **Type** | what *kind* of question was it? | clarifying, scope/generalization, mechanism ("why"), so-what/implication, method challenge, data-source, hostile/skeptical |
| **Target** | did it aim at your **claims** (does the evidence support what you say?) or your **communication** (could a listener follow you?)? | claims / communication |
| **Recurrence** | how many times, and from whom? | a count (an item asked once is an anecdote; asked three times, it is a finding about your project) |

The **target** code is the one that does the real work. A question like *"I
couldn't tell which bar was your treatment group"* is a **communication**
signal — your figure failed, not your evidence. A question like *"Would this
hold for anyone outside your sample?"* is a **claims** signal — the listener is
doubting the reach of what you can support. Fixing the first makes your poster
clearer; answering the second may change what you are allowed to claim. Sorting
them is the whole point.

```python
# The worked tally — six coded rows standing in for one afternoon.
# SWAP IN YOUR OWN: replace these rows with the questions you were actually
# asked. Keep the four columns; use the code values from the table above.
tally = pd.DataFrame([
    {"question": "Would this hold on other campuses too?",              "type": "scope/generalization", "target": "claims",        "count": 3},
    {"question": "Why would that mechanism produce the effect?",         "type": "mechanism",            "target": "claims",        "count": 2},
    {"question": "What's the takeaway for someone outside your field?",  "type": "so-what/implication",  "target": "claims",        "count": 3},
    {"question": "I couldn't tell which bar was the treatment group.",   "type": "clarifying",           "target": "communication", "count": 4},
    {"question": "Where did the data come from?",                        "type": "data-source",          "target": "claims",        "count": 1},
    {"question": "Your title was hard to read from a few feet away.",    "type": "clarifying",           "target": "communication", "count": 2},
])

total = int(tally["count"].sum())
by_type = tally.groupby("type")["count"].sum().sort_values(ascending=False)
by_target = tally.groupby("target")["count"].sum().sort_values(ascending=False)
loudest = tally.loc[tally["count"].idxmax()]

print("✓ Coded", len(tally), "distinct questions,", total, "asked in total")
print()
print("By TYPE (weighted by recurrence):")
print(by_type.to_string())
print()
print("By TARGET:")
print(by_target.to_string())
print()
print("Most-repeated single question ({}x): {}".format(int(loudest['count']), loudest['question']))

# Self-check: the counts must add up the same way no matter how we slice them.
assert by_type.sum() == total == by_target.sum(), "slices disagree — recount!"
assert by_type.index[0] == "clarifying", "dominant type drifted — investigate"
print()
print("✓ Self-check passed — the two slices agree on", total, "total questions")
```

## 3. From Feedback to a Redesign — Mapping onto MIDA

> *"Good. You have your data. Now do not bring me a wish list. 'More
> participants, more time' is how every researcher feels leaving a conference and
> it is not a redesign — it points at nothing you decided. Tell me instead which
> *part* of the design each comment implicates, and what changing it would buy."*
> — your **thesis advisor**, reading your tally on Friday

Today you turn coded feedback into a **plan**. The move a working researcher
makes after a conference is to take each important comment and map it onto the
part of the design it implicates — the same **MIDA** letters you have used all
semester:

- **M**odel — the world, the assumptions, the mechanism. A *"why would that
  happen?"* question usually lives here.
- **I**nquiry — the quantity or question itself. A *"so what?"* or *"is that
  even the right thing to measure?"* question.
- **D**ata — sampling, measurement, assignment, sources. A *"who was in your
  sample?"* or *"how did you measure that?"* question — and, crucially, most
  **generalization** questions, because how far a result reaches is decided by
  the data strategy, not by wishing.
- **A**nswer — the analysis and how it maps evidence to the claim. A *"would that
  result survive a different method?"* question.

Three worked mappings, the ones you will reach for most:

- A **generalization** question — *"Would this hold for graduate researchers, or
  other campuses?"* → **D**. It feels like an Inquiry question, but the fix lives
  in the **Data** strategy: who is in your sample and what frame they represent.
- A **mechanism** question — *"Why would social media drive anxiety, exactly?"*
  → **M**. It asks about the world your claim assumes, which is the Model.
- A **so-what** question — *"Why does this quantity matter to anyone?"* → **I**.
  It challenges the Inquiry itself: whether the thing you chose to estimate is
  worth estimating.

> **A question that often comes up here:** *"A single comment feels like it hits
> two letters at once — is that allowed?"* Often it does, and the useful move is
> to name the letter where the *fix* lives, not where the discomfort lives. "Your
> comparison group might differ on major" stings like an Answer problem (your
> analysis looks wrong) but is repaired in **D** (measure major, stratify or
> adjust). Map to where the redesign acts.

## 4. NOW vs NEXT — Redesign, Not Regret

Every redesign gets one more tag: **NOW** or **NEXT**.

- **NOW** — it fits inside the two weeks that remain, so it becomes a commitment
  in your dossier work plan (the analytical core you build in nb17, then M20–M23).
  A NOW item is a promise you will keep, listed where your December work can find
  it.
- **NEXT** — it is real and worth stating, but it belongs to the imagined next
  study you will not run for this course. Naming what you *would* do next is part
  of a mature redesign; pretending you can do it by Thanksgiving is not.

The line this meeting exists to enforce: **"more data, more time" is a lament,
not a redesign.** A redesign points at a Model, Inquiry, Data, or Answer decision
you would make *differently*, and it names the **diagnosed benefit** — which
design virtue improves: less **bias**, more **precision**, wider
**generalization**, or clearer **mechanism**. A redesign you could print on a
fortune cookie is not one.

> **A question that often comes up here:** *"Isn't it defeatist to tag my best
> idea NEXT?"* The opposite — it is the most honest thing in the plan. Three
> ambitious redesigns all tagged NOW guarantee at least two broken promises, and
> a broken promise in your dossier costs you more than a clearly-labeled NEXT
> ever would. Triage as if the calendar were real, because it is.

### 🛠️ Hands-On (optional): re-diagnose with post-conference parameters

One kind of feedback deserves a number, not just a paragraph: when the audience
made you admit that the effect you can honestly defend is **smaller** than your
pilot flattered you into hoping. If a redesign is "run it again at a larger
sample," the honest question is *what that buys* — and the semester's `diagnose()`
engine answers it. The cell reruns a two-arm diagnosis at a modest, post-Expo
effect across three sample sizes, so you can read the **power** a bigger study
would actually purchase.

```python
# diagnose() — the seeded Monte-Carlo engine from earlier in the course.
def diagnose(design_fn, reps=1000, seed=SEED):
    r = np.random.default_rng(seed)
    d = pd.DataFrame([design_fn(r) for _ in range(reps)],
                     columns=["est", "se", "truth"])
    return pd.Series({
        "bias":     (d.est - d.truth).mean(),
        "power":    (np.abs(d.est / d.se) > 1.96).mean(),
        "coverage": ((d.est - 1.96*d.se <= d.truth) &
                     (d.truth <= d.est + 1.96*d.se)).mean(),
    })

def two_arm(n, effect):
    """One simulated two-arm study of size n and true effect `effect`."""
    def _run(r):
        z = np.zeros(n, dtype=int); z[:n // 2] = 1
        z = r.permutation(z)
        y0 = r.normal(0, 1, n)
        y = np.where(z == 1, y0 + effect, y0)
        g1, g0 = y[z == 1], y[z == 0]
        est = g1.mean() - g0.mean()
        se = np.sqrt(g1.var(ddof=1) / len(g1) + g0.var(ddof=1) / len(g0))
        return est, se, effect
    return _run

# The audience talked you down from a hopeful 0.6 to a defensible 0.3.
honest_effect = 0.30
grid = pd.DataFrame({n: diagnose(two_arm(n, honest_effect)) for n in (60, 120, 240)}).T
grid.index.name = "n"
print("Diagnosis at the honest effect of", honest_effect, "(1000 sims each):")
print(grid.round(3).to_string())

# Self-check: the design is unbiased and well-calibrated; only power should climb.
assert grid["bias"].abs().max() < 0.05, "unexpected bias — the estimator should be clean"
assert grid.loc[60, "power"] < grid.loc[240, "power"], "power should rise with n"
print()
print("✓ Power climbs from {:.0%} (n=60) to {:.0%} (n=240) — that is what the "
      "redesign buys.".format(grid.loc[60, 'power'], grid.loc[240, 'power']))
```

Read the table honestly: **bias** stays near zero and **coverage** near
0.95 at every size — a bigger sample does not make the study more *correct*, only
more *powerful*. Power climbs from about 22% at n=60 to about 64% at n=240. So
"run it larger" is a real redesign only when you state the diagnosed benefit
(precision/power to detect an effect this modest) and tag it honestly — almost
always **NEXT**, since quadrupling a sample rarely fits in two weeks.
