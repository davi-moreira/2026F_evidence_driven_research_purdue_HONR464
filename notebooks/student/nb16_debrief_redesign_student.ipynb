## 2. Feedback Is Data. Code It Like Data.

**Guiding question:** *how do you turn a blur of an afternoon's questions into a table you can count, and keep what the crowd enjoyed separate from what your evidence can support?*

The tool is **coding**, in the qualitative-research sense: sorting free-form
text into a small set of labeled categories so you can count it. When you file
"I couldn't tell which bar was the treatment group" under *clarifying*, you are
coding. Every question on the feedback sheet you filled during the poster
session gets three codes:

| Code | What it captures | Values |
|---|---|---|
| **Type** | what *kind* of question was it? | clarifying, scope/generalization, mechanism ("why"), so-what/implication, method challenge, data-source, hostile/skeptical |
| **Target** | did it aim at your **claims** (does the evidence support what you say?) or your **communication** (could a listener follow you?) | claims / communication |
| **Recurrence** | how many times was it asked? | a count (asked once, it is an anecdote; asked three times, it is a finding about your project) |

The **target** code does the real work. "I couldn't tell which bar was your
treatment group" is a **communication** signal. Your figure failed, not your
evidence. "Would this hold for anyone outside your sample?" is a **claims**
signal. The listener doubts the reach of what you can support. Fixing the first
makes your poster clearer. Answering the second may change what you are allowed
to claim. Sorting them is the whole point.

## 3. From Feedback to a Redesign: Mapping onto MIDA

A **redesign** is a specific change to one design decision in your study, made
in response to what you learned. "Record each person's major and compare within
majors" is a redesign. "Get more data" is not; it names no decision. The move a
working researcher makes after a conference is **MIDA mapping**: matching each
important comment to the design component whose change would answer it, using
the same four letters you have carried all semester.

- **M**odel, the world your claim assumes: its assumptions and mechanism. A
  "why would that happen?" question usually lives here.
- **I**nquiry, the question itself. A "so what?" or "is that even the right
  thing to measure?" question.
- **D**ata, your sampling, measurement, and assignment choices. A "who was in
  your sample?" or "how did you measure that?" question. Crucially, most
  **generalization** questions live here too, because how far a result reaches
  is decided by the data strategy, not by wishing.
- **A**nswer, the analysis that maps evidence to the claim. A "would that
  result survive a different method?" question.

Three worked mappings, the ones you will reach for most:

- A **generalization** question ("would this hold on other campuses?") → **D**.
  It feels like an Inquiry question, but the fix lives in the data strategy:
  who is in your sample, and whom that sample can speak for.
- A **mechanism** question ("why would social media drive anxiety, exactly?")
  → **M**. It asks about the world your claim assumes, which is the Model.
- A **so-what** question ("why does this quantity matter to anyone?") → **I**.
  It challenges whether the thing you chose to estimate is worth estimating.

> **A question that often comes up here:** *"A single comment feels like it
> hits two letters at once. Is that allowed?"* Often it does, and the useful
> move is to name the letter where the *fix* lives, not where the discomfort
> lives. "Your comparison group might differ on major" stings like an Answer
> problem, because your analysis looks wrong. But it is repaired in **D**:
> measure major, then compare within majors. Map to where the redesign acts.

### NOW vs NEXT: redesign, not regret

Every redesign gets one more tag.

- **NOW**: it fits inside the two weeks that remain, so it becomes a commitment
  in your dossier work plan (the analytical core you build in nb17, then
  M20–M23). A NOW item is a promise you will keep.
- **NEXT**: it is real and worth stating, but it belongs to the next study, the
  one you will not run for this course. Naming what you *would* do next is part
  of a mature redesign. Pretending you can do it before the semester ends is
  not.

The line this hour exists to enforce: **"more data, more time" is a lament, not
a redesign.** A redesign points at a Model, Inquiry, Data, or Answer decision
you would make differently. It also names its **diagnosed benefit**, the design
virtue the change buys: less **bias**, more **precision**, wider
**generalization**, or clearer **mechanism**. A redesign you could print on a
fortune cookie is not one.

> **A question that often comes up here:** *"Isn't it defeatist to tag my best
> idea NEXT?"* The opposite. It is the most honest thing in the plan. Three
> ambitious redesigns all tagged NOW guarantee at least two broken promises,
> and a broken promise in your dossier costs you more than a clearly labeled
> NEXT ever would. Triage as if the calendar were real, because it is.

## 4. If You Want to Go Further *(optional)*

This section is optional; work through it after class if it fits your project.
It is for one situation in particular: the audience talked you down to a
smaller honest effect than your pilot flattered you into hoping for, and your
instinct is "run it again, bigger."

That kind of feedback deserves a number, not just a paragraph. If your redesign
is "a larger sample," the honest question is what that buys, and the semester's
`diagnose()` engine answers it. The cell below reruns a **two-arm** diagnosis
(one treatment group, one control group) at a modest post-Expo effect across
three sample sizes. Two reminders before you read the table. **Power** is the
chance your study detects an effect that is really there. **Bias** is a
systematic tilt: your procedure missing the truth in the same direction on
average.

> 💡 **Gemini Prompt:** "The audience talked me down to a smaller honest effect
> size for my project, and I'm considering re-running at a larger sample. Here
> are the effect size and the sample sizes I'm weighing: [paste them]. Explain
> what 'power' means here, and why a bigger sample would raise power but not
> fix bias. Then I'll confirm the pattern against the diagnose() table I run."
>
> **After running, verify (the responsible-AI habit):**
> - [ ] Run the cell and check that bias really does stay near zero while power
>   climbs. The printed table is the evidence, not Gemini's explanation of it.
> - [ ] Make sure any redesign you tag from this names its diagnosed benefit
>   (precision / power) and is triaged honestly. Usually NEXT, not NOW.
> - [ ] Log this use in your AI ledger: tool, task, verification.

```python
# diagnose() — the seeded Monte-Carlo engine from earlier in the course.
def diagnose(design_fn, reps=1000, seed=SEED):
    r = np.random.default_rng(seed)
    d = pd.DataFrame([design_fn(r) for _ in range(reps)],
                     columns=["est", "se", "truth"])
    return pd.Series({
        "bias":     (d.est - d.truth).mean(),
        "power":    (np.abs(d.est / d.se) > 1.96).mean(),
        "coverage": ((d.est - 1.96*d.se <= d.truth) &
                     (d.truth <= d.est + 1.96*d.se)).mean(),
    })

def two_arm(n, effect):
    """One simulated two-arm study of size n and true effect `effect`."""
    def _run(r):
        z = np.zeros(n, dtype=int); z[:n // 2] = 1
        z = r.permutation(z)
        y0 = r.normal(0, 1, n)
        y = np.where(z == 1, y0 + effect, y0)
        g1, g0 = y[z == 1], y[z == 0]
        est = g1.mean() - g0.mean()
        se = np.sqrt(g1.var(ddof=1) / len(g1) + g0.var(ddof=1) / len(g0))
        return est, se, effect
    return _run

# The audience talked you down from a hopeful 0.6 to a defensible 0.3.
honest_effect = 0.30
grid = pd.DataFrame({n: diagnose(two_arm(n, honest_effect)) for n in (60, 120, 240)}).T
grid.index.name = "n"
print("Diagnosis at the honest effect of", honest_effect, "(1000 sims each):")
print(grid.round(3).to_string())

# Self-check: the design is unbiased and well-calibrated; only power should climb.
assert grid["bias"].abs().max() < 0.05, "unexpected bias — the estimator should be clean"
assert grid.loc[60, "power"] < grid.loc[240, "power"], "power should rise with n"
print()
print("✓ Power climbs from {:.0%} (n=60) to {:.0%} (n=240) — that is what the "
      "redesign buys.".format(grid.loc[60, 'power'], grid.loc[240, 'power']))
```

Read the table honestly. **Bias** stays near zero and **coverage**
(how often the uncertainty interval catches the truth) stays near 0.95 at every
size. A bigger sample does not make the study more *correct*, only more
*powerful*. Power climbs from about 22% at n=60 to about 64% at n=240. So "run
it larger" is a real redesign only when you state the diagnosed benefit
(precision, the power to detect an effect this modest) and tag it honestly.
That tag is almost always **NEXT**, since quadrupling a sample rarely fits in
two weeks.
