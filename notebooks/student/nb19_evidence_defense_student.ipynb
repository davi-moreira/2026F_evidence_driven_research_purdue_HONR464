## 1. Why This Matters

> *"The defense is not where you prove you were right. It is where you show me
> you know exactly how far your evidence reaches — and that you have the nerve to
> stop there, out loud, while someone pushes you to go further. The researcher I
> trust is not the one who never gets pushed off balance. It is the one who
> concedes the right inch and holds the rest."*
> — a **thesis advisor**, on what a defense actually tests

Everything the semester built points here. Not a document — a **defense.** For
ten minutes you stand behind your **claims**, your **choices**, and your
**verification**: why this compass position, why this data, why you believe the number.
For five minutes the room and Professor Moreira **cross-examine** any of it — any
ledger row, any decision, any result. This is the moment the course's whole
ethic becomes personal: *a claim you cannot defend at its boundary was never
yours to make, and a verification you outsourced to an AI tool without checking
is a verification you do not have.*

It is not a debate you win by never yielding. It is a demonstration that you know
precisely how far your evidence reaches — which is why the rarest virtue it
rewards is **conceding gracefully, exactly where your own ledger says concede.**
Your ledger is not a liability in that room. It is your shield.

> **A question that often comes up here:** *"If I concede a limitation, haven't I
> just admitted my project is weak?"* The opposite. An unconceded limitation does
> not disappear when you refuse to name it — it sits there, visible to everyone
> but you, and every sentence you build on top of it looks naive. The researcher
> who says "this is descriptive; it cannot tell you it generalizes, and here is
> exactly why" looks *more* competent, not less, because they are demonstrating
> the one thing a defense is built to test: that they know the boundary and
> chose to stand on the right side of it.

## 2. How to Defend, and How to Cross-Examine

**Guiding question:** *how do you stand behind your claims, choices, and verification for ten minutes — and concede exactly the inch your evidence gives, without collapsing or bluffing?*

**The defender's structure — claims, choices, verification.** Lead with the
claims you are proud of, each stated at its real boundary. Then your choices —
why this compass position, this data strategy, this measurement — and what each can
and cannot establish. Then your verification — how you know each headline result is
real: recomputed by hand, cross-checked, reproduced by your partner at M20. Your
dossier draft and ledger are open in front of you the whole time. You are not
performing from memory; you are reading your own evidence back, fluently.

**The cross-examiner's craft — probe one, not everything.** When you are not
defending, you are a cross-examiner, and it is graded. A good question is
**ledger-grounded**: it names a row and aims at exactly one thing — one *claim*
("row 4 says the gap holds; does it hold in the population, or only your
sample?"), one *choice* ("why exclude the night-section respondents?"), or one
*verification* ("you say recomputed — recomputed how?"). A vague "did you think
about bias?" is not a question; it is a free ride, and free-riding in the
audience is a miss.

**Graceful concession — the strongest move in the room.** When a probe lands on a
real limit, concede it cleanly and hold the rest:

> *"That row is descriptive only. I can't tell you it generalizes — my sampling
> frame was one residence hall — so I won't claim that it does. Within the pilot,
> the 18-point gap is what I stand on, and here's the recompute."*

Boundary held, concession made where due, verification shown. Notice what it is
not: not a collapse ("I guess the whole thing is worthless"), and not a
white-knuckled refusal to yield. It gives the exact inch the evidence gives, and
not one inch more.

> **A question that often comes up here:** *"What if the room asks a question my
> ledger simply has no row for?"* Then you say so — that is the honest don't-know,
> and it is a strong move, not a weak one. "I don't have a row that licenses that
> claim, and I won't invent one; what I'd need to answer it is [the evidence]"
> shows the room you know exactly where your knowledge stops. The failure is not
> the missing row — it is minting a claim under pressure to cover it.

## 4. Synthesis: Every Compass Position, One Last Time

**Guiding question:** *one question, every compass position — what does each ask, need, permit, and forbid, and how do you hear which one a claim is really making?*

You began in August with a curiosity. You end with a defended, reproducible,
honestly-bounded claim. Between those two points sits the course's signature
skill — the one thing meant to outlast every dataset and deadline: **classify a
question by its kind and reach, place it on the inquiry compass, and hold every
claim inside that position's boundary.** Today you use it one last time, on a single
question, at every compass position at once — so you can feel that the positions are
not topics you studied but **questions you can ask of anything.**

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/rdss_fig_2_1.png" width="75%"/></center>

*The map from nb00, meeting 1: you now own every box on it — Model, Inquiry, Data,
and Answer (M·I·D·A), across the theoretical and empirical sides. The compass and
MIDA are partners, not rivals: the compass **names** the inquiry (its kind and its
reach); MIDA is how you **declare and defend** it. (Figure from the replication
materials of Blair, Coppock & Humphreys (2023),* Research Design in the Social
Sciences *— MIT-licensed archive; the book is free at book.declaredesign.org.)*

Take one concrete, campus-scale question: **"How is commute length related to
class attendance?"** Watch it become four different studies, each honest, each
bounded, none reducible to another:

| Compass position (kind · reach) | The question it asks here | Evidence it needs | What its answer PERMITS | What its answer FORBIDS |
|---|---|---|---|---|
| **Description** (descriptive · data at hand) | Among the undergraduates in my data, how does attendance differ across commute lengths? | one dataset, honestly summarized | "In these N undergraduates, those commuting over 40 minutes attended about X fewer sessions." | any leap to undergraduates in general; any "because" |
| **Generalization** (descriptive · a population) | In the population these were sampled from, does the attendance–commute gap hold, and how sure can I be? | a defensible sample + quantified uncertainty | "Across the population my frame reaches, the gap is about X, give or take a margin." | reaching past the frame; a causal reading |
| **Prediction** (descriptive · unseen cases) | Given a new undergraduate's commute, can I forecast their attendance? | an out-of-sample test against a baseline | "On held-out cases, commute forecasts attendance better than guessing the mean, by [margin]." | that commute *causes* attendance; importance as mechanism |
| **Causal reasoning** (causal) | If a long-commuter's commute were shortened, would their attendance rise? | an intervention + a stated identification argument | "Under [randomization / a defended design], shortening the commute changes attendance by [effect]." | an effect with no identification; "after, therefore because" |

Four sentences, four kinds of evidence, four different claims you are allowed to
make — and, crucially, four different overreaches to refuse. The same numbers
could sit under all four studies; what changes is the **question**, and the
question is what decides what the answer may say.

Read the four as a single motion. **Description** stays home: it speaks only
about the rows in front of you, and its whole discipline is not to wander past
them. **Generalization** travels — it uses those rows to speak about a wider
population — and it pays for the trip with stated uncertainty and a defensible
sample. **Prediction** turns to face the future: it asks only whether commute
helps *forecast* attendance for cases you have not seen, and it buys reach by
giving up any claim to *why*. **Causal reasoning** is the boldest and the hardest:
it asks about a *changed world* — the same person with a shorter commute — a world
you never fully observe, which is why it demands an identification argument before
you may say "would change." Each step across the compass is bought with a
license, and the failure that ends careers is never using the wrong tool — it is
using description's evidence and speaking in causation's voice, a crossing made
without paying. The skill you leave with is hearing which one a claim is really
making — in a paper, in a headline, in your own mouth.

> **A question that often comes up here:** *"Isn't causal reasoning just the best
> compass position — why would I ever settle for description?"* Because each
> position answers a different question, and description is not a weaker causal
> study — it is the honest, complete answer when your question is "what is here?" A
> precise description bounded to its data is worth more than a causal claim its
> design cannot support. The strongest researchers are not the ones always reaching
> for cause; they are the ones who match the position to the question and refuse to
> borrow a bolder claim than their evidence licenses.

You have written a claim ticket at the end of nearly every meeting since August —

standards changing. Before you write your own first-vs-last comparison, let's

### 🛠️ Hands-On: your claim-ticket archive, then vs now

Run the cell. It aggregates an example archive into a **then-vs-now** comparison:
the share of your earliest claims that both hedged uncertainty and stated a
boundary, versus the share of your latest ones. The numbers are illustrative —
swap in your own tickets — but the shape is the point.

```python
# An example claim-ticket archive (illustrative strings — swap in your own tickets).
# Two flags per ticket: does it HEDGE uncertainty, and does it state a BOUNDARY?
archive = [
    {"ticket": 1,  "hedge": False, "boundary": False,
     "claim": "Social media makes undergraduates anxious."},
    {"ticket": 5,  "hedge": False, "boundary": True,
     "claim": "My question is a descriptive question, for now."},
    {"ticket": 13, "hedge": False, "boundary": True,
     "claim": "In these 320 pilot respondents, 43% reported skipping office hours."},
    {"ticket": 39, "hedge": True,  "boundary": True,
     "claim": "A partner reproduced my headline number on Dec 2; one residual, now fixed."},
    {"ticket": 43, "hedge": True,  "boundary": True,
     "claim": ("In my 320-person pilot, first-generation respondents skipped office hours "
               "about 18 points more often — a gap that may not hold in the full population.")},
]

def disciplined_share(tickets):
    # A claim is 'disciplined' when it BOTH hedges uncertainty AND states a boundary.
    strong = [t for t in tickets if t["hedge"] and t["boundary"]]
    return len(strong) / len(tickets)

early = [t for t in archive if t["ticket"] <= 5]     # my first standards
late  = [t for t in archive if t["ticket"] >= 39]    # my last standards
then = disciplined_share(early)
now  = disciplined_share(late)

print("THEN — my earliest claim (Ticket 1):", archive[0]["claim"])
print("NOW  — my latest claim   (Ticket 43):", archive[-1]["claim"])
print()
print(f"Then (tickets ≤ 5) : {then:.0%} of my claims both hedged AND stated a boundary")
print(f"Now  (tickets ≥ 39): {now:.0%} of my claims both hedged AND stated a boundary")
```

```python
# Self-check — the standard is what changed, not the topic.
assert then == 0.0 and now == 1.0, "then/now drifted — check the archive flags"
print("✓ Self-check passed: 0% then, 100% now.",
      "Same curiosity, a completely different standard of proof.")
```

it, and say honestly what moved. You can make a package a stranger can run, a brief

gracefully, exactly where your evidence stops. That last one is the whole course.
