## 1. Why This Matters

> *"The defense is not where you prove you were right. It is where you show me
> you know exactly how far your evidence reaches, and that you have the nerve to
> stop there, out loud, while someone pushes you to go further. The researcher I
> trust is not the one who never gets pushed off balance. It is the one who
> concedes the right inch and holds the rest."*
> — a **thesis advisor**, on what a defense actually tests

Everything the semester built points here. Not a document, a performance. An
**evidence defense** is a live session where you stand behind your project's
claims and answer hard questions about them, the way a doctoral researcher
defends a dissertation. For example: you state your headline finding, and the
room asks how you know the number is real. For ten minutes you stand behind
your **claims**, your **choices**, and your **verification**: why this compass
position, why this data, why you believe the number. For five minutes the room
and Professor Moreira question any of it. Any row, any decision, any result.

Your anchor in that room is your **claim ledger**, the running table you have
kept all semester: one row per claim, each with its evidence, its verification,
and its boundary. One row might read: "In my 320-person pilot, first-generation
respondents skipped office hours 18 points more often; recomputed by hand;
descriptive only." The ledger is not a liability in the defense. It is your
shield.

This is where the course's whole ethic becomes personal. A claim you cannot
defend at its boundary was never yours to make. A verification you outsourced
to an AI tool without checking is a verification you do not have. And the
rarest virtue the defense rewards is **conceding gracefully, exactly where your
own ledger says concede.**

> **A question that often comes up here:** *"If I concede a limitation, haven't
> I just admitted my project is weak?"* The opposite. An unconceded limitation
> does not disappear when you refuse to name it. It sits there, visible to
> everyone but you, and every sentence you build on top of it looks naive. The
> researcher who says "this is descriptive; it cannot tell you it generalizes,
> and here is exactly why" looks *more* competent, not less. They are
> demonstrating the one thing a defense is built to test: they know the
> boundary, and they chose to stand on the right side of it.

First, the rules of the room: how a defense runs, and how to question one.

## 2. How to Defend, and How to Cross-Examine

**Guiding question:** *how do you stand behind your claims, choices, and
verification for ten minutes, then concede exactly the inch your evidence
gives, without collapsing or bluffing?*

The defense runs the same way across both of this topic's lectures. Half the
room defends on day one, half on day two, and everyone cross-examines on both
days.

**The defender's structure: claims, choices, verification.** Lead with the
claims you are proud of, each stated at its real boundary. Then your choices:
why this compass position, this data strategy, this measurement, and what each
can and cannot establish. Then your verification: how you know each headline
result is real. Recomputed by hand, cross-checked against a second source, or
reproduced by your partner at M20. Your **dossier** draft stays open in front
of you, next to your ledger. The dossier is the single organized folder that
holds your whole project: question, sources, design, analyses, ledger, brief. A
stranger could open it and retrace every decision you made. You are not
performing from memory. You are reading your own evidence back, fluently.

**The cross-examiner's craft: probe one thing, not everything.** When you are
not defending, you are a cross-examiner, and that role is graded on both days.
**Cross-examination** is the question period after each defense, when the room
probes specific claims. A good question is **ledger-grounded**: it names a row
and aims at exactly one thing. One *claim* ("row 4 says the gap holds; does it
hold in the population, or only your sample?"). One *choice* ("why exclude the
night-section respondents?"). One *verification* ("you say recomputed;
recomputed how?"). A vague "did you think about bias?" is not a question. It is
a free ride, and free-riding in the audience is a miss.

**Graceful concession: the strongest move in the room.** When a probe lands on
a real limit, concede it cleanly and hold the rest:

> *"That row is descriptive only. I can't tell you it generalizes, because my
> sampling frame (the list of people I could actually draw from) was one
> residence hall, so I won't claim that it does. Within the pilot, the 18-point
> gap is what I stand on, and here's the recompute."*

Boundary held, concession made where due, verification shown. Notice what it
is not. It is not a collapse ("I guess the whole thing is worthless"), and it
is not a white-knuckled refusal to yield. It gives the exact inch the evidence
gives, and not one inch more.

Three moves are forbidden, and each costs points on the rubric: **inventing a
new claim under pressure**, **retracting a concession without new evidence**,
and **blaming the AI tool** for a verification you own.

> **A question that often comes up here:** *"What if the room asks a question
> my ledger simply has no row for?"* Then you say so. That is the honest
> don't-know, and it is a strong move, not a weak one. "I don't have a row that
> licenses that claim, and I won't invent one; what I'd need to answer it is
> [the evidence]" shows the room you know exactly where your knowledge stops.
> The failure is not the missing row. The failure is minting a claim under
> pressure to cover it.

## 4. The Final Studio: Every Compass Position, One Last Time

**Guiding question:** *one question, every compass position: what does each
ask, need, permit, and forbid, and how do you hear which one a claim is really
making?*

The defenses are behind you; what remains is to fold the course into a single
skill. This closing section is the material for the final Friday studio, the
last time the course meets, and it is short on purpose: one synthesis exercise,
one piece of archaeology on your own claim tickets, and the exit claim.
**Synthesis** means pulling everything you learned into one view, so it works
as a single skill instead of a list of topics. The concrete form it takes here:
one question, read at all four compass positions at once.

You began in August with a curiosity. You end with a defended, reproducible,
honestly-bounded claim. Between those two points sits the course's signature
skill, the one thing meant to outlast every dataset and deadline: **classify a
question by its kind and reach, place it on the inquiry compass, and hold every
claim inside that position's boundary.** In the studio you use it one last
time, on a single question, at every position at once, so you can feel that the
positions are not topics you studied but **questions you can ask of anything.**

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/rdss_fig_2_1.png" width="75%"/></center>

*The map from nb00, the very first notebook: you now own every box on it —
Model, Inquiry, Data, and Answer (M·I·D·A), across the theoretical and
empirical sides. The compass and MIDA are partners, not rivals: the compass
**names** the inquiry (its kind and its reach); MIDA is how you **declare and
defend** it. (Figure from the replication materials of Blair, Coppock &
Humphreys (2023),* Research Design in the Social Sciences*, an MIT-licensed
archive; the book is free at book.declaredesign.org.)*

Take one concrete, campus-scale question: **"How is commute length related to
class attendance?"** Watch it become four different studies, each honest, each
bounded, none reducible to another:

| Compass position (kind · reach) | The question it asks here | Evidence it needs | What its answer PERMITS | What its answer FORBIDS |
|---|---|---|---|---|
| **Description** (descriptive · data at hand) | Among the undergraduates in my data, how does attendance differ across commute lengths? | one dataset, honestly summarized | "In these N undergraduates, those commuting over 40 minutes attended about X fewer sessions." | any leap to undergraduates in general; any "because" |
| **Generalization** (descriptive · a population) | In the population these were sampled from, does the attendance–commute gap hold, and how sure can I be? | a defensible sample + quantified uncertainty | "Across the population my frame reaches, the gap is about X, give or take a margin." | reaching past the frame; a causal reading |
| **Prediction** (descriptive · unseen cases) | Given a new undergraduate's commute, can I forecast their attendance? | a test on held-out cases (cases the model never saw) against a baseline | "On held-out cases, commute forecasts attendance better than guessing the mean, by [margin]." | that commute *causes* attendance; importance as mechanism |
| **Causal reasoning** (causal) | If a long-commuter's commute were shortened, would their attendance rise? | an intervention + a stated **identification argument** (the reason you may read a difference as cause) | "Under [randomization / a defended design], shortening the commute changes attendance by [effect]." | an effect with no identification; "after, therefore because" |

Four sentences, four kinds of evidence, four different claims you are allowed
to make, and, crucially, four different overreaches to refuse. The same numbers
could sit under all four studies. What changes is the **question**, and the
question is what decides what the answer may say.

Read the four as a single motion. **Description** stays home: it speaks only
about the rows in front of you, and its whole discipline is not to wander past
them. **Generalization** travels. It uses those rows to speak about a wider
population, and it pays for the trip with stated uncertainty and a defensible
sample. **Prediction** turns to face the future: it asks only whether commute
helps *forecast* attendance for cases you have not seen, and it buys that reach
by giving up any claim to *why*. **Causal reasoning** is the boldest and the
hardest: it asks about a *changed world*, the same person with a shorter
commute, a world you never fully observe. That is why it demands an
identification argument before you may say "would change." Each step across the
compass is bought with a license. The failure that ends careers is never using
the wrong tool. It is using description's evidence while speaking in
causation's voice: a crossing made without paying. The skill you leave with is
hearing which claim a sentence is really making: in a paper, in a headline, in
your own mouth.

> **A question that often comes up here:** *"Isn't causal reasoning just the
> best compass position — why would I ever settle for description?"* Because
> each position answers a different question, and description is not a weaker
> causal study. It is the honest, complete answer when your question is "what
> is here?" A precise description bounded to its data is worth more than a
> causal claim its design cannot support. The strongest researchers are not the
> ones always reaching for cause. They are the ones who match the position to
> the question and refuse to borrow a bolder claim than their evidence
> licenses.

### 🛠️ Hands-On: your claim-ticket archive, then vs now

Run the cell. It aggregates an example archive into a **then-vs-now**
comparison: the share of the earliest claims that both hedged uncertainty and
stated a boundary, versus the share of the latest ones. The numbers are
illustrative. Swap in your own tickets; the shape is the point.

```python
# An example claim-ticket archive (illustrative strings — swap in your own tickets).
# Two flags per ticket: does it HEDGE uncertainty, and does it state a BOUNDARY?
archive = [
    {"entry": 1,  "hedge": False, "boundary": False,
     "claim": "Social media makes undergraduates anxious."},
    {"entry": 5,  "hedge": False, "boundary": True,
     "claim": "My question is a descriptive question, for now."},
    {"entry": 13, "hedge": False, "boundary": True,
     "claim": "In these 320 pilot respondents, 43% reported skipping office hours."},
    {"entry": 28, "hedge": True,  "boundary": True,
     "claim": "A partner reproduced my headline number this week; one residual, now fixed."},
    {"entry": 30, "hedge": True,  "boundary": True,
     "claim": ("In my 320-person pilot, first-generation respondents skipped office hours "
               "about 18 points more often — a gap that may not hold in the full population.")},
]

def disciplined_share(tickets):
    # A claim is 'disciplined' when it BOTH hedges uncertainty AND states a boundary.
    strong = [t for t in tickets if t["hedge"] and t["boundary"]]
    return len(strong) / len(tickets)

early = [t for t in archive if t["entry"] <= 5]     # my first standards
late  = [t for t in archive if t["entry"] >= 28]    # my last standards
then = disciplined_share(early)
now  = disciplined_share(late)

print("THEN — my earliest claim (entry 1): ", archive[0]["claim"])
print("NOW  — my latest claim   (entry 30):", archive[-1]["claim"])
print()
print(f"Then (entries ≤ 5) : {then:.0%} of my claims both hedged AND stated a boundary")
print(f"Now  (entries ≥ 28): {now:.0%} of my claims both hedged AND stated a boundary")
```

```python
# Self-check — the standard is what changed, not the topic.
assert then == 0.0 and now == 1.0, "then/now drifted — check the archive flags"
print("✓ Self-check passed: 0% then, 100% now.",
      "Same curiosity, a completely different standard of proof.")
```

## 5. Wrap-Up — the last word

Look at what you can do now that you could not do in August. You can take a
curiosity too big to hold and walk it down to a question evidence can answer.
You can tell which compass position your question really occupies, and hold
your claim inside its boundary while someone pushes. You can find a real source
and catch an invented one. You can declare a model, name an inquiry, and
measure a concept without pretending the number is the thing. You can build a
design, diagnose it, redesign it, and say honestly what moved. You can make a
package a stranger can run, a brief a busy reader can act on, and a claim you
can defend out loud. And you can concede, gracefully, exactly where your
evidence stops. That last one is the whole course.
