## 1. Why This Matters: the 90-Second Reader

> *"I walk 60 posters in an hour. I stop at the ones whose claim I can read
> from four feet away. I fund the ones whose claim I still believe after I ask
> the hard question. If I have to hunt for what you found, I have already moved
> on."*
> — a **policy stakeholder** doing the rounds at a poster session

A poster is not a paper stapled to foam board. It is a claim delivery device
under a brutal constraint. A stranger gives it 90 seconds, reads it in whatever
order their eye chooses, and leaves with at most one claim. The route that
stranger's eye takes across the board is the **read path**, and this unit
teaches you to design it on purpose.

Two classes and one studio, three jobs. In Monday's class you decide what your
one claim is and rank the evidence that earns it. In Wednesday's class you
draft the poster and let trained readers attack it. At the Friday production
studio you build the final version to spec and submit it. That submission is
final. The poster you hand in at the studio is the poster strangers read at
the Expo.

Start with the idea the whole unit leans on. The **claim boundary** is the line
between what your evidence licenses you to say and what it does not. On a
poster, the boundary lives in the headline verb. "Is associated with" stays
inside the boundary for an observational design; "causes" steps over it.

The discipline this unit trains is **claim economy**. Your M09 pilot produced
more than a poster can hold. A **storyboard** is a one-page plan that fixes
what goes where before you build the real thing. Film directors sketch every
shot before the cameras roll; you will sketch every poster block before you
open any design software. Building the storyboard forces you to rank what
matters and cut what does not, including things you worked hard for. What
survives is one defensible claim and the shortest honest path to it.

> **A question that often comes up here:** *"My project has three interesting
> findings. Why do I only get one claim?"* Because the reader only takes one.
> You can put three findings on the board, but the eye elevates exactly one to
> "the point." If you do not choose which, the reader chooses for you, and they
> often pick the flashiest finding rather than the most defensible one. Ranking
> is you deciding, before a stranger does, which finding your poster is willing
> to be judged on.

## 2. The Six Principles of Evidence Communication

**Guiding question:** *what makes a chart, or a whole poster, tell the truth
efficiently?*

This section adapts the communication lecture Professor Moreira gives in every
course he teaches, tuned here for research posters. Six principles, each a
question you can run any chart or poster panel against:

| # | Principle | The question it asks |
|---|---|---|
| 1 | **Context matters** | *Compared to what?* What baseline turns this number into evidence? |
| 2 | **Visualization derives from data** | *What kind of data is this?* What chart respects its shape, and its claim boundary? |
| 3 | **Less is more: separate signal from noise** | *What can I delete and still understand the chart?* |
| 4 | **Hierarchy among data** | *Where does the eye land first? Should it?* |
| 5 | **Beauty counts: not all data are equally important** | *Which one number am I asking the reader to remember?* |
| 6 | **Telling your story** | *Can a non-expert restate the conclusion, with its boundary, in one sentence?* |

Principle 4 has a formal name: **visual hierarchy**. Visual hierarchy means
arranging size and position so the most important element gets seen first. A
newspaper front page does it with one giant headline above smaller columns.
Your poster does it by giving the headline claim the biggest type on the board.

Principle 3 has a name worth knowing too: the **data-ink ratio**. Of all the
ink on your chart, the data-ink ratio asks how much actually encodes data. A
gray background grid is ink that carries no data; the bars themselves are ink
that works. Watch the ratio double:

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/data_ink.png" width="85%"/></center>

*Same data, twice: the right panel deleted the background, the border, and the
grid clutter, and lost nothing a reader needs. (From Professor Moreira's
MGMT 47400 data-communication lecture, after Tufte's data-ink principle.)*

> **A question that often comes up here:** *"Isn't 'beauty counts' subjective,
> my aesthetic against yours?"* Not in this discipline. On a poster, "beauty"
> is not decoration. It is the visible result of principles 3, 4, and 5 applied
> together: nothing extra, a clear first landing point for the eye, and one
> number dressed to be remembered. Two well-trained readers rarely disagree
> about whether a panel achieves that. When they do, Wednesday's red-team
> round, where classmates attack your draft on purpose, settles it with
> evidence.

````python
# The data-ink cleanup, live: the same evidence drawn twice.
# Weekly response rates from a small pilot survey (illustrative, fixed values).
weeks = ["W1", "W2", "W3", "W4", "W5", "W6"]
response_rate = [62, 71, 58, 74, 69, 77]  # percent

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# Left: default-ish, ink everywhere
ax = axes[0]
ax.bar(weeks, response_rate, color="steelblue", edgecolor="black")
ax.set_facecolor("#dfeef0")
ax.grid(True, which="both", color="gray", linewidth=0.8)
ax.set_title("Pilot response rate (default ink)")
ax.set_ylabel("percent")

# Right: data-ink cleanup — delete what encodes nothing, elevate what matters
ax = axes[1]
bars = ax.bar(weeks, response_rate, color="0.6")
bars[-1].set_color("#2c7fb8")                      # hierarchy: the week that matters
ax.spines[["top", "right", "left"]].set_visible(False)
ax.tick_params(left=False)
ax.set_yticks([])
for b, v in zip(bars, response_rate):
    ax.annotate(f"{v}%", (b.get_x() + b.get_width()/2, v), ha="center",
                va="bottom", fontsize=10)
ax.set_title("Pilot response rate (data-ink cleanup)")

plt.tight_layout()
plt.show()
print("✓ Same numbers, twice — count what the right panel deleted, and what it kept.")
````

### The architecture of a research poster: eleven sections, one spine

The same principles scale from a chart to the whole board. A research-evidence
poster has a recognizable architecture: eleven sections, each with one job.
Your storyboard grid (next section) is how you decide what goes in each:

| # | Section | Its one job |
|---|---|---|
| 1 | **Title + author** | the claim's trailer: searchable, honest, 15 words or fewer |
| 2 | **The hook** | why a stranger should spend their 90 seconds here |
| 3 | **Question + compass position** | your research question (built to the four marks from Topic 1) and its declared compass position |
| 4 | **The gap** | one bounded sentence plus the 2–3 sources that anchor it |
| 5 | **Design in miniature** | your MIDA declaration (Model, Inquiry, Data strategy, Answer strategy), compressed to a panel |
| 6 | **Headline evidence** | THE figure, with principles 1–5 at full strength |
| 7 | **Supporting evidence** | the two blocks that survived your ranking |
| 8 | **Uncertainty & limitations** | visible on the board, not exiled to a footnote |
| 9 | **The claim** | one sentence, inside its boundary: what you defend |
| 10 | **What would change my mind** | the honest next step your evidence points to |
| 11 | **References + QR** | retrievable sources plus a code to your reproducibility package |

Sections 6 through 9 own the center of the board. That is where the 90-second
read path must land. Everything else is margin territory. The storyboard grid
you build next is these eleven jobs, fought over by your actual evidence.

## 3. The Storyboard Grid: Deciding the Read Path Before You Build

Your storyboard is milestone **M10**. You pitch it in today's class in 90
seconds, and you submit it tonight on Brightspace. It is a *plan for a poster*,
not the poster. On one page it fixes four things:

- **The headline claim.** The one sentence a 90-second reader must leave with,
  15 words or fewer, worded to match your compass position's boundary. A
  descriptive project describes; an associational project associates; only an
  identified causal design says *cause*. The reader never sees your footnote,
  so the boundary has to live in the headline itself.
- **Three evidence blocks, ranked by evidential value.** Evidential value is
  how much weight a piece of output carries for the headline claim. An effect
  estimate with its interval outweighs a photo of your field site. Order the
  blocks so the strongest sits nearest the claim, and write one line on why
  each block earns its rank. If you cannot say why block 1 beats block 3, you
  have not decided what the poster argues.
- **The uncertainty line.** The interval, frame limit, or measurement caveat
  that governs the headline. Place it on the read path, not in a bottom-corner
  footnote nobody reaches in 90 seconds.
- **The read-path arrows.** The order you intend the eye to travel: entry →
  headline → evidence 1 → 2 → 3 → uncertainty → exit.

All of this is poster real-estate economics. The top-center is your most
expensive space, and it goes to the claim, not the university logo. Every
block below competes for a reader who is already walking away.

> **A question that often comes up here:** *"Isn't ranking evidence just
> picking favorites?"* No. Ranking is an argument. Block 1 beats block 3 for a
> *stated reason*: it speaks more directly to the headline, rests on more data,
> or rules out more alternatives. Writing that reason down is how you find out
> whether your poster has a spine or just a pile of outputs.

### 🛠️ Render the storyboard-grid wireframe

The cell below draws the grid as labeled rectangles, with the read-path arrows
numbered in order. It is a visual template you can rearrange by editing the
coordinates. Move a block, and the reader's path moves with it. Run it, then
change one block's position or the order of the arrows and re-run. Feel how the
read path shifts.

> 💡 **Gemini Prompt:** "Here is a matplotlib cell that draws a poster
> storyboard as labeled rectangles with numbered arrows: [paste the next cell].
> Show me exactly which coordinates to edit to swap the positions of the
> uncertainty line and the second evidence block, and tell me what that swap
> would do to the order a reader meets them."
>
> **After running, verify (the responsible-AI habit):**
> - [ ] Make the edit yourself and re-run. The rendered wireframe, not Gemini's
>   description of it, is the proof the swap worked.
> - [ ] Check its reading-order claim against the arrow numbers you can see.
> - [ ] Log this use in your AI ledger: tool, task, verification.

````python
# A storyboard-grid wireframe — rearrange it by editing the coordinates below.
from matplotlib.patches import Rectangle

fig, ax = plt.subplots(figsize=(11, 7.5))
ax.set_xlim(0, 12)
ax.set_ylim(0, 9)
ax.axis("off")

def block(x, y, w, h, label, fc, ec):
    ax.add_patch(Rectangle((x, y), w, h, facecolor=fc, edgecolor=ec, linewidth=1.6))
    ax.text(x + w / 2, y + h / 2, label, ha="center", va="center", fontsize=9.5)

# Headline strip: the most expensive real estate, spent on the ONE claim.
block(0.4, 7.5, 11.2, 1.1,
      "HEADLINE CLAIM   ( <= 15 words, inside your compass position's boundary )",
      "#D7E3F0", "#3B5C86")

# Three evidence blocks, ranked left-to-right by evidential value.
evidence_blocks = [
    (0.4,  4.2, "EVIDENCE 1\nstrongest\n(nearest the claim)"),
    (4.35, 4.2, "EVIDENCE 2\nsupporting"),
    (8.3,  4.2, "EVIDENCE 3\nweakest;\ncut first if tight"),
]
for x, y, label in evidence_blocks:
    block(x, y, 3.3, 2.7, label, "#EAEFF5", "#4C72B0")

# Uncertainty strip: ON the read path, never a footnote.
block(0.4, 2.7, 11.2, 1.0,
      "UNCERTAINTY LINE   ( interval / frame limit — the reader MEETS it )",
      "#F3EAD6", "#C89B3C")

# Entry and exit: where the eye arrives, and the one claim it carries away.
block(0.4, 0.9, 2.6, 1.1, "ENTRY\neye lands here", "#E4F0E4", "#4C9C4C")
block(9.0, 0.9, 2.6, 1.1, "EXIT\nthe one claim\nthey carry away", "#E4F0E4", "#4C9C4C")

# The read path: the order you INTEND the stranger's eye to travel.
read_path = [(1.7, 2.0), (6.0, 8.05), (2.05, 5.55),
             (6.0, 5.55), (9.95, 5.55), (6.0, 3.2), (10.3, 2.0)]
for i in range(len(read_path) - 1):
    ax.annotate("", xy=read_path[i + 1], xytext=read_path[i],
                arrowprops=dict(arrowstyle="->", color="#B4462E", lw=1.8,
                                connectionstyle="arc3,rad=0.08"))
    mx = (read_path[i][0] + read_path[i + 1][0]) / 2
    my = (read_path[i][1] + read_path[i + 1][1]) / 2
    ax.text(mx, my, str(i + 1), color="#B4462E", fontsize=11, fontweight="bold",
            ha="center", va="center",
            bbox=dict(boxstyle="circle", fc="white", ec="#B4462E"))

ax.set_title("Storyboard grid — a template you rearrange by editing the coordinates",
             fontsize=11)
plt.tight_layout()
plt.show()

n_arrows = len(read_path) - 1
print(f"Blocks drawn: 1 headline + {len(evidence_blocks)} ranked evidence + "
      f"1 uncertainty + entry/exit.")
print(f"Read path: {n_arrows} arrows, entry -> headline -> evidence 1-2-3 "
      f"-> uncertainty -> exit.")

# Self-check: the structure a URC poster storyboard commits to.
assert len(evidence_blocks) == 3, "a poster storyboard ranks exactly three evidence blocks"
assert n_arrows == 6, "the read path should connect all seven waypoints in order"
print("✓ Wireframe built — one claim, three ranked blocks, uncertainty on the path.")
````

## 4. Red-Team: Attack Your Own Poster Before Strangers Do

> *"Every flaw a friendly room finds today is a flaw a hostile stranger never
> sees at the Expo. I would rather be embarrassed by you now than by an Expo
> visitor later. So come at my poster like you are trying to sink it."*
> — a **skeptical peer** on red-team day

A **red-team** is a group whose assigned job is to attack your work on purpose,
so the weaknesses surface while you can still fix them. Banks hire people to
break into their own vaults; today, your classmates break into your poster.
Adversarial review is cheaper than public embarrassment. Before a stranger at
the Expo finds the weakest thing on your board, a room of trained-hostile
readers finds it first.

This class is milestone **M11**: your draft plus the red-team, submitted
Wednesday night on Brightspace. You run a four-audit protocol against two
classmates' posters, and two reviewers run it against yours. The four audits
are the four ways a poster fails:

1. **Claim boundary.** Does any claim, the headline most of all, outrun the
   compass position's evidence? Concrete check: circle every verb. Does any of
   them (*causes, improves, boosts, leads to*) claim more than the design
   licenses?
2. **Figure honesty.** Does any figure's visual impression say more than the
   numbers? Concrete check: where does each axis start? Is a small gap drawn
   as a chasm? Does a color or a 3-D effect editorialize?
3. **Read path.** Can a stranger find the headline in 90 seconds? Concrete
   check: walk it cold, out loud, and mark the first spot your eye got lost or
   doubled back.
4. **Uncertainty.** Is it visible, or hidden? Concrete check: point to the
   interval or frame limit *on the read path*. If you can only point to a
   footnote, that is a hit.

The norm that makes this work: **hard on the poster, warm on the person.** A
rigorous audit names *where* and *why*. "My eye went headline, then Figure 3,
then back to the method box; I never reached your uncertainty line; move it
under the headline." Never "looks good, maybe add color." During your own walk
you stay **silent**. You learn what a real reader sees only when you cannot
defend yourself.

> **A question that often comes up here:** *"What if my reviewer just didn't
> get my project?"* Then that IS a read-path finding about your poster, not a
> complaint about your reviewer. A trained reader who misreads your poster is
> telling you a stranger will too. The defensive move, "they didn't understand
> it," is the single fastest way to waste a free review.

### 🛠️ Build the flawed figure, then audit it

The cell constructs a bar chart with three deliberate flaws, each labeled in a
comment: a **truncated axis** (a figure-honesty flaw), an **overclaiming causal
title** (a claim-boundary flaw), and **no uncertainty shown** (an uncertainty
flaw). Nothing here is a trick of the data. The flaws live in how honest
numbers are *drawn and described*. Run it, then bring the four audits to it.

````python
# A deliberately flawed mini-poster figure — every flaw is labeled, by design.
groups = ["Program A", "Program B"]
scores = [3.1, 3.5]          # a 0.4-point gap on a 1-5 satisfaction scale

fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(groups, scores, color=["#8899AA", "#C44E52"], edgecolor="white")
ax.set_ylim(3.0, 3.6)                          # FLAW 1 (figure honesty): truncated axis
ax.set_ylabel("Satisfaction")                  # no scale range shown
ax.set_title("Program B BOOSTS satisfaction")  # FLAW 2 (claim boundary): a causal verb
# FLAW 3 (uncertainty): no interval, no error bar, no frame note anywhere.
plt.tight_layout()
plt.show()

floor = 3.0
exaggeration = (scores[1] - floor) / (scores[0] - floor)
print(f"Truncated at {floor}: the taller bar is drawn {exaggeration:.1f}x bigger "
      f"than the shorter one — for a real gap of {scores[1] - scores[0]:.1f} "
      f"on a 1-5 scale.")

title_text = "Program B BOOSTS satisfaction"
# Self-check: confirm the flaws are really there (this is the answer key).
assert ax.get_ylim()[0] > 0, "the axis should start above zero — that IS the deliberate truncation"
assert any(w in title_text.lower() for w in ("boost", "cause", "improve", "lead")), \
    "the title should overclaim with a causal verb, by design"
print("✓ Three deliberate flaws: truncated axis, causal title, and no uncertainty shown.")
````

### Traceability, demonstrated

Before you claim a number is traceable, feel what that means. Below is the kind
of cell that *produces* a headline number. When a partner, or Professor
Moreira, or a stranger points at the poster and asks "where did that come
from?", you open the notebook and point *here*. Run it. The assertion is the
whole idea: the poster's number must equal this cell's output, or it does not
go on the board.

````python
# Number-to-notebook traceability — every poster number must trace to a cell.
# Say your headline reports the share of people who showed the outcome. This is the
# cell that produced it — the one you point to when someone asks.
responses = np.array([1] * 38 + [0] * 62)     # 100 people; 38 showed the outcome
point = responses.mean()
se = np.sqrt(point * (1 - point) / len(responses))
lo, hi = point - 1.96 * se, point + 1.96 * se
print(f"Poster reads:  {point:.0%}   (95% CI {lo:.0%} to {hi:.0%})")

# Self-check: the poster number IS this cell's output — that is what traceable means.
assert round(point, 2) == 0.38, "the poster number must equal this cell's output — that is traceability"
print("✓ Traceable: 38% was produced here, not typed from memory or carried from a draft.")
````
