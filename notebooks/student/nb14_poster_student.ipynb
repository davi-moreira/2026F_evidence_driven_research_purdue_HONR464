## 1. Why This Matters — the 90-Second Reader

> *"I walk 60 posters in an hour. I stop at the ones whose one claim I can read
> from four feet away, and I fund the ones whose claim I still believe after I
> ask the hard question. If I have to hunt for what you found, I've already moved
> on."*
> — a **policy stakeholder** doing the rounds at your poster session

A poster is not a paper stapled to foam board. It is a **claim delivery device**
operating under a brutal constraint: a stranger gives it 90 seconds, reads it in
whatever order their eye chooses, and leaves with — at most — **one claim**. Your
job across the next three meetings is to decide what that one claim is, rank the
evidence that earns it, place the uncertainty where the reader actually meets it,
and then build the thing to spec and submit it. Friday, **November 6, is the real
production deadline** — the poster you submit that night is the poster strangers
read on November 17. There is no draft after it.

The discipline this unit trains is **claim economy**. Your M09 pilot produced
more than a poster can hold, and the storyboard forces you to rank what matters
and cut what doesn't — including things you worked hard for. What survives is one
defensible claim and the shortest honest path to it.

> **A question that often comes up here:** *"My project has three interesting
> findings. Why do I only get one claim?"* Because the reader only takes one. You
> can put three findings on the board, but the eye elevates exactly one to
> "the point," and if you don't choose which, the reader chooses for you — often
> the flashiest, not the most defensible. Ranking is you deciding, before a
> stranger does, which finding your poster is willing to be judged on.

## 2. The Six Principles of Evidence Communication

**Guiding question:** *what makes a chart — or a whole poster — tell the truth
efficiently?*

This section is the communication lecture Professor Moreira gives in every
course he teaches, adapted here for research posters. Six principles, each a
question you can run any chart or poster panel against:

| # | Principle | The question it asks |
|---|---|---|
| 1 | **Context matters** | *Compared to what?* What baseline turns this number into evidence? |
| 2 | **Visualization derives from data** | *What kind of data is this?* What chart respects its shape — and its claim boundary? |
| 3 | **Less is more: separate signal from noise** | *What can I delete and still understand the chart?* |
| 4 | **Hierarchy among data** | *Where does the eye land first? Should it?* |
| 5 | **Beauty counts: not all data are equally important** | *Which one number am I asking the reader to remember?* |
| 6 | **Telling your story** | *Can a non-expert restate the conclusion — with its boundary — in one sentence?* |

Principle 3 has a name worth knowing: the **data-ink ratio** — of all the ink
on your chart, how much actually encodes data? Watch it double:

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/data_ink.png" width="85%"/></center>

*Same data, twice: the right panel deleted the background, the border, and the
grid clutter — and lost nothing a reader needs. (From Professor Moreira's
MGMT 47400 data-communication lecture, after Tufte's data-ink principle.)*

> **A question that often comes up here:** *"Isn't 'beauty counts' subjective —
> my aesthetic against yours?"* Not in this discipline. On a poster, "beauty"
> is not decoration; it is the visible result of principles 3–5 applied
> together: nothing extra, a clear first landing point for the eye, and one
> number dressed to be remembered. Two well-trained readers rarely disagree
> about whether a panel achieves that — and when they do, your red-team round
> on Wednesday will settle it with evidence.

````python
# The data-ink cleanup, live: the same evidence drawn twice.
# Weekly response rates from a small pilot survey (illustrative, fixed values).
weeks = ["W1", "W2", "W3", "W4", "W5", "W6"]
response_rate = [62, 71, 58, 74, 69, 77]  # percent

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# Left: default-ish, ink everywhere
ax = axes[0]
ax.bar(weeks, response_rate, color="steelblue", edgecolor="black")
ax.set_facecolor("#dfeef0")
ax.grid(True, which="both", color="gray", linewidth=0.8)
ax.set_title("Pilot response rate (default ink)")
ax.set_ylabel("percent")

# Right: data-ink cleanup — delete what encodes nothing, elevate what matters
ax = axes[1]
bars = ax.bar(weeks, response_rate, color="0.6")
bars[-1].set_color("#2c7fb8")                      # hierarchy: the week that matters
ax.spines[["top", "right", "left"]].set_visible(False)
ax.tick_params(left=False)
ax.set_yticks([])
for b, v in zip(bars, response_rate):
    ax.annotate(f"{v}%", (b.get_x() + b.get_width()/2, v), ha="center",
                va="bottom", fontsize=10)
ax.set_title("Pilot response rate (data-ink cleanup)")

plt.tight_layout()
plt.show()
print("✓ Same numbers, twice — count what the right panel deleted, and what it kept.")
````

### The architecture of a research poster — eleven sections, one spine

The same principles scale from a chart to the whole board. A research-evidence
poster has a recognizable architecture — eleven sections, each with one job.
Your storyboard grid (next section) is how you'll decide what goes in each:

| # | Section | Its one job |
|---|---|---|
| 1 | **Title + author** | the claim's trailer — searchable, honest, ≤15 words |
| 2 | **The hook** | why a stranger should spend their 90 seconds here |
| 3 | **Question + compass position** | the four-mark question and its declared compass position |
| 4 | **The gap** | one bounded sentence + the 2–3 sources that anchor it |
| 5 | **Design in miniature** | your MIDA declaration, compressed to a panel |
| 6 | **Headline evidence** | THE figure — principles 1–5 at full strength |
| 7 | **Supporting evidence** | the two blocks that survived your ranking |
| 8 | **Uncertainty & limitations** | visible on the board, not exiled to a footnote |
| 9 | **The claim** | one sentence, inside its boundary — what you defend |
| 10 | **What would change my mind** | the honest next step (your ULN's N) |
| 11 | **References + QR** | retrievable sources + a code to your reproducibility package |

Sections 6–9 own the center of the board — that is where the 90-second read
path must land. Everything else is margin territory. The storyboard grid you
build next is these eleven jobs, fought over by your actual evidence.

## 3. The Storyboard Grid — Deciding the Read Path Before You Build

The storyboard (milestone **M10**, due **Monday, November 2, 11:59 PM**, pitched
in class today) is a *plan for a poster*, not the poster. On one page it fixes
four things:

- **The headline claim** — the one sentence a 90-second reader must leave with,
  **≤ 15 words**, worded to match your compass position's boundary. A descriptive project
  describes; an associational project associates; only an identified causal design
  says *cause*. The reader never sees your footnote, so the boundary has to live
  in the headline itself.
- **Three evidence blocks, ranked by evidential value** — ordered so the **most
  probative sits closest to the claim**. For each, one line on *why it earns its
  rank*. If you can't say why block 1 beats block 3, you haven't decided what the
  poster argues.
- **The uncertainty line** — the interval, the frame limit, or the measurement
  caveat that governs the headline, placed **on** the read path, not exiled to a
  bottom-corner footnote nobody reaches in 90 seconds.
- **The read-path arrows** — the order you intend the eye to travel: entry →
  headline → evidence 1 → 2 → 3 → uncertainty → exit.

All of this is **poster real-estate economics**: the top-center is your most
expensive space, and it goes to the claim, not the university logo. Every block
below competes for a reader who is already walking away.

> **A question that often comes up here:** *"Isn't ranking evidence just picking
> favorites?"* No — ranking is an argument. Block 1 beats block 3 for a *stated
> reason*: it speaks more directly to the headline, rests on more data, or rules
> out more alternatives. Writing that reason down is how you find out whether your
> poster actually has a spine or just a pile of outputs.

### 🛠️ Render the storyboard-grid wireframe

The cell below draws the grid as labeled rectangles with the read-path arrows
numbered in order. It is a **visual template you can rearrange by editing the
coordinates** — move a block, and the reader's path moves with it. Run it, then
try changing one block's position or the order of the arrows and re-running to
feel how the read path shifts.

> 💡 **Gemini Prompt:** "Here is a matplotlib cell that draws a poster
> storyboard as labeled rectangles with numbered arrows: [paste the next
> cell]. Show me exactly which coordinates to edit to swap the positions of
> the uncertainty line and the second evidence block — and tell me what that
> swap would do to the order a reader meets them."
>
> **After running, verify (the responsible-AI habit):**
> - [ ] Make the edit yourself and re-run — the rendered wireframe, not
>   Gemini's description of it, is the proof the swap worked.
> - [ ] Check its reading-order claim against the arrow numbers you can see.
> - [ ] Log this use in your AI ledger: tool, task, verification.

````python
# A storyboard-grid wireframe — rearrange it by editing the coordinates below.
from matplotlib.patches import Rectangle

fig, ax = plt.subplots(figsize=(11, 7.5))
ax.set_xlim(0, 12)
ax.set_ylim(0, 9)
ax.axis("off")

def block(x, y, w, h, label, fc, ec):
    ax.add_patch(Rectangle((x, y), w, h, facecolor=fc, edgecolor=ec, linewidth=1.6))
    ax.text(x + w / 2, y + h / 2, label, ha="center", va="center", fontsize=9.5)

# Headline strip: the most expensive real estate, spent on the ONE claim.
block(0.4, 7.5, 11.2, 1.1,
      "HEADLINE CLAIM   ( <= 15 words, inside your compass position's boundary )",
      "#D7E3F0", "#3B5C86")

# Three evidence blocks, ranked left-to-right by evidential value.
evidence_blocks = [
    (0.4,  4.2, "EVIDENCE 1\nmost probative\n(nearest the claim)"),
    (4.35, 4.2, "EVIDENCE 2\nsupporting"),
    (8.3,  4.2, "EVIDENCE 3\nweakest —\ncut first if tight"),
]
for x, y, label in evidence_blocks:
    block(x, y, 3.3, 2.7, label, "#EAEFF5", "#4C72B0")

# Uncertainty strip: ON the read path, never a footnote.
block(0.4, 2.7, 11.2, 1.0,
      "UNCERTAINTY LINE   ( interval / frame limit — the reader MEETS it )",
      "#F3EAD6", "#C89B3C")

# Entry and exit: where the eye arrives, and the one claim it carries away.
block(0.4, 0.9, 2.6, 1.1, "ENTRY\neye lands here", "#E4F0E4", "#4C9C4C")
block(9.0, 0.9, 2.6, 1.1, "EXIT\nthe one claim\nthey carry away", "#E4F0E4", "#4C9C4C")

# The read path: the order you INTEND the stranger's eye to travel.
read_path = [(1.7, 2.0), (6.0, 8.05), (2.05, 5.55),
             (6.0, 5.55), (9.95, 5.55), (6.0, 3.2), (10.3, 2.0)]
for i in range(len(read_path) - 1):
    ax.annotate("", xy=read_path[i + 1], xytext=read_path[i],
                arrowprops=dict(arrowstyle="->", color="#B4462E", lw=1.8,
                                connectionstyle="arc3,rad=0.08"))
    mx = (read_path[i][0] + read_path[i + 1][0]) / 2
    my = (read_path[i][1] + read_path[i + 1][1]) / 2
    ax.text(mx, my, str(i + 1), color="#B4462E", fontsize=11, fontweight="bold",
            ha="center", va="center",
            bbox=dict(boxstyle="circle", fc="white", ec="#B4462E"))

ax.set_title("Storyboard grid — a template you rearrange by editing the coordinates",
             fontsize=11)
plt.tight_layout()
plt.show()

n_arrows = len(read_path) - 1
print(f"Blocks drawn: 1 headline + {len(evidence_blocks)} ranked evidence + "
      f"1 uncertainty + entry/exit.")
print(f"Read path: {n_arrows} arrows, entry -> headline -> evidence 1-2-3 "
      f"-> uncertainty -> exit.")

# Self-check: the structure a URC poster storyboard commits to.
assert len(evidence_blocks) == 3, "a poster storyboard ranks exactly three evidence blocks"
assert n_arrows == 6, "the read path should connect all seven waypoints in order"
print("✓ Wireframe built — one claim, three ranked blocks, uncertainty on the path.")
````

## 4. Red-Team — Attack Your Own Poster Before Strangers Do

> *"Every flaw a friendly room finds today is a flaw a hostile stranger doesn't
> find on November 17. I would rather be embarrassed by you now than by a Expo
> visitor then — so come at my poster like you're trying to sink it."*
> — a **skeptical peer** on red-team day

Before a stranger at the Expo finds the weakest thing on your poster, a room of
trained-hostile classmates finds it first — while there is still time to fix it.
That is the whole logic of a **red-team**: adversarial review is cheaper than
public embarrassment. Today (milestone **M11**, draft + red-team, **due
Wednesday, November 4, 11:59 PM**) you run a four-audit protocol against two
classmates' posters, and two reviewers run it against yours. The four audits are
the four ways a poster fails:

1. **Claim boundary** — does any claim (the headline most of all) outrun the
   compass position's evidence? Concrete check: circle every verb; does any of them
   (*causes, improves, boosts, leads to*) claim more than the design licenses?
2. **Figure honesty** — does any figure's *visual impression* say more than the
   numbers? Concrete check: where does each axis start; is a small gap drawn as a
   chasm; does a color or a 3-D effect editorialize?
3. **Read path** — can a stranger find the headline in 90 seconds? Concrete
   check: walk it cold, out loud, and mark the first spot your eye got lost or
   doubled back.
4. **Uncertainty** — is it visible, or hidden? Concrete check: can you point to
   the interval or frame limit *on the read path*, or is it a footnote?

The norm that makes this work: **hard on the poster, warm on the person.** A
rigorous audit names *where* and *why* — "my eye went headline → Figure 3 → back
to the method box; I never reached your uncertainty line; move it under the
headline" — not "looks good, maybe add color." During your own walk you stay
**silent**: you learn what a real reader sees only when you can't defend yourself.

> **A question that often comes up here:** *"What if my reviewer just didn't get
> my project?"* Then that IS a read-path finding about your poster, not a
> complaint about your reviewer. A trained reader who misreads your poster is
> telling you a stranger will too. The defensive move — "they didn't understand
> it" — is the single fastest way to waste a free review.

### 🛠️ Build the flawed figure, then audit it

The cell constructs a bar chart with three planted flaws, each labeled in a
comment: a **truncated axis** (figure-honesty flaw), an **overclaiming causal
title** (claim-boundary flaw), and **no uncertainty shown** (uncertainty flaw).
Nothing here is a trick of the data — the flaws are all in how honest numbers are
*drawn and described*. Run it, then bring the four audits to it.

````python
# A deliberately flawed mini-poster figure — every flaw is labeled, by design.
groups = ["Program A", "Program B"]
scores = [3.1, 3.5]          # a 0.4-point gap on a 1-5 satisfaction scale

fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(groups, scores, color=["#8899AA", "#C44E52"], edgecolor="white")
ax.set_ylim(3.0, 3.6)                          # FLAW 1 (figure honesty): truncated axis
ax.set_ylabel("Satisfaction")                  # no scale range shown
ax.set_title("Program B BOOSTS satisfaction")  # FLAW 2 (claim boundary): a causal verb
# FLAW 3 (uncertainty): no interval, no error bar, no frame note anywhere.
plt.tight_layout()
plt.show()

floor = 3.0
exaggeration = (scores[1] - floor) / (scores[0] - floor)
print(f"Truncated at {floor}: the taller bar is drawn {exaggeration:.1f}x bigger "
      f"than the shorter one — for a real gap of {scores[1] - scores[0]:.1f} "
      f"on a 1-5 scale.")

title_text = "Program B BOOSTS satisfaction"
# Self-check: confirm the flaws are actually planted (this is the answer key).
assert ax.get_ylim()[0] > 0, "the axis should start above zero — that IS the planted truncation"
assert any(w in title_text.lower() for w in ("boost", "cause", "improve", "lead")), \
    "the title should overclaim with a causal verb, by design"
print("✓ Three planted flaws: truncated axis, causal title, and no uncertainty shown.")
````

### Traceability, demonstrated

Before you claim a number is traceable, feel what that means: below is the kind of
cell that *produces* a headline number, so that when a partner (or Professor
Moreira, or a stranger) points at the poster and asks "where did that come from?"
you open the notebook and point *here*. Run it — the assertion is the whole idea:
the poster's number must equal this cell's output, or it does not go on the
board.

````python
# Number-to-notebook traceability — every poster number must trace to a cell.
# Say your headline reports the share of people who showed the outcome. This is the
# cell that produced it — the one you point to when someone asks.
responses = np.array([1] * 38 + [0] * 62)     # 100 people; 38 showed the outcome
point = responses.mean()
se = np.sqrt(point * (1 - point) / len(responses))
lo, hi = point - 1.96 * se, point + 1.96 * se
print(f"Poster reads:  {point:.0%}   (95% CI {lo:.0%} to {hi:.0%})")

# Self-check: the poster number IS this cell's output — that is what traceable means.
assert round(point, 2) == 0.38, "the poster number must equal this cell's output — that is traceability"
print("✓ Traceable: 38% was produced here, not typed from memory or carried from a draft.")
````
