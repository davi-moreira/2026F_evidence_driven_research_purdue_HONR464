```python
# Minimum-viable-analysis scaffold — edit the EDIT ME block to use YOUR own data.
# It computes ONE estimate (a difference in means) and a randomization-inference
# interval around it. Every label flows from a single source of truth below, so
# renaming a column in ONE place updates the whole analysis; there is no second
# hardcoded name to fall out of sync.

# --- Single source of truth: name your columns and arms ONCE. -------------
GROUP_COL   = "arm"        # EDIT ME — the column that says which group a unit is in
OUTCOME_COL = "outcome"    # EDIT ME — the column holding the number you measure
TREATED     = "treated"    # EDIT ME — the label for your treatment / exposed group
CONTROL     = "control"    # EDIT ME — the label for your comparison group

# --- PLACEHOLDER data: replace this whole block with YOUR data. -----------
# A seeded stand-in so the cell runs before your real data arrive. Swap it for
# load_course_data(...) or your own CSV, keeping the column names above. This
# placeholder is NOT evidence about anything; it only proves the pipeline runs.
n_per_arm = 60             # EDIT ME — your per-group sample size
true_gap  = 3.0            # EDIT ME — used only by the placeholder generator
df = pd.DataFrame({
    GROUP_COL:   [TREATED] * n_per_arm + [CONTROL] * n_per_arm,
    OUTCOME_COL: np.concatenate([rng.normal(true_gap, 10, n_per_arm),
                                 rng.normal(0.0,      10, n_per_arm)]),
})
# --------------------------------------------------------------------------

# --- The estimate: a difference in means (treated minus control). ---------
treated_vals = df.loc[df[GROUP_COL] == TREATED, OUTCOME_COL].to_numpy()
control_vals = df.loc[df[GROUP_COL] == CONTROL, OUTCOME_COL].to_numpy()
observed_diff = treated_vals.mean() - control_vals.mean()

# --- The interval: randomization inference under the sharp null of NO effect.
# Re-shuffle who is labeled "treated" many times assuming the treatment does
# nothing, and read the spread of differences pure chance produces. (Mirrors the
# method you built in nb08.)
pooled = df[OUTCOME_COL].to_numpy()
n_treated = len(treated_vals)
REPS = 2000
null_diffs = np.empty(REPS)
for i in range(REPS):
    shuffled = rng.permutation(pooled)
    null_diffs[i] = shuffled[:n_treated].mean() - shuffled[n_treated:].mean()

se = null_diffs.std(ddof=1)
lo, hi = observed_diff - 1.96 * se, observed_diff + 1.96 * se
p_two_sided = float(np.mean(np.abs(null_diffs) >= abs(observed_diff)))

print(f"Estimate (difference in means): {observed_diff:+.2f}")
print(f"Randomization 95% interval:     [{lo:+.2f}, {hi:+.2f}]")
print(f"Two-sided randomization p:      {p_two_sided:.3f}")
print("✓ Ran on PLACEHOLDER data — swap in your own before you report anything.")
```
