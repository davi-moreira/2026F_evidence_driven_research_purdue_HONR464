## 1. The Gallery Walk: Reading Every Poster as a Stranger Would

**Guiding question:** *walking the room as a stranger would, does each poster's
evidence actually earn its headline, and can everyone read it?*

> *"I walk sixty posters in an hour. I stop at the ones whose claim I can read
> from four feet away, and I believe the ones whose claim survives one hard
> question. If your figure flatters your data, I catch it, and then I distrust
> the whole board."*
> — a **skeptical peer** doing the rounds at a poster session

A **gallery-walk audit** is a structured pass through every poster in the room,
where you stop at each one and run the same fixed checklist instead of glancing
and moving on. It turns vague admiration into located, fixable findings. The core
of it is **the four audits**, four ways a poster fails, each a question you ask of
one specific poster:

- **Claim-boundary audit.** Does the headline stay inside the poster's compass
  position? A **claim boundary** is the line between what the evidence licenses
  and what it does not. Example: an observational project may say "is associated
  with," never "causes." The boundary lives in the headline verb.
- **Figure-honesty audit.** Does the figure's visual impression match its number?
  **Figure honesty** means the picture cannot say more than the data. Example: an
  axis that starts just below the data draws a tiny gap as a canyon.
- **Read-path audit.** Can a stranger find the headline in 90 seconds? The **read
  path** is the order a reader's eye actually travels. Example: the eye should
  reach the claim before the methods box, not after.
- **Uncertainty audit.** Is the bound visible, or buried? **Uncertainty on the
  page** means the interval or caveat sits in the same eye-span as the claim.
  Example: a "±0.7" that lives only in a footnote is doing hidden work.

A **hit** is a specific, located finding with a fix: "the y-axis starts just below
the shorter bar, so a gap of about half a point looks like a canyon; run it 1 to
5." A review with no hits did not spare the poster. It failed it.

> **A question that often comes up here:** *"Isn't a gallery walk just being
> harsh about my classmates' work?"* No. The rule of the room is **hard on the
> poster, warm on the person.** Every hit a friendly room finds today is a hit a
> hostile stranger never gets to find at the Expo. You are giving each other a
> free stranger's eyes, which is the most useful gift a draft can get.

Two more checks ride alongside the four audits, and they decide whether your
claim reaches *every* reader, not just a fast sighted one.

- **Visual design.** How the layout guides the eye: size, position, and contrast.
  The one number you want remembered gets the biggest type and the top-center
  slot. Example: the headline claim, not the university logo, owns the center.
- **Accessibility.** Whether the poster's message reaches readers with different
  perception. **Accessibility** means the claim is legible to a color-blind reader
  and to a screen-reader user, not only to someone who sees every color and leans
  in close. Example: two series told apart by red versus green alone vanish for a
  red-green color-blind reader, who is roughly 1 in 12 men.

The fix for the color trap has a name worth knowing. **Redundant encoding** means
carrying the same distinction in more than one channel, so no single channel is
load-bearing. Example: label each bar with its value and give each a distinct
pattern, so color is a bonus, not the only signal.

> **A question that often comes up here:** *"Accessibility sounds like a
> compliance checkbox. Is it really an evidence issue?"* It is the same issue as
> every other audit. Your claim is only as strong as the readers it reaches. A
> figure that encodes its result in color alone makes a claim that a color-blind
> reviewer literally cannot receive, which is an overclaim by exclusion. Reaching
> every reader is not politeness. It is the claim doing its job.

```python
# A poster's headline figure, drawn two ways from the SAME pilot numbers.
# LEFT = "as drawn" (the flaws a gallery walk must catch).
# RIGHT = "as it should read" (honest axis, labels, uncertainty, reachable colors).

# A small satisfaction pilot on a 1-to-5 scale, two program cohorts.
scale = np.array([1, 2, 3, 4, 5])
p_a = np.array([0.16, 0.22, 0.30, 0.20, 0.12])   # cohort A: lower satisfaction
p_b = np.array([0.08, 0.14, 0.26, 0.26, 0.26])   # cohort B: a little higher
a = rng.choice(scale, size=200, p=p_a)
b = rng.choice(scale, size=200, p=p_b)

mean_a, mean_b = a.mean(), b.mean()
ci_a = 1.96 * a.std(ddof=1) / np.sqrt(len(a))     # 95% interval half-widths:
ci_b = 1.96 * b.std(ddof=1) / np.sqrt(len(b))     # the uncertainty the honest panel shows
gap = mean_b - mean_a

# One source of truth for the truncation: a floor just below the shorter bar.
floor = min(mean_a, mean_b) - 0.05
exaggeration = (mean_b - floor) / (mean_a - floor)

print(f"Cohort A mean = {mean_a:.2f}   Cohort B mean = {mean_b:.2f}   gap = {gap:.2f} on a 1-to-5 scale")
print(f"Truncated at {floor:.2f}, the taller bar is drawn {exaggeration:.1f}x the shorter one.")

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))

# LEFT: "as drawn." Bring the audits to it YOURSELF in the exercise below; these
# comments only say what the code does, never which audit each choice trips.
axL = axes[0]
axL.bar(["Program A", "Program B"], [mean_a, mean_b],
        color=["#D62728", "#2CA02C"])              # two bars told apart by red vs green only
axL.set_ylim(floor, mean_b + 0.05)                 # y-axis floored just below the shorter bar
axL.set_title("Program B BOOSTS satisfaction")     # title worded the way a program might pitch it
axL.set_ylabel("Satisfaction")
# no interval drawn, no value labels, no full scale shown.

# RIGHT: as it should read — honest, and reachable by every reader.
axR = axes[1]
bars = axR.bar(["Program A", "Program B"], [mean_a, mean_b],
               yerr=[ci_a, ci_b], capsize=6,
               color=["#4C72B0", "#DD8452"],        # colorblind-distinguishable pair
               hatch=["", "//"], edgecolor="white")  # redundant encoding: pattern backs up color
axR.set_ylim(1, 5)                                   # honest scale: the real 1-to-5 range
for bar, m in zip(bars, [mean_a, mean_b]):
    axR.annotate(f"{m:.2f}", (bar.get_x() + bar.get_width() / 2, m + 0.10),
                 ha="center", fontsize=11)           # direct labels: a screen reader gets the number
axR.set_ylabel("Mean satisfaction (1 = low, 5 = high)")
axR.set_title("Program A vs B: means with 95% intervals")

plt.tight_layout()
plt.show()

# Self-check: the flaws and the honest reading are both really present.
assert mean_b > mean_a, "cohort B should read a little higher in this pilot"
assert exaggeration > 3, "the truncated axis should visibly inflate the gap"
assert axes[0].get_ylim()[0] > 1, "the left axis is truncated by design"
assert axes[1].get_ylim() == (1.0, 5.0), "the right axis shows the honest full scale"
print("✓ Same numbers, drawn two ways: the left panel overclaims; the right one reads honestly.")
```

## 2. The Oral Defense: Three Lenses, One Poster

**Guiding question:** *standing at your poster, can you answer the outsider, the
methodologist, and the skeptic, and concede gracefully where you must?*

> *"I ask three kinds of question at a poster, and I learn the most from the one
> the author cannot answer. A confident bluff tells me to distrust everything. A
> clean 'here is what I know, here is where it stops' tells me to trust the rest."*
> — a **methods reviewer** working a poster session

An **oral poster defense** is the live conversation at your board, where you
answer questions you cannot see coming and cannot edit afterward. The written
poster hid your gaps behind layout. The spoken defense exposes them. Three lenses
account for almost every hard question you will get, and each needs a different
answer:

- **The interdisciplinary lens (the outsider).** Someone outside your field who
  does not share your vocabulary. Their question tests whether your claim and its
  relevance survive translation. Example: "I study materials, not surveys. Why
  should I care what your number means?"
- **The methods lens (the specialist).** Someone who knows your tools and probes a
  choice. Their question tests whether your **specification**, the specific design
  and analysis choices you made, was defensible. Example: "why this cutoff, this
  measure, this comparison group?"
- **The skeptic lens (the doubter).** Someone who grants your data and proposes an
  **alternative explanation**, a different story that fits the same evidence. Their
  question tests your claim boundary. Example: "couldn't a third thing cause both
  of these?"

> **A question that often comes up here:** *"If I just knew my project cold, would
> I not answer all three the same way?"* No, and this is the trap. Knowing your
> project answers the outsider and often the specialist. The skeptic is different:
> a good alternative explanation is not a knowledge gap you can study your way out
> of at the poster. It is a boundary question, and the honest answer is often to
> concede the boundary, not to defend past it.

Two moves turn a hard question into a credible answer.

The **honest-don't-know formula** is what a strong "I don't know" contains. It is
not a shrug. It names what you *do* know, states the boundary plainly, and points
to what evidence would settle the rest. Example: "my design shows these two move
together, but it cannot rule out a common cause, so I would not claim direction
without a randomized follow-up." That answer concedes and stays expert at once.

**Defend-versus-concede** is the judgment call underneath every hard question.
You **defend** when the objection asks for something your evidence already
licenses, and you hold your ground with the number. You **concede** when the
objection asks for a crossing your design never paid for, and you narrow the claim
on the spot. The failure to avoid is upgrading your claim under pressure, saying
"causes" out loud because a skeptic pushed, when your written poster only ever
said "is associated with."

> **A question that often comes up here:** *"Doesn't conceding make me look like I
> do not know my own project?"* The opposite. A concession that names the exact
> boundary is the clearest signal that you *do* know your project, because you can
> see where its evidence stops. The bluff is what exposes you. A reviewer who
> catches one overreach stops trusting every other claim on the board.

> **After running, verify (counters *silent scope change*: a suggested answer that
> quietly upgrades your claim to settle a harder question than your design supports):**
> - [ ] Read each suggested answer next to your compass position, word for word.
>   A skeptic's alternative explanation on an observational design is a *concede*,
>   so reject any answer that tells you to "defend" it as settled.
> - [ ] Confirm each question truly belongs to its named lens. A methods question
>   dressed as an outsider question is mislabeled, and the mislabel will send you
>   into the wrong defense at the board.
> - [ ] Check that the "cannot answer" question is honestly the one your evidence
>   stops short of. If the tool marks a question you *can* answer as unanswerable,
>   trust your own read of your design.
> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# The three lenses, the poster part each probes, and the claim they all test.
D = nx.DiGraph()
lenses = {
    "Interdisciplinary\n(the outsider)": (0.0, 2.0),
    "Methods\n(the specialist)":         (0.0, 1.0),
    "Skeptic\n(the doubter)":            (0.0, 0.0),
}
parts = {
    "Question &\nrelevance":     (1.7, 2.0),
    "Design &\nspecification":   (1.7, 1.0),
    "Alternative\nexplanations": (1.7, 0.0),
}
claim = {"Headline\nclaim": (3.4, 1.0)}
pos = {**lenses, **parts, **claim}

lens_to_part = list(zip(lenses, parts))          # each lens aims at one poster part
D.add_edges_from(lens_to_part)
D.add_edges_from((p, next(iter(claim))) for p in parts)  # every part feeds the claim

fig, ax = plt.subplots(figsize=(10, 4.6))
nx.draw_networkx_nodes(D, pos, nodelist=list(lenses), node_color="#DCE6F1",
                       edgecolors="#4C72B0", linewidths=1.5, node_size=5200, ax=ax)
nx.draw_networkx_nodes(D, pos, nodelist=list(parts), node_color="#EAF3E5",
                       edgecolors="#55A868", linewidths=1.5, node_size=5200, ax=ax)
nx.draw_networkx_nodes(D, pos, nodelist=list(claim), node_color="#F3EAD6",
                       edgecolors="#C89B3C", linewidths=1.8, node_size=6000, ax=ax)
nx.draw_networkx_labels(D, pos, font_size=8.5, ax=ax)
nx.draw_networkx_edges(D, pos, arrowstyle="-|>", arrowsize=16, edge_color="#555555",
                       width=1.5, node_size=5200, ax=ax)
ax.set_title("Three lenses, one claim: each question aims at a different poster part")
ax.axis("off")
plt.tight_layout()
plt.show()

print("Each lens probes a different part of the poster:")
for lens, part in lens_to_part:
    print(f"  {lens.splitlines()[0]:<18} -> {part.replace(chr(10), ' ')}")
assert D.number_of_edges() == 6, "three lens->part edges plus three part->claim edges"
print("✓ Map drawn: the outsider tests relevance, the specialist tests the design, the skeptic tests the boundary.")
```

*The map shows three labeled entry points on the left, the interdisciplinary,
methods, and skeptic lenses, each arrow reaching one middle box, the poster part
it attacks, and all three middle boxes feeding the single headline claim on the
right. Reading it tells you a defense is not one skill but three, aimed at three
different targets.*

## 3. The Final Lock: Every Number to a Cell, Every Citation to a Source

**Guiding question:** *before your poster becomes permanent, can you prove every
number, resolve every citation, and disclose every place an AI helped?*

> *"At the Expo I point at one number and ask, 'where did this come from?' If you
> can walk me to the line of code that made it, I believe the whole board. If you
> cannot, I stop believing all of it."*
> — a **research-integrity judge** working the poster hall

The gallery walk fixed how your poster reads. The oral defense fixed what you say
about it. The lock fixes the last thing a stranger can break: whether the poster's
claims are actually *backed*. Four habits carry a poster through the lock, and each
one is a graded row in your M10 final poster.

- **Claim-evidence traceability.** Every number on your poster traces back to the
  exact notebook cell that produced it. Example: the correlation printed on your
  board points to the one line of code that computed it, and you can show that line
  on request. A number you cannot walk back to a cell, one typed from memory,
  rounded by hand, or carried from an old draft, is an **untraceable number**, and
  it is an M10 hard cap: it cannot go on the board.
- **Citation verification.** Every source your poster names resolves to a real,
  retrievable work you opened yourself. Example: you click the link, the paper
  loads, and it actually says what your poster claims it says.
- **AI-use disclosure block.** A short, honest note that records which AI tool did
  which task, and how you checked its output. Example: "Gemini drafted the skeptic
  questions; I mapped each to an audit and verified it against my own figure."
- **Terminal lock.** The moment your poster becomes final and unchangeable. Example:
  once you submit the file for printing, the Expo audience reads exactly that
  version, with no chance to fix a number afterward.

A locked poster is a promise you cannot take back, so the audit that precedes
the lock is the most consequential one you run all week.

> **A question that often comes up here:** *"Tracing every single number to a cell
> sounds like busywork. Why does it matter for a poster no one will audit line by
> line?"* Because one stranger might. A judge who points at a figure and hears
> "I think that came from the analysis somewhere" has just learned to distrust
> every other number on the board. The reverse is also true. Walk one pointed
> finger straight to its cell and you buy trust for the whole poster. Traceability
> is not paperwork. It is the thing that lets a claim survive a hard question in
> public.
