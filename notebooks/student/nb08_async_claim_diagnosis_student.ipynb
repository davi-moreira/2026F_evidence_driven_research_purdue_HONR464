The skill today is **diagnosis without cruelty**. Anyone can shout "fake news"
at a claim they dislike. A researcher does something harder and more useful:
take the claim apart, keep the piece that survives, name the piece that does
not, and say what evidence would settle the rest. That posture has a name.
**Charitable reconstruction** means stating the strongest honest version of a
claim before you critique it. It is the difference between debunking and
understanding.

> **A question that often comes up here:** *"If a study is flawed, why not just
> throw the whole thing out?"* Because "flawed" is not a verdict. It is a
> location: it tells you WHERE the claim outran its evidence, not that nothing
> was learned. Almost every flawed study still supports some weaker, honest
> claim. Finding that surviving claim is the whole point. Discarding the study
> wholesale throws away the true part along with the oversold part.

## 3. The Anatomy of a Claim, and Where It Sits on the Compass

**Guiding question:** *What is this claim made of, and which compass position
is it reaching to occupy?*

Before touching the arithmetic, take the claim apart structurally. Every claim
has three parts, and naming them shows you where to aim:

- **The claim** is what is asserted: skipping breakfast drives course failure.
- **The evidence** is what is offered in support: a one-term survey of 1,000
  first-years, and a table of failure counts.
- **The warrant** is the unstated bridge that says the evidence supports the
  claim: "if skippers fail more in this survey, then skipping *causes* failure."

Most overclaims break at the **warrant**. The evidence is real, but the bridge
is asked to carry far too much weight. So the decisive question becomes a
compass question. What kind of answer is the claim, descriptive or causal? How
far does it reach: the data at hand, a population beyond the data, or cases not
yet seen? And which design would that position require?

- A **description** claim ("in this survey, skippers failed at a higher rate")
  needs only honest counting of the data at hand.
- A **generalization** claim ("first-years in general skip and struggle") needs
  a defensible sample and stated uncertainty. That is the sample-to-population
  crossing, and a sampling design is its license.
- A **prediction** claim ("this flags who will struggle next term") needs a
  check on cases the model never saw. That is the crossing to unseen cases, and
  a **held-out test** (an honest test on data kept aside) is its license.
- A **causal** claim ("skipping breakfast *causes* failure") needs an
  **identification argument**: a stated reason the comparison isolates the
  cause, ideally **random assignment** (a coin flip decides who gets what, so
  the groups start alike). That is the descriptive-to-causal crossing.

Here is the sharp version of the whole diagnosis: **every overclaim is a
compass crossing made without paying for it.** The course names the three
classic smuggles:

- **The silent upgrade.** The words drift from "in my sample" to "people in
  general," and no sampling design paid for the trip.
- **Leakage.** A claim about unseen cases rides on a check that quietly used
  information you would not have at prediction time, or on no held-out check
  at all.
- **After-therefore-because.** A pattern is narrated as a cause, and no
  assignment or identification argument paid for that trip either.

Name the crossing and you have named the flaw. The breakfast story's language
("driver," "protect your GPA," "eat breakfast") is causal. Its evidence is a
one-term **observational** survey: nobody was assigned to skip breakfast, the
survey just watched. That is descriptive evidence about the data at hand. Hold
that gap in mind while the practice below warms up your classifier.

## 4. Recompute the Central Number

**Guiding question:** *When you rebuild the headline numbers from scratch,
which survive and which fall apart?*

A structural diagnosis tells you the claim reached too high. Now check whether
the numbers hold up on their own terms. The story made two arithmetic claims,
"twice as likely" and "two-thirds of failures," and each either does or does
not come out of that table. This is the **verify** habit made literal: never
repeat a number you have not reproduced yourself.

### 🛠️ Hands-On: recompute the headline from its own table

Run the cell below. It rebuilds the story's table and recomputes each claimed
figure from scratch. You will see the two group failure rates, the **relative
risk** (one group's rate divided by the other's, the "twice as likely" number),
the **absolute difference** (the plain gap between the two rates, in percentage
points), and the decisive one: the skippers' true share of all course failures.
Read each printed line against what the story claimed.

```python
# The story's own table, rebuilt exactly as printed.
case = pd.DataFrame({
    "group": ["Breakfast-skippers", "Breakfast-eaters"],
    "n": [200, 800],
    "failed": [24, 48],
})
case["passed"] = case["n"] - case["failed"]
case["failure_rate"] = case["failed"] / case["n"]
print("The survey's own numbers:")
print(case.to_string(index=False))
print()

rate_skip = 24 / 200        # skippers' failure rate
rate_eat  = 48 / 800        # eaters' failure rate
print(f"Skipper failure rate : {rate_skip:.0%}   (24 of 200)")
print(f"Eater   failure rate : {rate_eat:.0%}    (48 of 800)")
print()

# Claim (a): "twice as likely to fail"
relative_risk = rate_skip / rate_eat
print(f"CLAIM (a) 'twice as likely'  → relative risk = {relative_risk:.1f}×   → TRUE, 12% vs 6%")

# ...but what the relative framing hides:
absolute_gap = rate_skip - rate_eat
print(f"   BUT the ABSOLUTE gap is only {absolute_gap:.0%} "
      f"(6 percentage points); {1 - rate_skip:.0%} of skippers did NOT fail.")
print()

# Claim (b): "two-thirds of all course failures came from skippers"
total_failures = case["failed"].sum()
true_share = 24 / total_failures
print(f"CLAIM (b) 'two-thirds of failures' → skippers' TRUE share = "
      f"24 / {total_failures} = {true_share:.0%}   → FALSE, that is a MINORITY")

# How the story likely got 'two-thirds': it compared the RATES, ignoring group size.
wrong_share = rate_skip / (rate_skip + rate_eat)
print(f"   The story's error: rate / (rate+rate) = {wrong_share:.0%} — a ratio of")
print(f"   RATES, not a share of PEOPLE. Eaters outnumber skippers 4-to-1, so their")
print(f"   6% is far more failures (48) than the skippers' 12% (24).")
```

```python
# Self-check: pin down every recomputed figure so the diagnosis rests on facts.
assert relative_risk == 2.0, "the doubling claim should check out exactly"
assert abs(absolute_gap - 0.06) < 1e-9, "absolute gap should be 6 percentage points"
assert round(true_share, 3) == 0.333, "skippers are one-third of failures, not two-thirds"
assert total_failures == 72, "the table's failures must total 72"
print("✓ Self-check passed: the '2× likelier' claim is TRUE but small in absolute")
print("  terms; the 'two-thirds of failures' claim is FALSE — skippers are 33%, a")
print("  minority. One claim survives, one does not, and now you can prove both.")
```

## 5. The Structured Diagnosis: What Stands, What Falls, What Would Settle It

A diagnosis is not a verdict of "good" or "bad." It is a three-part memo that a
fair reader could act on:

1. **What stands:** the claim the evidence genuinely supports, stated at its
   true (usually more modest) altitude.
2. **What falls:** the specific overreach, named precisely, with the reason it
   does not follow.
3. **What would settle it:** the design that could actually answer the question
   the headline was reaching for.

That third part is what separates a researcher from a heckler. Debunking stops
at "this is wrong." Diagnosis continues: *here is what would make it right, or
prove it wrong for good.* For the breakfast claim, settling it means breaking
the **confounding**. A confounder is a third factor, like sleep or stress, that
plausibly produces both the skipped breakfasts and the failed courses. A design
that randomly assigns some first-years to a free-breakfast program breaks that
link; a survey that measures and adjusts for the leading confounders at least
narrows it.

> **A question that often comes up here:** *"Isn't 'charitable reconstruction'
> just letting a bad claim off the hook?"* The opposite. Steel-manning a claim,
> stating its strongest honest version before you critique it, makes your
> critique *harder* to dismiss, because no one can say you attacked a
> caricature. It also protects you. The version of you that always finds the
> surviving true claim is the version reviewers trust when you say something
> genuinely does not hold.
