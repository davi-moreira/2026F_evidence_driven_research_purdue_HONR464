The skill today is **diagnosis without cruelty**. Anyone can shout "fake news" at
a claim they dislike. What a researcher does is harder and more useful: take the
claim apart, keep the piece that survives, name the piece that does not, and say
precisely what evidence would settle the rest. That posture — **charitable
reconstruction** — is the difference between debunking and understanding.

> **A question that often comes up here:** *"If a study is flawed, why not just
> throw the whole thing out?"* Because "flawed" is not a verdict, it is a
> location — it tells you WHERE the claim outran its evidence, not that nothing
> was learned. Almost every flawed study still supports some weaker, honest claim.
> Finding that surviving claim is the whole point; discarding the study wholesale
> throws away the part that was true along with the part that was oversold.

## 3. The Anatomy of a Claim — and Which Approach It Is Using

Before touching the arithmetic, take the claim apart structurally. Every claim has
three parts, and naming them shows you where to aim:

- **The claim** — what is asserted (skipping breakfast drives course failure).
- **The evidence** — what is offered in support (a one-term survey of 1,000
  first-years; a table of failure counts).
- **The warrant** — the unstated bridge that says the evidence supports the claim
  ("if skippers fail more in this survey, then skipping *causes* failure").

Most overclaims break at the **warrant** — the evidence is real, but the bridge it
is asked to carry is far too heavy. So the decisive question becomes: *what kind of
claim is this, and what design would that kind of claim require?* That is the
four-approach classification you have practiced all semester:

- A **descriptive** claim ("in this survey, skippers failed at a higher rate")
  needs only honest counting of the data at hand.
- An **inferential** claim ("first-years in general…") needs a defensible sample
  and stated uncertainty.
- A **causal** claim ("skipping breakfast *causes* failure") needs an
  identification argument — ideally random assignment, at minimum serious control
  of confounders.

The Beacon's language — "driver," "protect your GPA," "eat breakfast" — is
**causal**. Its evidence is a single-term observational survey. Hold that mismatch
in mind; it is the heart of the diagnosis.

## 4. Recompute the Central Number

A structural diagnosis tells you the claim reached too high. Now check whether the
numbers even hold up on their own terms — because the Beacon made two *arithmetic*
claims ("twice as likely," "two-thirds of failures") that either do or do not come
out of that table. This is the **verify** habit made literal: never repeat a number
you have not reproduced yourself.

The next cell rebuilds the Beacon's own table and recomputes each claimed figure
from scratch.

### 🛠️ Hands-On: recompute the headline from its own table

Run the cell below. It reconstructs the study's table, computes the two group
failure rates, the relative risk ("twice as likely"), the *absolute* difference,
and — the decisive one — the skippers' true share of all course failures. Read each
printed line against what the story claimed.

```python
# The Beacon's own table, rebuilt exactly as printed.
case = pd.DataFrame({
    "group": ["Breakfast-skippers", "Breakfast-eaters"],
    "n": [200, 800],
    "failed": [24, 48],
})
case["passed"] = case["n"] - case["failed"]
case["failure_rate"] = case["failed"] / case["n"]
print("The study's own numbers:")
print(case.to_string(index=False))
print()

rate_skip = 24 / 200        # skippers' failure rate
rate_eat  = 48 / 800        # eaters' failure rate
print(f"Skipper failure rate : {rate_skip:.0%}   (24 of 200)")
print(f"Eater   failure rate : {rate_eat:.0%}    (48 of 800)")
print()

# Claim (a): "twice as likely to fail"
relative_risk = rate_skip / rate_eat
print(f"CLAIM (a) 'twice as likely'  → relative risk = {relative_risk:.1f}×   → TRUE, 12% vs 6%")

# ...but what the relative framing hides:
absolute_gap = rate_skip - rate_eat
print(f"   BUT the ABSOLUTE gap is only {absolute_gap:.0%} "
      f"(6 percentage points); {1 - rate_skip:.0%} of skippers did NOT fail.")
print()

# Claim (b): "two-thirds of all course failures came from skippers"
total_failures = case["failed"].sum()
true_share = 24 / total_failures
print(f"CLAIM (b) 'two-thirds of failures' → skippers' TRUE share = "
      f"24 / {total_failures} = {true_share:.0%}   → FALSE, that is a MINORITY")

# How the story likely got 'two-thirds': it compared the RATES, ignoring group size.
wrong_share = rate_skip / (rate_skip + rate_eat)
print(f"   The story's error: rate / (rate+rate) = {wrong_share:.0%} — a ratio of")
print(f"   RATES, not a share of PEOPLE. Eaters outnumber skippers 4-to-1, so their")
print(f"   6% is far more failures (48) than the skippers' 12% (24).")
```

```python
# Self-check: pin down every recomputed figure so the diagnosis rests on facts.
assert relative_risk == 2.0, "the doubling claim should check out exactly"
assert abs(absolute_gap - 0.06) < 1e-9, "absolute gap should be 6 percentage points"
assert round(true_share, 3) == 0.333, "skippers are one-third of failures, not two-thirds"
assert total_failures == 72, "the table's failures must total 72"
print("✓ Self-check passed: the '2× likelier' claim is TRUE but small in absolute")
print("  terms; the 'two-thirds of failures' claim is FALSE — skippers are 33%, a")
print("  minority. One claim survives, one does not, and now you can prove both.")
```

```python
# One picture makes the framing trick visible: the same data, two stories.
fig, axes = plt.subplots(1, 2, figsize=(11, 4.2))

# Left: the scary "relative" framing — bars of the two failure rates.
axes[0].bar(["Skippers", "Eaters"], [rate_skip, rate_eat],
            color=["#C44E52", "#4C72B0"])
axes[0].set_ylabel("Failure rate")
axes[0].set_ylim(0, 1)
axes[0].set_title("The headline's framing:\n'skippers fail at DOUBLE the rate'")
for i, r in enumerate([rate_skip, rate_eat]):
    axes[0].text(i, r + 0.02, f"{r:.0%}", ha="center", fontweight="bold")

# Right: the honest "absolute" framing — what happened to each 100 skippers.
axes[1].bar(["Failed", "Did NOT fail"], [rate_skip, 1 - rate_skip],
            color=["#C44E52", "#B0B0B0"])
axes[1].set_ylabel("Share of skippers")
axes[1].set_ylim(0, 1)
axes[1].set_title("The same skippers, whole picture:\n88% did not fail a course")
for i, r in enumerate([rate_skip, 1 - rate_skip]):
    axes[1].text(i, r + 0.02, f"{r:.0%}", ha="center", fontweight="bold")

plt.tight_layout()
plt.show()
```

## 5. The Structured Diagnosis — What Stands, What Falls, What Would Settle It

A diagnosis is not a verdict of "good" or "bad." It is a three-part memo that a fair
reader could act on:

1. **What stands** — the claim the evidence genuinely supports, stated at its true
   (usually more modest) altitude.
2. **What falls** — the specific overreach, named precisely, with the reason it does
   not follow.
3. **What evidence would settle it** — the design that could actually answer the
   question the headline was reaching for.

That third part is what separates a researcher from a heckler. Debunking stops at
"this is wrong." Diagnosis continues: *here is what would make it right, or prove it
wrong for good.* For the breakfast claim, settling it would take a design that
breaks the confounding — randomly assigning some first-years to a free-breakfast
program and comparing course outcomes, or at minimum an observational study that
measures and adjusts for sleep, work hours, and prior performance.

> **A question that often comes up here:** *"Isn't 'charitable reconstruction' just
> letting a bad claim off the hook?"* The opposite. Steel-manning a claim — stating
> the strongest honest version before you critique it — makes your critique
> *harder* to dismiss, because no one can say you attacked a caricature. It also
> protects you: the version of you that always finds the surviving true claim is the
> version reviewers trust when you say something genuinely does not hold.

## 6. Preview of Monday: The Estimand / Estimator / Estimate Triad

Everything you just did by hand — deciding what the study *should* have measured,
what procedure would measure it, and what number came out — has three formal names
that Monday's class (nb09) will make precise. Learn the words now, using a cooking
metaphor that will carry all week:

- **Estimand** — the true quantity you are after, the number the world actually
  holds. *The dish the genie would serve you if you could ask.* For the breakfast
  question, it is the real causal effect of breakfast on course failure — which no
  survey ever handed us.
- **Estimator** — the *recipe*: the exact procedure you apply to data to produce an
  answer (a difference in failure rates, a regression, a difference-in-means).
- **Estimate** — the *dish you actually cooked*: the specific number the recipe
  returns on your data (12% minus 6% = 6 percentage points).

The Beacon's deepest problem, in these words: it reported an **estimate** (a rate
difference) as if it were the **estimand** (the causal effect) — serving the dish it
cooked as though it were the dish the genie promised. Monday you will build the
recipe yourself, on a world where you actually get to peek at the genie's number and
measure exactly how far your recipe lands from it.

> **A question that often comes up here:** *"If we can never see the estimand, how do
> we know any recipe is any good?"* Because you can build a pretend world where you
> DID plant the true number yourself, run your recipe on it, and check the miss —
> which is precisely what nb09 does. Testing your recipe on a world whose truth you
> control is how you earn trust in it before pointing it at a world whose truth you
> do not.
