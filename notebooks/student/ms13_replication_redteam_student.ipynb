## 🎯 Definition of Done

You are done with this sprint when your **Replication and Red-Team Report** on your
assigned peer package contains all five parts, each pointing at a number your own
run produced:

1. **Reproduction log.** Did **restart-and-run-all** (clear memory, run every cell
   top to bottom, no manual fixes) succeed? The **headline number** as *your* run
   produced it, whether it matched the write-up, and every break logged to the
   exact cell. If part did not reproduce, that is a logged finding, not a failure.
2. **Claim-to-output trace.** Each written claim beside the output that backs it,
   with every gap flagged, including a rounded or restated number and any missing
   uncertainty (a settled number reported with no interval).
3. **One alternative specification and one hidden assumption.** The headline
   computed a second defensible way, and one assumption the analysis leans on
   quietly, each labeled as it **moves** the estimate, **widens** its uncertainty,
   or **breaks** the claim.
4. **The verified GenAI Studio pass.** What the Reproducibility Auditor flagged, and
   which flags survived *your own run*, not the tool's trace.
5. **Prioritized recommendations.** Your findings ranked by **threat to the claim,
   not ease of fixing**, plus the single most-threatening weakness (and the one
   check that would settle it) that you post to the async board.

Plus the **AI Research Ledger** rows and the **dossier update line**. Then post
your most-threatening weakness to the board and reply to one classmate.

> **A question that often comes up here:** *"The package reproduced with a green
> check. Isn't the sprint basically over?"* No — that is where it starts. A clean
> run proves the code executes; it says nothing about whether the write-up's
> sentence matches the output. Reproduction earns you a trustworthy number; parts
> 2 through 5 are the red-team, and that is where the grade lives.

### 🛠️ Hands-On: organize your reproduced numbers

Run the scaffold below on its default numbers first to see what it does. Then edit
the **one dict** and the **one list** at the top with the figures from *your*
assigned package, after your own cold run, and rerun. Everything downstream derives
from those two edits, so there is a single source of truth for every number: change
a value once at the top and the whole readout stays consistent.

```python
# --- Reproduction scaffold — edit ONLY the dict and the list below.
# Everything after them is DERIVED, so there is one source of truth per number.
# Plug in the figures from YOUR assigned package AFTER your own restart-and-run-all.

MY_PACKAGE = {
    "claim_label":      "canvassing raised turnout",         # EDIT ME: the headline claim, in words
    "reported_value":   4.00,   # EDIT ME: the number the WRITE-UP states
    "reproduced_value": 3.41,   # EDIT ME: the number YOUR run produced
    "alt_spec_label":   "weighted instead of unweighted",    # EDIT ME: your alternative specification
    "alt_spec_value":   2.83,   # EDIT ME: the headline under that alternative choice
    "units":            "percentage points",
}

# Per-cluster estimates from YOUR run: one number per independent group (a street,
# a site, a class — whatever your package's clustering unit is). EDIT ME with your
# package's per-cluster differences. The defaults let the cell run before you do.
cluster_estimates = [5.1, 2.2, -0.4, 4.8, 1.3, 6.0, 0.2, 3.7, -1.1, 2.9,
                     4.4, 1.8, 0.6, 3.1, 5.5]

# ---- derived below; do not retype any number, read it from the two edits above ----
u = MY_PACKAGE["units"]

# 1) Claim-to-output: does the write-up's number match your run?
claim_gap = MY_PACKAGE["reported_value"] - MY_PACKAGE["reproduced_value"]
matched = abs(claim_gap) < 0.005
print(f"CLAIM-TO-OUTPUT TRACE for '{MY_PACKAGE['claim_label']}'")
print(f"  write-up states  : {MY_PACKAGE['reported_value']:.2f} {u}")
print(f"  your run produced: {MY_PACKAGE['reproduced_value']:.2f} {u}")
print(f"  gap = {claim_gap:+.2f} {u}  -> " +
      ("MATCH" if matched else
       "MISMATCH: the write-up reports a number your run does not produce"))

# 2) Alternative specification: how far does one undisclosed choice move the number?
spec_swing = MY_PACKAGE["reproduced_value"] - MY_PACKAGE["alt_spec_value"]
print(f"\nALTERNATIVE SPECIFICATION ({MY_PACKAGE['alt_spec_label']})")
print(f"  as reported: {MY_PACKAGE['reproduced_value']:.2f} {u}  |  "
      f"alternative: {MY_PACKAGE['alt_spec_value']:.2f} {u}  |  "
      f"swing = {spec_swing:+.2f} {u}")

# 3) Hidden assumption: honor the clustering. Resample WHOLE clusters (seed 464),
#    not individual rows, and read whether the honest interval reaches zero.
arr = np.array(cluster_estimates, dtype=float)
K = len(arr)
boot = np.array([rng.choice(arr, size=K, replace=True).mean() for _ in range(2000)])
lo, hi = np.percentile(boot, [2.5, 97.5])
reaches_zero = lo <= 0 <= hi
print(f"\nHIDDEN ASSUMPTION (clustering): cluster-aware 95% interval over {K} clusters")
print(f"  [{lo:+.2f}, {hi:+.2f}] {u}  -> " +
      ("reaches zero: the clear-effect reading weakens once clusters count"
       if reaches_zero else "stays above zero"))

# 4) Threat-ranking template — YOU write the order; the numbers above are the evidence.
print("\nRANK BY THREAT (most dangerous first) — replace with YOUR ordering:")
print("  1. <weakness that could BREAK the claim>   (evidence: a number above)")
print("  2. <weakness that MOVES the estimate>       (evidence: a number above)")
print("  3. <weakness that only WIDENS uncertainty>  (evidence: a number above)")
print("\n✓ Scaffold ran on the default numbers. Now edit MY_PACKAGE and")
print("  cluster_estimates with YOUR package's figures and rerun.")
```

code. Only your own run breaks the tie.
