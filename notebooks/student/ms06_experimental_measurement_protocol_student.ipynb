## 🎯 Definition of Done

You are done with this sprint when your submission contains all six parts of the
protocol, applied to **your own project**:

1. **The construct and the measurement design.** The **latent characteristic** you
   must measure (a level, a prevalence, a weight, a threshold), and the
   design that reveals it: an experiment (audit, list, conjoint, behavioral game,
   calibration, A/B instrument test) or a direct data-acquisition route, with one
   sentence of why it fits.
2. **The descriptive quantity it reports.** The exact number your design is built
   to report (the **estimand**) and its units, kept descriptive for the units you
   observe. If a random assignment sits inside it, one line on why it serves
   measurement and not a causal inquiry.
3. **The measurement DAG and artifact plan.** An image showing the stimulus (or
   source) and the latent characteristic feeding the measured response, plus at
   least one dashed **artifact** arrow. In words: the **demand effect** and the
   **instrument effect** you most fear, each with the redesign that defuses it.
4. **The claim boundary (two sentences).** The descriptive claim your design
   licenses, with its uncertainty, and the causal or population sentence you
   forbid yourself to write.
5. **The URC abstract.** Roughly 150 to 250 words, describing your project and its
   measurement, worded so it stays inside the Component 4 boundary. This clears an
   internal gate at the studio before it goes out.
6. **AI Research Ledger rows.** One row for every AI-assisted step, each with a
   named verification method. A missing ledger returns the submission.

Plus the **dossier update line** naming where each part now lives in your Research
Project Dossier. You walk a partner and the Experimental Design Reviewer through
the protocol at the studio, then polish from the red-team.

```python
# Measurement-artifact scaffold — edit the four numbers to match YOUR
# project. It shows, on one picture, why a raw measurement can be mostly artifact:
# your instrument reads the TRUE construct value PLUS a demand effect PLUS an
# instrument effect, and a redesign is meant to shrink both. Replace every EDIT ME
# number with your own project's rough guess, then read whether your redesign
# actually recovers the construct.

# EDIT ME — four numbers, on your own measurement scale, are the single source of
# truth for this picture. Change these and every bar rebuilds itself; nothing else
# is hardcoded.
true_value        = 5.0   # the latent characteristic you are trying to measure
demand_effect     = 1.2   # how far respondents shift to "look right" to you
instrument_effect = 0.8   # how far the tool or wording itself pushes the reading
redesign_removes  = 0.85  # FRACTION (0 to 1) of BOTH artifacts your redesign kills

# A seeded noise draw so the honest noise band is identical every run (SEED is set
# in the setup cell, so this is reproducible and edit-safe).
n_resp = 200
noise_band = float(rng.normal(0.0, 0.4, size=n_resp).std())

# The readings are all BUILT from the four numbers above, so they stay in sync.
artifact_total     = demand_effect + instrument_effect
naive_reading      = true_value + artifact_total
residual_artifact  = (1 - redesign_removes) * artifact_total
redesigned_reading = true_value + residual_artifact

labels = ["True\nconstruct", "Naive\nmeasurement", "Redesigned\nmeasurement"]
values = [true_value, naive_reading, redesigned_reading]
colors = ["#1a7f37", "#b3261e", "#33475b"]

fig, ax = plt.subplots(figsize=(9, 4.5))
bars = ax.bar(labels, values, color=colors, width=0.6)
ax.errorbar(range(3), values, yerr=noise_band, fmt="none",
            ecolor="#555555", capsize=6, linewidth=1.2)
ax.axhline(true_value, color="#1a7f37", linestyle="dashed", linewidth=1.0)
for b, v in zip(bars, values):
    ax.text(b.get_x() + b.get_width() / 2, v + 0.08, f"{v:.2f}",
            ha="center", fontsize=10)
ax.set_ylabel("Reading on your measurement scale")
ax.set_title("How much of your number is construct, and how much is artifact?",
             fontsize=11)
plt.tight_layout()
plt.show()

print(f"✓ Scaffold rendered — naive reading is inflated by {artifact_total:.2f} "
      f"of pure artifact; your redesign leaves {residual_artifact:.2f} behind")
print("  Now edit the four EDIT ME numbers for your own project.")
```

### 🗡️ Red-Team Exchange

The studio's **29 to 41 minute block**. Pair with one partner and run two rounds of
attack. Bring in an AI reviewer as a third voice using the **Experimental Design
Reviewer** role card (`ai_resources/agent_role_cards.md`; full briefing in
`genai_studio/roles/experimental_design_reviewer.md`). This role is the recommended
reviewer for the M6 measurement or data-acquisition protocol.

**Round 1: attack the boundary and the abstract (about 6 minutes).**
Your partner reads your licensed sentence and your URC abstract and hunts for a
place where either quietly claims a **causal** effect or a **population** your
design never reached. They also paste the boundary to the Reviewer, which returns
an **assignment read**: whether your assignment supports a descriptive claim or is
smuggling in a causal one. You defend or concede, then rewrite so the abstract sits
inside the boundary.

**Round 2: attack the artifact plan (about 6 minutes).**
Your partner questions your DAG's dashed arrows. Is the **demand effect** you named
the real worst one, or drawn to look complete? Is there an **instrument effect** you
are hiding? Does each redesign actually rule out its artifact? The Reviewer's
**structure check** and its **ethics questions** feed this round; you answer the
ethics questions yourself, never on a model's say-so.

**What to log.** Write down the single strongest hit from each round and how you
changed the protocol in response. An attack that changed nothing still gets a line
saying why it did not land. That record is part of what you submit.

> convincing and why, in one sentence. And remember the correlated-error warning: if
> you also ask Gemini and it agrees your design is sound, that is not confirmation.
