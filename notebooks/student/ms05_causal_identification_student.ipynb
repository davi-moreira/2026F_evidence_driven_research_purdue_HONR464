## 🎯 Definition of Done

You are done with this sprint when your submission contains all six parts of the
deliverable, applied to **your own project**:

1. **The causal question and estimand.** Your treatment, your outcome, and the
   estimand (the average causal effect, the average of Y(1) minus Y(0)), one line
   each, plus one sentence naming the **counterfactual** you never observe.
2. **The confounding causal diagram.** An image showing your treatment-to-outcome
   arrow *and* at least one **confounder** with arrows into both treatment and
   outcome: the **back door** your design must close or your boundary must
   confess.
3. **The identification argument OR the honest causal-language boundary.** If you
   have leverage (selection on observables, a natural experiment, DiD, RDD, or
   IV), three sentences naming the design, its **one untestable assumption**, and
   how a skeptic attacks it. If you have none, the one-sentence boundary that
   keeps your claim on the *associated with* side of the line.
4. **The Causal Identification Skeptic critique with your resolutions.** The
   ranked threats it returned, each answered in writing.
5. **One recompute.** The single number you would recompute yourself to check any
   AI or collaborator, and how.
6. **AI Research Ledger rows.** One row for every AI-assisted step, each with a
   named verification method. A missing ledger returns the submission.

Plus the **dossier update line** naming where each part now lives in your
Research Project Dossier. You record a 90-second causal-strategy statement, post
it to the board, and revise from the red-team.

> **A question that often comes up here:** *"What if my honest answer is that my

```python
# Confounding-diagram scaffold — edit the three labels to match YOUR project.
# It renders the causal picture at the center of M5: the treatment -> outcome
# arrow you want to estimate, plus ONE confounder with arrows into BOTH treatment
# and outcome. That double arrow is the back door your design must close or your
# boundary must confess. Replace every "EDIT ME" with your own project's words.
import networkx as nx

# EDIT ME — the three names below are the SINGLE SOURCE OF TRUTH for the diagram.
# Rename them to your project and every arrow, label, and position rebuilds itself.
treatment  = "Treatment\n(EDIT ME)"
outcome    = "Outcome\n(EDIT ME)"
confounder = "Confounder\n(EDIT ME)"

G = nx.DiGraph()
# The effect you WANT to estimate: treatment -> outcome.
G.add_edge(treatment, outcome)
# The back door: the confounder points into BOTH treatment and outcome. This is
# selection into treatment (why some units got treated) colliding with the thing
# you measure. We reference the three variables above, never a hardcoded string,
# so renaming any label cannot break these edges.
G.add_edge(confounder, treatment)
G.add_edge(confounder, outcome)

# Positions are built FROM your three variables, so they are the single source of
# truth: rename a label above and the layout still works (deterministic layout,
# no randomness needed).
pos = {treatment: (0, 0), outcome: (2, 0), confounder: (1, 1.15)}

# Colour the back-door edges (from the confounder) differently from the effect
# edge, so the trap is visible at a glance.
effect_edges   = [(treatment, outcome)]
backdoor_edges = [(confounder, treatment), (confounder, outcome)]

fig, ax = plt.subplots(figsize=(9, 4.5))
nx.draw_networkx_nodes(G, pos, node_size=5200, node_color="#e8eef6",
                       edgecolors="#33475b", linewidths=1.5, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=9, ax=ax)
nx.draw_networkx_edges(G, pos, edgelist=effect_edges, arrowstyle="-|>",
                       arrowsize=20, width=2.0, edge_color="#1a7f37",
                       node_size=5200, ax=ax)
nx.draw_networkx_edges(G, pos, edgelist=backdoor_edges, arrowstyle="-|>",
                       arrowsize=20, width=1.6, edge_color="#b3261e",
                       style="dashed", node_size=5200,
                       connectionstyle="arc3,rad=0.08", ax=ax)
ax.set_title("Confounding diagram — green: the effect you want · "
             "red dashed: the back door", fontsize=11)
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ Diagram rendered — now edit the three EDIT ME labels for your own project")
```

### 🗡️ Red-Team Exchange

M5 is the **async module**, so the red-team runs on the **Brightspace board**
rather than in a live studio block. Pair with one partner in the thread and run
two rounds of attack, with the **Causal Identification Skeptic** as a third voice
(role card in `ai_resources/agent_role_cards.md`; full briefing in
`genai_studio/roles/causal_identification_skeptic.md`). This role is a **required**
reviewer for M5.

**Round 1: attack the untestable assumption.**
Your partner, and the Skeptic, read your identification argument and press on its
one **untestable assumption** (no hidden confounder / parallel trends / as-if
random at the cutoff / exclusion). Is it stated at all? Could it fail in your
setting? The Skeptic returns a ranked **Threats to identification** list. You
answer each threat in writing: show why your design rules it out, or carry it as
a limitation. An unanswered top threat is an open hole.

**Round 2: attack the language for design mimicry.**
Your partner and the Skeptic hunt for **design mimicry**: a causal label or the
word *because* doing work your stated assumption has not earned. They catch any
**Language-boundary flag** where your wording claims more than your license holds.
You defend the wording or rewrite it to the boundary your design actually reaches.

**What to log.** Write down the single strongest hit from each round and how you
changed the argument in response. An attack that changed nothing still gets a line
saying why it did not land. Those threats and your resolutions are the required M5
content of your ledger.

> **A question that often comes up here:** *"What if my partner and the Skeptic
> disagree?"* Good. You are the judge, not either of them. Say which you found
> more convincing and why, in one sentence. And remember the correlated-error
> warning: if you also ask Gemini and it agrees "looks identified," that is not
> confirmation — two models can share the same blind spot on causal reasoning.
> Defend against the course license and a real worked example, not a second
> model's agreement.
