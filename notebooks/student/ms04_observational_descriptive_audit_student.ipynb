## 🎯 Definition of Done

You are done with this sprint when your submission contains all five parts of
the audit, applied to **your own project**:

1. **Population / frame / sample stack.** Your target population, accessible
   population, sampling frame, and sample, each one shown to narrow the layer
   above it, plus one sentence naming the **silent exclusion** your frame drops.
2. **A selection DAG.** An image of how a unit lands in your data, showing the
   funnel *and* at least one **selection influence**: a trait that bends both
   who appears in your data and what you measure.
3. **A measurement plan.** The concept → construct → indicator ladder for your
   key construct, with the dominant data risk named as **coverage**,
   **nonresponse**, or **missingness**, and whether it runs in a direction.
4. **The generalization boundary.** Two sentences: the population claim your
   frame licenses (with its uncertainty), and the **silent upgrade** your design
   forbids.

> 💡 **Gemini Prompt:** "Here is Python that draws a selection DAG with
> networkx and matplotlib. Explain what each edge represents, then suggest one

> frame is [your frame]. Do not rewrite the whole cell; describe the single edge
> to add and why."
>
> **After running, verify (the responsible-AI habit):**
> - [ ] Run the edited cell yourself and confirm the new arrow renders where you

> - [ ] Confirm the added influence is real for your project, not just plausible

```python
# Selection-DAG scaffold — edit the labels and arrows to match YOUR project.
# It renders the funnel from target population down to your recorded data, plus
# ONE selection influence: a trait that bends BOTH who lands in your data AND
# what you measure. Replace every "EDIT ME" with your own project's words.
import networkx as nx

# EDIT ME — rename these four layers to your project's real ones.
funnel = [
    "Target\npopulation",
    "Accessible\npopulation",
    "Sampling\nframe",
    "Responded\n& recorded",
]
# EDIT ME — name the trait that drives selection AND touches your measure.
selector = "Selection influence\n(e.g., who opts in)"
measure = "Your key\nmeasure"

G = nx.DiGraph()
# The funnel: each layer narrows into the next.
for a, b in zip(funnel[:-1], funnel[1:]):
    G.add_edge(a, b)
# The dangerous part: the selection influence points into BOTH the sample and
# the measure. That double arrow is what a plain description can be fooled by.
G.add_edge(selector, "Responded\n& recorded")
G.add_edge(selector, measure)
G.add_edge("Responded\n& recorded", measure)

# Deterministic left-to-right layout. No randomness needed; SEED (set in setup)
# keeps the whole notebook reproducible.
pos = {
    "Target\npopulation": (0, 1.0),
    "Accessible\npopulation": (1, 1.0),
    "Sampling\nframe": (2, 1.0),
    "Responded\n& recorded": (3, 1.0),
    "Your key\nmeasure": (4, 1.0),
    selector: (3, 2.15),
}

fig, ax = plt.subplots(figsize=(11, 4))
nx.draw_networkx_nodes(G, pos, node_size=3400, node_color="#e8eef6",
                       edgecolors="#33475b", linewidths=1.5, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=8, ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=18, width=1.4,
                       edge_color="#33475b", node_size=3400,
                       connectionstyle="arc3,rad=0.06", ax=ax)
ax.set_title("Selection DAG scaffold — replace every EDIT ME with your project",
             fontsize=11)
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ DAG scaffold rendered — now edit the EDIT ME nodes for your own project")
```

### 🗡️ Red-Team Exchange

The studio's 29–41 minute block. Pair with one partner and run two rounds of
attack. Bring in an AI reviewer as a third voice using the **Observational
Descriptive Auditor** role card (`ai_resources/agent_role_cards.md`; full
briefing in `genai_studio/roles/observational_descriptive_auditor.md`).

**Round 1: attack the generalization sentence (about 6 minutes).**
Your partner reads your licensed population sentence and tries to catch a
**silent upgrade**: a place where the sentence quietly covers people your frame
never reached, or drifts toward cause. They also paste your sentence to the
Auditor and ask it to name the reach it is not entitled to. You defend or
concede, then rewrite the sentence.

**Round 2: attack the DAG's arrows (about 6 minutes).**
Your partner questions every arrow. Is the selection influence real, or drawn to
look complete? Is there a second influence you are hiding? They challenge one
arrow you cannot defend, and you either justify it or cut it.

**What to log.** Write down the single strongest hit from each round and how you
changed the audit in response. An attack that changed nothing still gets a line
saying why it did not land. That record is part of what you submit.
