## 🎯 Definition of Done

You are done with this sprint when your charter declares all six parts, applied
to **your own project**, and the four MIDA parts **align**, all pointing at the
same quantity:

1. **The Model (M).** Your world in a short paragraph (units, the thing that
   varies, the outcome, the forces that move it) plus a **3-node DAG** image: your
   treatment, your outcome, and at least one **confounder** with an arrow into
   *both*.
2. **The Inquiry (I).** Your estimand stated once in plain words and once as a
   quantity that passes the **genie test**, labeled **descriptive** or **causal**,
   with the units it averages over.
3. **The Data strategy (D).** One paragraph: who is sampled and who gets which
   condition (a coin flip, a lottery, self-selection, or plain observation).
4. **The Answer strategy (A).** One or two sentences turning your data into an
   estimate of the inquiry, targeting *that* quantity and not an easier neighbor.
5. **The diagnosis and one redesign.** Which of **bias**, **variance**, or
   **power** is your biggest threat, what it does to your answer, and the single
   redesign that reads the diagnosis (more units for variance or power; a changed
   comparison or assignment for bias).
6. **Compass position and claim boundary.** Your position (Description,
   Generalization, Prediction, or Causal reasoning) and two sentences: what your
   design *can* establish, and the crossing it does **not** license — including
   whether it **identifies** a cause or only shows an **association**.

Plus **AI Research Ledger rows** for every AI-assisted step, and the **dossier
update line** naming where each part now lives. You deliver a 3-minute charter
declaration at the studio, then revise from the red-team.

> **A question that often comes up here:** *"What if my honest diagnosis says my
> design is weak?"* Then say so, and bring the one redesign that fixes it. A
> charter that names its biggest threat and the fix earns full credit. The defect
> this milestone catches is not a weak design. It is a design you declared
> "rigorous" without ever diagnosing it. Rigor is a diagnosed property, not a
> vibe.

```python
# DAG scaffold — edit the three labels to match YOUR project.
# It renders the model at the center of M3: the treatment -> outcome arrow you
# want to estimate, plus ONE confounder with arrows into BOTH treatment and
# outcome. Export the figure and drop it into your charter. Replace every
# "EDIT ME" with your own project's words.

# EDIT ME — these three names are the SINGLE SOURCE OF TRUTH for the diagram.
# Rename them to your project and every arrow, label, and position rebuilds itself.
treatment  = "Treatment\n(EDIT ME)"
outcome    = "Outcome\n(EDIT ME)"
confounder = "Confounder\n(EDIT ME)"

G = nx.DiGraph()
G.add_edge(treatment, outcome)      # the effect you WANT to estimate
G.add_edge(confounder, treatment)   # the confounder steers who gets treated ...
G.add_edge(confounder, outcome)     # ... and moves the outcome on its own

# Positions are built FROM your three variables, so renaming a label above cannot
# break the layout (deterministic, no randomness needed).
pos = {treatment: (0, 0), outcome: (2, 0), confounder: (1, 1.15)}
effect_edges   = [(treatment, outcome)]
backdoor_edges = [(confounder, treatment), (confounder, outcome)]

fig, ax = plt.subplots(figsize=(9, 4.5))
nx.draw_networkx_nodes(G, pos, node_size=5200, node_color="#e8eef6",
                       edgecolors="#33475b", linewidths=1.5, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=9, ax=ax)
nx.draw_networkx_edges(G, pos, edgelist=effect_edges, arrowstyle="-|>",
                       arrowsize=20, width=2.0, edge_color="#1a7f37",
                       node_size=5200, ax=ax)
nx.draw_networkx_edges(G, pos, edgelist=backdoor_edges, arrowstyle="-|>",
                       arrowsize=20, width=1.6, edge_color="#b3261e",
                       style="dashed", node_size=5200,
                       connectionstyle="arc3,rad=0.08", ax=ax)
ax.set_title("Your model — green: the effect you want · "
             "red dashed: the confounder's two arrows", fontsize=11)
ax.axis("off")
plt.tight_layout()
plt.show()

print("Diagram rendered — now edit the three EDIT ME labels for your own project.")
```

```python
# Diagnosis scaffold — declare a design like YOURS, then read its bias,
# variance, and power. Edit only the constants in this block; the simulation is
# seeded, so your numbers are reproducible. This is a stand-in you tune toward
# your own design, not a substitute for reasoning about it.

# EDIT ME — describe a design shaped like yours.
GUESSED_EFFECT   = 2.0          # the true effect you are hoping to detect (in outcome units)
N_PER_GROUP      = 15           # units in each condition (raise this to cure variance)
OUTCOME_NOISE    = 8.0          # person-to-person wobble in the outcome
CONFOUND_STRENGTH = 6.0         # how hard the confounder pushes on the outcome
ASSIGNMENT       = "randomized"  # "randomized" (coin flip) or "confounded" (units self-select)

def diagnose(effect, n_per_group, noise, confound, assignment, reps=1500, seed=SEED):
    rng_d = np.random.default_rng(seed)   # own reproducible stream
    n = n_per_group * 2
    ests = np.empty(reps)
    detected = np.zeros(reps, dtype=bool)
    for r in range(reps):
        prior = rng_d.normal(0, 1, n)                      # the confounder
        Y0 = 50 + confound * prior + rng_d.normal(0, noise, n)
        Y1 = Y0 + effect
        if assignment == "randomized":
            Z = np.zeros(n, dtype=int)
            Z[rng_d.permutation(n)[:n // 2]] = 1           # a coin flip: half treated
        else:
            Z = (rng_d.random(n) < 1 / (1 + np.exp(-1.5 * prior))).astype(int)
        Y = np.where(Z == 1, Y1, Y0)                       # observe only the assigned world
        est = Y[Z == 1].mean() - Y[Z == 0].mean()          # answer strategy: difference in means
        se = np.sqrt(Y[Z == 1].var(ddof=1) / max((Z == 1).sum(), 1)
                     + Y[Z == 0].var(ddof=1) / max((Z == 0).sum(), 1))
        ests[r] = est
        detected[r] = se > 0 and abs(est) / se > 1.96      # would this run 'find' an effect?
    return ests.mean() - effect, ests.std(ddof=1), detected.mean()

bias, spread, power = diagnose(GUESSED_EFFECT, N_PER_GROUP, OUTCOME_NOISE,
                               CONFOUND_STRENGTH, ASSIGNMENT)
print(f"Design diagnosed over 1500 runs ({ASSIGNMENT}, {N_PER_GROUP} per group):")
print(f"  BIAS      {bias:+.3f}   near 0 means on target; a tilt no extra data can fix")
print(f"  VARIANCE  {spread:.3f}   how much one run wobbles; more units shrinks it")
print(f"  POWER     {power:.1%}   how often the design detects the effect")
print("Your biggest threat is the number that looks worst. Name it in your charter.")
```
