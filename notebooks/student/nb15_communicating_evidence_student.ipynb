## 2. Two Architectures — the Hook and the Walk

**Guiding question:** *what are the two spoken pieces a poster session demands — and how do you compress a claim for speech without inflating it?*

A poster session asks for two different spoken pieces, and they do different jobs
(milestone **M13, Presentation Plan, due Wednesday, November 11, 11:59 PM**;
delivery rounds run in class today):

**The 1-minute hook** — the 60 seconds that earn a stop. Four beats:

1. **Hook** — one sentence that earns attention (a stakes line, a surprising
   number, a shared frustration).
2. **Claim** — your headline, said plainly.
3. **Boundary sentence** — the words that keep the claim honest ("this is an
   association, not proof that…"). **Mark this sentence** and rehearse it until
   it's automatic; it is the one pressure will try to drop.
4. **Invitation** — a reason to look closer ("come see the pattern and tell me
   what you'd expect").

**The 3-minute walk** — the guided tour for the visitor who stopped. It moves
their eye and yours along the poster's read path: headline → the evidence blocks,
in your ranked order → the **ULN moment** (uncertainty, limitation, next step —
you reach it out loud, you don't skip it) → and out. The walk maps onto your
poster's read path and spends its time on the evidence that earns it.

The rule that governs both: **compress, never inflate.** Speaking is faster than
reading, so you drop words — but every word that survives must sit inside the same
boundary your poster's headline sits inside.

> **A question that often comes up here:** *"How is compressing different from
> inflating? Both change the claim."* Compression drops *detail* while keeping the
> claim's *boundary* — "in my sample of 812 commuters, longer commutes go with
> lower attendance" compresses to "in my survey, longer commutes track with
> skipping more class." Inflation moves the *boundary* — "longer commutes *make*
> people skip class." Same length, different sin: one lost a number, the other
> lost the truth.

### 🛠️ The pacing calculator

The cell converts words into seconds: words ÷ words-per-second. It uses a clear,
unrushed presentation pace — deliberately on the slow side, because under pressure
people speed up, and a script that only fits when you rush is a script that
doesn't fit. Run it, see how many words one honest minute holds, then paste your
own hook into `sample_hook` and re-run to test your script's length.

```python
# Pacing calculator — how long will your script actually take to say?
RATE_WPM = 130   # a clear, unrushed pace; under pressure people speed up, so budget slow

def duration_seconds(script_text, wpm=RATE_WPM):
    # Words in the script, divided by words-per-second, gives seconds aloud.
    n_words = len(script_text.split())
    return n_words / (wpm / 60.0)

# A sample 1-minute hook: hook -> claim -> boundary sentence -> invitation.
sample_hook = (
    "Every morning thousands of Purdue commuters trade sleep for a drive. "
    "In my sample of 812 of them, the longer the commute, the lower the "
    "reported class attendance. I want to be precise: this is an association "
    "in a survey, not proof that the drive itself causes the absences. "
    "Come see the pattern and tell me what you think could explain it."
)

words = len(sample_hook.split())
secs = duration_seconds(sample_hook)
budget_60 = int(60 * RATE_WPM / 60)
budget_180 = int(180 * RATE_WPM / 60)
print(f"Sample hook: {words} words -> {secs:.0f} seconds at {RATE_WPM} wpm (budget 60s).")
print(f"At {RATE_WPM} wpm, 60s holds about {budget_60} words; 180s holds about {budget_180} words.")

# Self-checks: the math is an identity, and the sample hook fits its minute.
assert abs(duration_seconds("w " * 130) - 60.0) < 1e-9, "pacing math is wrong"
assert secs <= 60, "sample hook overruns its minute — cut words, never speed up"
print("✓ Pacing math checks out — 130 words is one honest minute; the sample hook fits.")
```

spoken verb outruns the written one:

| The poster wrote (honest) | The mouth said (inflated) | The upgrade |
|---|---|---|
| "is associated with" | "causes" / "makes" | correlation → causation |
| "in my sample of 812" | "in undergraduates everywhere" | sample → population |
| "predicts, on our data" | "will happen" | forecast → certainty |

Every row is the same crime: the boundary moved while nobody was watching the
words. Your boundary sentence exists to hold that line out loud.

## 3. Uncertainty and Limitations, Spoken Like a Scientist

> *"When a presenter tells me, unprompted, exactly what their study *can't* show,
> I trust everything else they said more. When they insist it's airtight, I start
> looking for the crack myself. Knowing where your claim stops is the most expert
> thing you can say."*
> — a **policy stakeholder** who has heard a thousand pitches

Here is the counterintuitive heart of scientific communication: **stating your
limitations makes you more believable, not less.** A limitation, said first-person
with a straight spine, is *expertise* — it proves you know exactly what your
evidence can and cannot bear. The move that does this is the **ULN move**
(milestone **M14** begins here; the M13 scripts are **due tonight, Wednesday,
November 11**):

- **U — Uncertainty:** how much wiggle is around your estimate, and why (the
  interval, the sample size).
- **L — Limitation:** what your design cannot establish (the boundary itself,
  named plainly).
- **N — Next step:** the design that *could* push the boundary — the study you'd
  run next.

The next step is what turns a limitation from an apology into a research agenda.
And the ULN move has **two failure modes**, both of which misstate the evidence:

- **Hiding** — omit the limit and hope no one asks. Someone always asks, and the
  omission looks worse than the limit ever would.
- **Drowning — the apology spiral:** *"honestly this is probably all wrong, my
  sample's tiny, I'm not a statistician…"* This hides a real, defensible finding
  behind self-erasure and tells the visitor nothing about where the claim
  actually stops.

A limitation is not an apology. It is the edge of your claim, drawn by someone who
knows where it is.

> **A question that often comes up here:** *"If I lead with what my study can't
> do, won't people just dismiss it?"* They dismiss the *drowner*, not the
> *namer*. "This is an association in a convenience sample; it can't establish
> cause; a panel study could" is three seconds of pure competence — it tells the
> visitor you're the person in the room who most precisely knows the finding's
> shape. Self-erasure ("it's probably all wrong") is what gets dismissed, because
> you dismissed it first.

---

### Sentence stems for the ULN move

Speak the move with a straight spine. These stems keep you out of the apology
spiral:

- **U:** "My estimate carries a wide interval because ___ — so I'd read it as ___,
  not as a precise number."
- **L:** "This is a(n) ___ design in a ___ frame, so it cannot establish ___ —
  what it *does* show is ___."
- **N:** "The study that could settle ___ is ___; that's the next step."

Notice each stem *ends on strength*: the U ends on what to read it as, the L ends
on what it does show, the N ends on a concrete study. You never leave the listener
on the doubt.

## 4. The Hard Questions — Building and Drilling Your Bank

> *"I ask every presenter the question I think their poster is dodging. The ones
> who have an honest answer ready — even if it's 'I don't know, but here's how
> you'd find out' — I remember. The ones who bluff, I catch, and then I doubt the
> rest."*
> — a **skeptical peer** working the poster hall

At the Expo a stranger will ask you the hardest fair question about your project,
and how you answer decides whether they leave trusting you. The instrument that
turns that from an ambush into a strength is the **hard-question bank** (milestone
**M14**, completed tonight — **due Friday, November 13, 11:59 PM**). Five hardest
fair questions, one of each recurring type, each with a prepared answer:

1. **Method** — "Why this analysis and not another?"
2. **Alternative explanation** — "Couldn't X explain your result instead?"
3. **So-what** — "Why does this matter?"
4. **Generalization** — "Does this hold beyond your sample?"
5. **AI-use** — "What did you use AI for, and how do you know it's right?"

At least one answer must be an **honest don't-know**, in the full form:
*"I don't know — but here is what I DO know that bears on it, and here is how one
would find out."* A prepared don't-know reads as confidence; a **bluff**, once
caught, sinks everything you said. It is the single move that can destroy a
presenter's credibility entirely.

**Defend vs. concede** is the judgment underneath all five. You *defend* when the
question rests on a misreading and your evidence answers it. You *concede* when the
question names a real limit — and a concession, spoken in ULN form, is not a loss;
it's the expertise from the last section, on demand.

> **A question that often comes up here:** *"Isn't saying 'I don't know' at my own
> poster a disaster?"* The opposite — but only in the full form. "I don't know" as
> a shrug ends the conversation and reads as unpreparedness. "I don't know whether
> it holds for graduate peers — my frame was undergraduates; what I *do* know is
> the pattern was stable across three undergraduate years; to find out you'd re-run
> it on a graduate sample" reads as someone who knows *exactly* where their
> knowledge stops. That precision is the most credible thing at the poster.
