## 2. The Two Architectures: the Hook and the Walk

**Guiding question:** *what are the two spoken pieces a poster session demands, and how do you compress a claim for speech without inflating it?*

A poster session asks for two different spoken pieces, and they do different
jobs. Both go into milestone **M13**, your presentation plan, submitted on
Brightspace. You deliver the first one to a partner in class today.

**The 1-minute hook** is the 60 seconds that earn a stop. Four beats:

1. **Hook.** One sentence that earns attention: a stakes line, a surprising
   number, a shared frustration.
2. **Claim.** Your headline, said plainly.
3. **Boundary sentence.** The words that keep the claim honest ("this is an
   association, not proof that…"). Mark this sentence and rehearse it until it
   is automatic. It is the one pressure will try to drop.
4. **Invitation.** A reason to look closer ("come see the pattern and tell me
   what you'd expect").

**The 3-minute walk** is the guided tour for the visitor who stopped. It moves
their eye and yours along the poster's read path: headline first, then the
evidence blocks in your ranked order, then the uncertainty-and-limitations
moment (you reach it out loud, you do not skip it), and out. The walk spends
its time on the evidence that earns it.

The rule that governs both: **compress, never inflate.** Speaking is faster
than reading, so you drop words. But every word that survives must sit inside
the same boundary your poster's headline sits inside.

> **A question that often comes up here:** *"How is compressing different from
> inflating? Both change the claim."* Compression drops detail while keeping
> the claim's boundary. "In my sample of 812 commuters, longer commutes go with
> lower attendance" compresses to "in my survey, longer commutes track with
> skipping more class." Inflation moves the boundary: "longer commutes *make*
> people skip class." Same length, different sin. One lost a number; the other
> lost the truth.

### 🛠️ The pacing calculator

The cell converts words into seconds: words divided by words-per-second. It
uses a clear, unrushed presentation pace, deliberately on the slow side. Under
pressure people speed up, and a script that only fits when you rush is a script
that does not fit. Run it, see how many words one honest minute holds, then
paste your own hook into `sample_hook` and re-run to test your script's
length.

```python
# Pacing calculator — how long will your script actually take to say?
RATE_WPM = 130   # a clear, unrushed pace; under pressure people speed up, so budget slow

def duration_seconds(script_text, wpm=RATE_WPM):
    # Words in the script, divided by words-per-second, gives seconds aloud.
    n_words = len(script_text.split())
    return n_words / (wpm / 60.0)

# A sample 1-minute hook: hook -> claim -> boundary sentence -> invitation.
sample_hook = (
    "Every morning thousands of Purdue commuters trade sleep for a drive. "
    "In my sample of 812 of them, the longer the commute, the lower the "
    "reported class attendance. I want to be precise: this is an association "
    "in a survey, not proof that the drive itself causes the absences. "
    "Come see the pattern and tell me what you think could explain it."
)

words = len(sample_hook.split())
secs = duration_seconds(sample_hook)
budget_60 = int(60 * RATE_WPM / 60)
budget_180 = int(180 * RATE_WPM / 60)
print(f"Sample hook: {words} words -> {secs:.0f} seconds at {RATE_WPM} wpm (budget 60s).")
print(f"At {RATE_WPM} wpm, 60s holds about {budget_60} words; 180s holds about {budget_180} words.")

# Self-checks: the math is an identity, and the sample hook fits its minute.
assert abs(duration_seconds("w " * 130) - 60.0) < 1e-9, "pacing math is wrong"
assert secs <= 60, "sample hook overruns its minute — cut words, never speed up"
print("✓ Pacing math checks out — 130 words is one honest minute; the sample hook fits.")
```

### The excited-mouth upgrade: the one failure to hunt

Because pacing is blind to it, you hunt the upgrade by ear. It is the moment
the spoken verb outruns the written one:

| The poster wrote (honest) | The mouth said (inflated) | The upgrade |
|---|---|---|
| "is associated with" | "causes" / "makes" | correlation → causation |
| "in my sample of 812" | "in undergraduates everywhere" | sample → population |
| "predicts, on our data" | "will happen" | forecast → certainty |

Every row is the same mistake: the boundary moved while nobody was watching the
words. Your boundary sentence exists to hold that line out loud.

## 3. Uncertainty and Limitations, Spoken Like a Scientist

> *"When a presenter tells me, unprompted, exactly what their study can't show,
> I trust everything else they said more. When they insist it's airtight, I
> start looking for the crack myself. Knowing where your claim stops is the
> most expert thing you can say."*
> — a **policy stakeholder** who has heard a thousand pitches

Here is the counterintuitive heart of scientific communication: stating your
limitations makes you more believable, not less. A **limitation** is a true
sentence about what your design cannot establish. Example: "a one-time survey
cannot show that commuting causes absences." Said in first person with a
straight spine, a limitation is expertise. It proves you know exactly what your
evidence can and cannot bear.

Its partner is the **uncertainty statement**: a sentence that says how much
your number could wobble and why. Example: "with 812 people, my estimate could
easily be a few points higher or lower." Together they draw the honest edge
around your claim. The move that delivers both out loud is the **ULN move**:

- **U, uncertainty:** how much wiggle is around your estimate, and why (the
  interval, the sample size).
- **L, limitation:** what your design cannot establish, named plainly. This is
  the boundary itself.
- **N, next step:** the design that *could* push the boundary. It is the study
  you would run next.

The next step is what turns a limitation from an apology into a research
agenda.

### Hedging vs. precision: the two ways to fail the move

The ULN move has two failure modes, and both misstate the evidence:

- **Hiding.** Omit the limit and hope no one asks. Someone always asks, and
  the omission looks worse than the limit ever would.
- **Drowning, the apology spiral.** *"Honestly this is probably all wrong, my
  sample's tiny, I'm not a statistician…"* This buries a real, defensible
  finding under self-erasure and tells the visitor nothing about where the
  claim actually stops.

The difference underneath is **hedging vs. precision**. Hedging is vague
self-protection: "sort of," "probably wrong," "I guess," words that blur where
the claim stops. Precision names the exact boundary: "this is an association in
one campus sample." Hedging protects you; precision informs the listener. A
limitation is not an apology. It is the edge of your claim, drawn by someone
who knows where it is.

> **A question that often comes up here:** *"If I lead with what my study can't
> do, won't people just dismiss it?"* They dismiss the drowner, not the namer.
> "This is an association in a convenience sample; it can't establish cause; a
> panel study could" is three seconds of pure competence. It tells the visitor
> you are the person in the room who most precisely knows the finding's shape.
> Self-erasure ("it's probably all wrong") is what gets dismissed, because you
> dismissed it first.

---

### Sentence stems for the ULN move

Speak the move with a straight spine. These stems keep you out of the apology
spiral:

- **U:** "My estimate carries a wide interval because ___, so I'd read it as
  ___, not as a precise number."
- **L:** "This is a(n) ___ design in a ___ frame, so it cannot establish ___.
  What it *does* show is ___."
- **N:** "The study that could settle ___ is ___. That's the next step."

Notice each stem *ends on strength*. The U ends on what to read it as, the L
ends on what it does show, and the N ends on a concrete study. You never leave
the listener on the doubt.

### From limits to questions: the question bank

At the Expo a stranger will ask you the hardest fair question about your
project, and how you answer decides whether they leave trusting you. The
instrument that turns that from an ambush into a strength is the **question
bank**: a short written list of the hardest fair questions a visitor could ask
you, each with a prepared answer. Think of it as a study guide you write for an
exam only you will take. Yours needs five questions, one of each recurring
type:

1. **Method.** "Why this analysis and not another?"
2. **Alternative explanation.** "Couldn't X explain your result instead?"
3. **So-what.** "Why does this matter?"
4. **Generalization.** "Does this hold beyond your sample?"
5. **AI-use.** "What did you use AI for, and how do you know it's right?"

At least one answer must be an **honest don't-know**, in the full form: *"I
don't know, but here is what I DO know that bears on it, and here is how one
would find out."* A prepared don't-know reads as confidence. A bluff, once
caught, sinks everything you said. It is the single move that can destroy a
presenter's credibility entirely.

**Defend vs. concede** is the judgment underneath all five. You *defend* when
the question rests on a misreading and your evidence answers it. You *concede*
when the question names a real limit. A concession spoken in ULN form is not a
loss; it is the expertise from this lecture, on demand. You drill the bank
under pressure at the hot-seat Friday studio, where M14 is due: a few minutes
of rapid questions from the room, drawn from your own bank plus surprises. What
the hot seat exposes is exactly what you revise before you submit.

> **A question that often comes up here:** *"Isn't saying 'I don't know' at my
> own poster a disaster?"* The opposite, but only in the full form. "I don't
> know" as a shrug ends the conversation and reads as unpreparedness. "I don't
> know whether it holds for graduate peers; my frame was undergraduates; what I
> *do* know is the pattern was stable across three undergraduate years; to find
> out you'd re-run it on a graduate sample" reads as someone who knows exactly
> where their knowledge stops. That precision is the most credible thing at the
> poster.

## 4. Dress Rehearsal: the Full Expo Simulation

> *"The difference between the presenters who shine and the ones who freeze is

The Expo is next. Today is the last chance to find out, in a safe room with

the logistics nailed down. **Rehearsal is the difference between nerves and
panic.** "I'll wing it" is the one sentence milestone **M15**, the
dress-rehearsal readiness audit, exists to make impossible.

**How the simulation runs.** Two **20-minute station rounds**. In each round, a
few posters go *live*, with their presenters standing beside them, while
everyone else rotates past as **conference visitors**, exercising the 1-minute

to real conditions. Then you swap, so everyone both presents and visits.

**The visitor role card.** When you visit, you are a critical-but-kind

**The readiness audit** is milestone **M15** itself: a signed in-class sheet,
not a Brightspace upload, completed before you leave today. Sign only what is

> is one you now get to fix with time to spare instead of discovering at the
> Expo. The presenters who "wing it" are not braver; they just move the fumble
> from the practice room to the live audience. Sign the audit for what is
