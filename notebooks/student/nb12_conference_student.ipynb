## 1. The Last Gate: Rehearsal and the Readiness Audit

**Guiding question:** *the day before you present, which piece of your work is
actually ready, and which one just looks ready?*

> *"Do not tell me it is basically ready. Walk it for me, cold, inside the clock,
> and take three hard questions without notes. Whatever cracks under that is your
> tonight. The poster being printed tells me nothing about whether you can defend
> it."*
> — your **thesis advisor**, the day before the conference

The gap in the puzzle has a name. A **readiness audit** is a signed check that
every piece of your presentation is either confirmed ready or has a named plan to
be fixed tonight. Example: if you cannot yet check the box "rehearsed the full walk
twice, inside time," that unchecked box is not a failure. It is a task with a
deadline of tonight. The audit forbids exactly one claim: "I'll wing it tomorrow."

What almost never survives a real audit is the part a printed poster cannot show:
answering an unrehearsed question out loud, in time, without notes. So today you
stand your poster through a **full dress run**, a simulated conference round with
pitches, hard questions, and an uncertainty-and-limitations moment, all on the
clock, while everyone else visits you as a critical-but-kind stranger.

A visitor is not there to be nice. A **visitor role card** asks each person to
spread five kinds of question across the posters they visit, so you feel each kind
before a stranger asks it for real:

- **method**: "how exactly did you estimate that?"
- **alternative-explanation**: "could something you did not measure explain the gap?"
- **so-what**: "what changes if you are right?"
- **generalization**: "would this hold beyond the sample you studied?"
- **AI-use**: "did you use AI in the analysis, and how did you check it?"

> **A question that often comes up here:** *"Isn't rehearsing the day before too
> late to change anything?"* The day before is too late to change your design, and
> that is not what the dress run is for. It is for finding the one piece that is not
> ready while there is still a tonight to fix it: a pitch that runs long, a question
> that makes you stall, a backup file you never actually saved. Small, fixable, and
> invisible until you run it cold.

The other thing a dress run gives you is a plan for tomorrow. You cannot code
feedback you did not capture, so before you present you decide how you will record
it. **Capturing questions as data** means keeping a light log in the moment: after
each visitor leaves, jot one line, the question's gist plus a one-word tag, on a
phone note or an index card you keep in your pocket. Star anything you hear more
than once. That log is your **coding sheet**, and it is the difference between real
data on the day and a vague memory by the time reflection comes around.

The next section turns that sheet into a table you can count.

## 2. Feedback Is Data. Code It Like Data.

**Guiding question:** *how do you turn a blur of an afternoon's questions into a
table you can count, and keep what the crowd enjoyed separate from what your
evidence can actually support?*

> *"Show me the questions, not the mood. A room can love a poster whose central
> claim is wrong. I want to know which of your claims a stranger refused to give
> you for free, and how many strangers refused it."*
> — a **journal reviewer**, reading your reflection

**Feedback as data** is the idea that audience questions are observations about
your project, to be recorded and counted rather than merely felt. Example: "would
this hold on another campus?" is neither an insult nor a compliment. It is a data
point about which claim a stranger will not grant you without more evidence.

The tool is **coding**, in the qualitative-research sense: sorting free-form
questions into a small set of labeled categories so you can count them. When you
file "I couldn't tell which line was your comparison group" under a clarity
problem, you are coding. Every question on your sheet gets three codes:

- **Type**: what kind of question was it? One of *method*, *alternative-explanation*,
  *so-what*, *generalization*, *AI-use*, or *other*.
- **Target**: did it aim at your **claims** (whether your result is true or properly
  bounded) or your **communication** (whether a listener could follow you)?
- **Recurrence**: how many times you got a version of it. Asked once, it is an
  anecdote. Asked three times, it is a finding about your project.

The **target** code does the real work, and it splits into two signals you must
never pile together. A **communication signal** is a question that says your
telling failed: "I couldn't tell which line was the comparison group." Your figure
failed, not your evidence, and the fix is the poster. A **claims signal** is a
question that says your evidence was doubted: "would this hold outside your sample?"
The fix may change what you are allowed to claim. Sorting the two is the whole
point.

> **A question that often comes up here:** *"A comment feels like it hits both my
> claims and my communication at once. Which column does it go in?"* Tag the target
> where the **fix** lives, not where the sting lives. "Your comparison group might
> differ from your treatment group" stings like a communication problem, because it
> sounds like you explained badly. But it doubts your evidence, so the fix is in
> your claim, and it codes to **claims**. When in doubt, ask: would a clearer poster
> make this question go away? If yes, communication. If a clearer poster leaves the
> doubt standing, claims.

```python
# A routing diagram: every audience question flows to one of three places.
# (networkx is imported once in the setup cell.)
G = nx.DiGraph()
pos = {
    "An audience\nquestion":              (0.0, 1.0),
    "CLAIMS signal\n(probe the evidence)":     (3.2, 1.9),
    "COMMUNICATION signal\n(fix the poster)":  (3.2, 1.0),
    "NOISE\n(log it, no fix)":                 (3.2, 0.1),
}
edges = [
    ("An audience\nquestion", "CLAIMS signal\n(probe the evidence)"),
    ("An audience\nquestion", "COMMUNICATION signal\n(fix the poster)"),
    ("An audience\nquestion", "NOISE\n(log it, no fix)"),
]
G.add_edges_from(edges)

fig, ax = plt.subplots(figsize=(10, 4.8))
node_colors = ["#DCE6F1", "#4C72B0", "#55A868", "#B0B0B0"]
nx.draw_networkx_nodes(G, pos, node_color=node_colors, edgecolors="#333333",
                       linewidths=1.5, node_size=6600, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=8.5, ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=18, edge_color="#555555",
                       width=1.6, node_size=6600, ax=ax)
edge_labels = {
    ("An audience\nquestion", "CLAIMS signal\n(probe the evidence)"): "doubts your evidence",
    ("An audience\nquestion", "COMMUNICATION signal\n(fix the poster)"): "could not follow you",
    ("An audience\nquestion", "NOISE\n(log it, no fix)"): "no question, nothing to fix",
}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8,
                             label_pos=0.55, ax=ax)
ax.set_title("Every audience question routes to one of three targets")
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ Routing diagram drawn.")
print("  Claims and communication demand DIFFERENT fixes, so they never share a column.")
print("  Noise (warmth, an aside, no question) is logged but changes nothing.")
```

**Reading the diagram.** Every question you were asked flows to exactly one of
three targets. A **claims** question doubts your evidence and may change your claim.
A **communication** question could not follow you and is fixed on the poster.
**Noise** is a reaction with no question and nothing to fix, such as "nice colors."
The danger is piling all three into one warm blur, because then the recurring
claims doubt hides inside the applause.

*In plain terms, the picture says sort every question by what it would take to
answer it: better evidence, a clearer poster, or nothing at all.*

## 3. From the Tally to Your Reflection and Your Criticism Portfolio

**Guiding question:** *once you can see the pattern, what does an honest researcher
do with it, and how do you criticize a peer's poster the same fair way?*

A tally often hands you a **new research idea**: the recurring generalization doubt
suggests a second site to sample, the alternative-explanation doubt suggests a
variable you should have measured. That idea is worth stating. It is also easy to
overstate, because it feels like the audience "agreed" with it. The audience did not
agree with anything; it asked questions. Whether your idea actually follows from the
pattern is your call to defend.

You will also read three peers' posters the same disciplined way, and assemble a
**poster-criticism portfolio**: three structured, kind-but-critical readings. Each
practices **structured criticism**, which means feedback tied to a specific axis and
to the evidence, not a vibe. Example: instead of "the poster was confusing," a
structured criticism says "the central claim is causal, but the design is
observational, so the poster shows association where it states cause." That names the
axis (claims), points at the evidence, and could be checked by anyone.

> **A question that often comes up here:** *"Isn't harsh criticism of a classmate
> unkind?"* Structured criticism is the opposite of unkind. Vague praise leaves your
> peer with nothing to fix. Naming exactly where their evidence stops short, on a
> stated axis, is the most useful thing a reader can give them, and it is exactly
> what you want in return. Kind-but-critical means you attack the claim, never the
> person.

The failure mode that ruins a reflection is not bad coding. It is having nothing to
code, because you trusted memory and reconstructed the afternoon a day later. The
next cell tests that. It expands the worked tally into one row per asked question,
then simulates "remembering only some of them" by keeping a random subset, and

**🏠 Homework depth: run this on your own.**
**Before you ask:** predict, in one sentence, whether a partial recall can make the
communication column look heavier than the claims column when it truly is not.

> 💡 **Gemini Prompt:** "Here is a Python cell that expands a coded tally into one row
> per question, then draws random subsets to imitate recalling only part of an
> afternoon: [paste the next cell]. Do not just re-explain each line. Instead: (1)
> point to the one line that fixes the full-sheet claims share and the one line that
> makes each memory-only recall wobble, and (2) give me one independent way to confirm
> the FULL-sheet claims share the cell reports, by hand, without rerunning this code."
>
> **After running, verify (counters *confident fabrication*: a fluent, confident
> answer can name a full-sheet claims share the code never produced):**
> - [ ] Read the printed full-sheet claims share off your own output. Confirm
>   Gemini's answer uses that number, not one it guessed.
> - [ ] Run its "independent way to confirm" yourself. Counting the claims rows by
>   hand should give the same share; if its method gives a different number, trust
>   your own count.
> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# Why you capture in the moment: memory resamples your afternoon.
# Expand the tally to one row per asked question (17 total), then keep a random
# subset to imitate recalling only part of the day.
instances = tally.loc[tally.index.repeat(tally["count"])].reset_index(drop=True)
true_claims_share = (instances["target"] == "claims").mean()
print(f"Full coding sheet: {len(instances)} questions logged, "
      f"claims share = {true_claims_share:.2f}  ({int((instances['target']=='claims').sum())} of {len(instances)})")
print()

remembered_shares = []
for i in range(3):
    idx = rng.choice(len(instances), size=8, replace=False)  # remember only 8
    share = (instances.iloc[idx]["target"] == "claims").mean()
    remembered_shares.append(share)
    print(f"  Memory-only recall #{i + 1} (8 of {len(instances)} questions): "
          f"claims share = {share:.2f}")

# Self-check: the full sheet is fixed; the from-memory recalls disagree with each other.
assert abs(true_claims_share - 11 / 17) < 1e-9, "full-sheet claims share should be 11/17"
assert len(set(round(s, 3) for s in remembered_shares)) > 1, "recalls should not all agree"
print()
print("✓ Self-check passed: the full sheet's claims share is fixed at 0.65, but the "
      "three\n  from-memory recalls disagree. Capture in the moment; do not reconstruct.")
```

**Reading the evidence.** The full coding sheet is fixed: its claims share is
about 0.65, that is 11 of 17 questions. The three from-memory recalls do not agree
with each other, because each one is a small sample of the afternoon, and small
samples wobble. Reconstructing your Expo from memory a day later is exactly this kind
of resampling, and it can quietly flip which target looks heavier. The lesson is the
one your coding sheet exists to enforce: log every question in the moment, so your
reflection reads the whole afternoon and not a lucky handful of it.
