## 🎯 Definition of Done

You are done with this sprint when your **presentation package**, built for *your
own* project, carries all five parts, each sitting inside the compass boundary
your poster locked:

1. **Three pitch layers.** The **30-second hook** with its **boundary sentence
   marked**, the **90-second walk**, and the **full pitch** (3 to 5 minutes), each
   sized to the same poster and none upgrading its claim.
2. **The poster-walk map.** What your hand or eye is on at each beat, mapped to a
   poster element, so a stranger could follow the walk with their eyes.
3. **The spoken uncertainty-and-limitations statement.** Your **ULN move**:
   uncertainty, your single most threatening limitation, and a next step, in a
   straight spine, no **apology spiral**.
4. **The five-type question bank.** One hard, fair question per type (method,
   alternative explanation, so-what, generalization, AI-use), each with a two-line
   answer marked **defend** or **concede** consistent with your claim ledger, plus
   one prepared **honest don't-know**.
5. **The AI-use answer** that points a visitor at your ledger and your verify step.

Plus the **AI Research Ledger** rows and the **dossier update line**. Then deliver
the 30-second hook at the mock symposium and revise from every spoken upgrade a
partner or the AI reviewer flags.

> **A question that often comes up here:** *"The poster is locked. Isn't the hard
> part behind me?"* No. The poster is a claim you stand behind; the package is a
> claim you stand beside, cold, under a stranger's eye contact. That is where a
> rehearsed presenter compresses and, compressing, quietly upgrades the claim. The
> five parts above are the discipline that keeps the spoken version inside the
> boundary the written one already bought.

### 🛠️ Hands-On: pace the hook, catch the apology spiral

Run the scaffold below on its default hook and uncertainty line first, to see what
it does. Then edit the **three values** at the top with *your own* words and rerun.
Everything downstream is derived from those edits, so there is one source of truth
per number: paste once at the top and the whole readout stays consistent. The
counts are deterministic, and the printed number, not your guess, settles the
truth.

```python
# --- Delivery scaffold — edit ONLY the three values below with YOUR words.
# Everything after them is DERIVED, so there is one source of truth per number.
# The defaults let the cell run before you paste anything of your own.

MY_HOOK = (
    "In my sample of 812 commuters, longer commutes go hand in hand with lower "
    "attendance. That is an association, not proof that commuting causes it. "
    "Come see the pattern, and tell me what would break it."
)  # EDIT ME: your 30-second hook, boundary sentence included

MY_ULN = (
    "This holds for the commuters my survey reached. Whether it reaches every "
    "commuter is the open question, and the check that would settle it is a "
    "sampling frame that includes the ones I missed."
)  # EDIT ME: your spoken uncertainty-and-limitations line

SPEAKING_WORDS_PER_MIN = 130   # a calm conference pace; raise it if you speak fast

# ---- derived below; do not retype any number, read it from the edits above ----

# 1) Pacing: does the hook fit its 30-second window at a calm pace?
hook_words = len(MY_HOOK.split())
hook_secs = round(hook_words / SPEAKING_WORDS_PER_MIN * 60, 1)
print("PACING — the 30-second hook")
print(f"  {hook_words} words -> {hook_secs} s at {SPEAKING_WORDS_PER_MIN} wpm  -> " +
      ("fits the 30-second window" if hook_secs <= 30 else
       "OVER: cut words until it fits, but never cut the boundary sentence"))

# 2) Apology spiral: self-erasing phrases that delete a finding instead of bounding it.
APOLOGY_PHRASES = ["sorry", "just", "probably", "i guess", "i'm no", "not sure",
                   "sort of", "kind of", "maybe", "possibly", "all wrong",
                   "can't really"]
uln_low = MY_ULN.lower()
hits = [p for p in APOLOGY_PHRASES if p in uln_low]
print("\nAPOLOGY-SPIRAL CHECK — your spoken uncertainty line")
print(f"  self-erasing phrases found: {len(hits)}" +
      (f"  -> {hits}: swap each for a bounded next step" if hits
       else "  -> none; the spine stays straight"))
NEXT_STEP_CUES = ["would settle", "would answer", "next step", "the check", "the test"]
ends_on_next_step = any(cue in uln_low for cue in NEXT_STEP_CUES)
print("  ends on a next step: " +
      ("yes, the boundary lands as a plan, not an apology" if ends_on_next_step
       else "no, add the one check that would move the limitation forward"))

print("\n✓ Scaffold ran on the defaults. Now paste YOUR hook and ULN above,")
print("  rerun, and trust the printed counts over any tool's prediction. seed", SEED)
```

### 🗡️ Red-Team Exchange

The mock symposium's **29–41 minute block**. Pair with one partner and run two
rounds of attack, then bring in an AI reviewer as a third voice using the **Poster
Critic** role card (`ai_resources/agent_role_cards.md`; full briefing in
`genai_studio/roles/poster_critic.md`). It reads how your pitch *communicates*:
untraceable numbers, claim boundaries, and whether uncertainty is audible. It
proposes; it cannot decide what your evidence supports.

**Round 1: attack the spoken boundary (about 6 minutes).**
Your partner is a hurried visitor. You deliver the 30-second hook cold; they flag
every place the spoken claim reached past your poster, the **excited-mouth
upgrade** most of all (a *make* or *causes* where the poster says *is associated
with*). Paste the flagged line to the Poster Critic and ask it to name the reach
the words are not entitled to. You defend or concede, then fix the wording and
mark the boundary sentence.

**Round 2: attack the question bank and the ULN (about 6 minutes).**
Your partner carries the **visitor role card** and asks one question of each of
the five types (method, alternative explanation, so-what, generalization, AI-use).
They press the two that scare you most: is your answer grounded in real evidence,
or does it lean on a number with no path back to your data? Then they hear your
uncertainty line and check that it lands as a bounded next step, not an apology
spiral. You either bound the limitation honestly or concede it.

**What to log.** Write down the single strongest hit from each round and how you
changed the package in response. A challenge that changed nothing still earns a
line saying why it did not land. That record is part of what you submit.

**Mind the correlated-error trap.** If your partner and the Poster Critic both call
your pitch bounded, that is not two independent votes. Neither delivers the pitch
under real pressure. Only the mock symposium's live traffic breaks the tie.

> **A question that often comes up here:** *"What if my partner and the Poster
> Critic disagree about whether a line upgrades my claim?"* Good. You are the
> judge, not either of them. Say in one sentence which case you found more
> convincing and why, checked against your own compass classification. Adjudicating
> that gap is stronger proof you understand your boundary than agreeing with a
> single reviewer would be.
