## 🎯 Definition of Done

You are done with this sprint when your submission carries a **locked poster**
plus the records behind it, all applied to **your own project**. The lock is
terminal, so the honesty has to be built in before the file leaves your hands.
Six pieces are graded, and every one is a skill you practiced this week:

1. **The locked poster.** Print-ready to URC spec and legible at print scale. The
   headline sits inside its **compass position** (the kind and reach of the
   question your project answered), the **uncertainty rides in the same eye-span**
   as the claim, and the board is **reachable** by a color-blind reader and a
   screen-reader user, not only a fast sighted one.
2. **Every number traces to a cell.** Each percentage, estimate, interval, and
   count points to the notebook cell that produced it. A number you cannot trace is
   an **untraceable number**, and it cannot go on the board.
3. **Every citation resolves.** Each source your poster names loads as a real,
   retrievable work you opened yourself, and it says what your poster claims.
4. **The gallery-walk audit record.** The four audits (claim-boundary,
   figure-honesty, read-path, uncertainty) plus accessibility, run on your own
   poster, each written as a located hit with its fix.
5. **The oral-defense record.** The three lenses (outsider, methods specialist,
   skeptic), your **defend-versus-concede** call for each, one **honest-don't-know**
   for the question your evidence cannot fully settle, and the claims you softened.
6. **The AI-use disclosure block, the AI Research Ledger, and the dossier line.**
   A short disclosure on the board naming which tool did which task and how you
   checked it, the full ledger behind it, and one line locating your updated
   dossier.

> **A question that often comes up here:** *"My poster reads fine already. Why is
> the lock a separate graded step?"* Because a poster that reads well can still
> carry a number no cell produces or a citation that does not resolve. Reading was
> last week's work. The lock asks a different question: is every claim on the board
> actually **backed**? A judge who points at one number and hears "it came from the
> analysis somewhere" stops trusting the whole board.

```python
# Poster number-to-cell traceability check — the M10 lock discipline, made runnable.
# The lock rule: every number on the board must reproduce from a cell, and the
# headline's uncertainty must ride in the same eye-span as the claim. This scaffold
# recomputes your headline number from a seeded stand-in, then checks each number you
# plan to print against what the cell actually produces. A number that does not
# reproduce is a STALE or UNTRACEABLE number, the M10 hard cap. Everything flows from
# ONE source of truth (CELL_VALUE), so there is no second copy to drift out of sync.

# --- The CELL side: recompute your headline number from your data. ----------
# PLACEHOLDER: a seeded stand-in so the check runs before your real data are wired in.
# Swap this block for your own M8/M9 computation, keeping CELL_VALUE and the interval.
n_per    = 400                                  # EDIT ME — your per-group size
group_hi = rng.normal( 0.35, 1.0, n_per)        # EDIT ME — replace with your real numbers
group_lo = rng.normal(-0.35, 1.0, n_per)
CELL_VALUE = group_hi.mean() - group_lo.mean()
se = np.sqrt(group_hi.var(ddof=1) / n_per + group_lo.var(ddof=1) / n_per)
CELL_LOW, CELL_HIGH = CELL_VALUE - 1.96 * se, CELL_VALUE + 1.96 * se
# ----------------------------------------------------------------------------

# --- The POSTER side: the numbers you plan to PRINT on the board. -----------
# Map each label to the value your poster shows. Here the first is set from the cell
# (so it traces) and the second is left deliberately stale to show the warning path.
# Replace BOTH with the numbers literally on YOUR poster.
poster_numbers = {
    "headline gap (traces)":    round(CELL_VALUE, 2),         # EDIT ME
    "old draft number (stale)": round(CELL_VALUE + 0.40, 2),  # EDIT ME — delete once real
}

print(f"Cell recomputes the headline: {CELL_VALUE:+.2f} "
      f"(95% interval {CELL_LOW:+.2f} to {CELL_HIGH:+.2f})\n")
for label, printed in poster_numbers.items():
    traces = abs(printed - CELL_VALUE) < 0.01
    mark = "✓ traces to the cell" if traces else "⚠️ does NOT reproduce: fix before the lock"
    print(f"  {label:26s}: poster prints {printed:+.2f}  ->  {mark}")

print("\nThe interval above is what must ride beside your headline, not in a footnote.")
print("✓ Ran on PLACEHOLDER data — swap in your own number before you trust this.")
```

### 🗡️ Red-Team Exchange

The studio's **29–41 minute block**. Pair with one partner and bring in the GenAI
Studio **Poster Critic** as a third voice. Cite it in your ledger by its role card
(`genai_studio/roles/poster_critic.md`): it audits how your poster *communicates*,
hunting untraceable numbers and boundary slips on the pasted draft, while your

**Round 1: attack the board (about 6 minutes).**
Your partner reruns your notebook and checks that every number on your poster matches

(`project/poster/redteam_review_protocol.md`): does the headline stay inside its

**Round 2: stand the oral defense (about 6 minutes).**
Your partner runs the Poster Critic on your headline and figure descriptions and reads

or fix it, and reword any sentence that upgrades a sample to a population or an

methods specialist, skeptic), and you call **defend or concede** out loud on each,
