## 1. The Four Parts of Every Research Design

**Guiding question:** *what are the four parts every research design must have,
and what must the world be like for your question to have an answer at all?*

> *"Before you collect a single data point, walk me through four things: the
> world you are assuming, the exact number you want out of it, how the data will
> come to you, and how you will turn that data into your number. Leave one out
> and you do not have a design yet. You have a hope."*
> — a **thesis advisor**, on the first meeting about your project

Every research design, from a two-person survey to a national experiment, has
the same four parts. The course calls them **MIDA**, one letter each.

- **Model (M)**: your written-down picture of how the world could work, meaning
  which things exist and what could affect what. Example: "belonging runs 0 to
  100, and a peer mentor could nudge it up."
- **Inquiry (I)**: the exact quantity you want from that world, named before any
  data arrive. Example: "the average change in belonging a mentor causes."
- **Data strategy (D)**: your plan for how the data come to exist, meaning who
  gets sampled and who gets which condition. Example: "flip a coin to decide who
  gets a mentor, then survey everyone."
- **Answer strategy (A)**: your plan for turning the data into an estimate of the
  inquiry. Example: "subtract the average belonging of the unmentored from the
  mentored."

The four are not a checklist you fill in any order. They have to **align**:
point at the same quantity. **Alignment** means your data and answer strategies
actually reach the inquiry your model makes askable. A population inquiry served
by a convenience sample is misaligned. A causal inquiry with a data strategy that
never varies the cause is misaligned. Run the next cell to see the four parts as
one picture.

```python
# A plain schematic of the four parts of a research design (MIDA).
fig, ax = plt.subplots(figsize=(11, 3.2))
parts = [
    ("M", "Model", "the world your\nquestion assumes"),
    ("I", "Inquiry", "the exact number\nyou want"),
    ("D", "Data strategy", "how the data\ncome to exist"),
    ("A", "Answer strategy", "data turned into\nan estimate"),
]
x_centers = [1.5, 4.0, 6.5, 9.0]
for (letter, name, blurb), xc in zip(parts, x_centers):
    ax.add_patch(plt.Rectangle((xc - 1.0, 0.4), 2.0, 1.7, facecolor="#DCE6F1",
                               edgecolor="#4C72B0", linewidth=1.6))
    ax.text(xc, 1.72, letter, ha="center", va="center", fontsize=20, fontweight="bold")
    ax.text(xc, 1.28, name, ha="center", va="center", fontsize=11, fontweight="bold")
    ax.text(xc, 0.78, blurb, ha="center", va="center", fontsize=8.5)

for xc in x_centers[:-1]:
    ax.annotate("", xy=(xc + 1.5, 1.25), xytext=(xc + 1.0, 1.25),
                arrowprops=dict(arrowstyle="-|>", color="#555555", lw=1.6))

ax.text(5.25, 2.55, "the four parts must ALIGN — all point at the same quantity",
        ha="center", va="center", fontsize=10, style="italic", color="#333333")
ax.set_xlim(0, 10.5)
ax.set_ylim(0, 3.0)
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ MIDA drawn: Model -> Inquiry -> Data strategy -> Answer strategy.")
print("  This notebook builds M and I in Lecture 1, then diagnoses whether D and A")
print("  actually reach that inquiry in Lecture 2.")
```

*In plain terms, the picture says a research design is four commitments in a row,
and each one has to serve the same quantity. The rest of this notebook builds the
first two parts, then stress-tests whether the last two can deliver on them.*

> **A question that often comes up here:** *"Isn't 'the data strategy' just how I
> collect data? Why give it a fancy name?"* Because naming it as a design part
> forces you to decide it on purpose, before you are attached to a result. Most
> weak studies are not weak because someone collected data badly. They are weak
> because the data strategy was never designed to reach the inquiry, and no one
> noticed until the answer was already in hand.

## 2. The Model: Two Worlds You Can Write Down

**Guiding question:** *when you say a mentor "had an effect," which two worlds
are you secretly comparing, and could you ever stand in both?*

Start with the **model**, the M of MIDA. For a causal question, the model has to
hold two versions of each person. For every first-year, imagine two numbers.

- **Y0** is the belonging score they *would* show **without** a mentor.
- **Y1** is the belonging score they *would* show **with** one.

These are **potential outcomes**: the two outcomes a person could show you,
depending on which condition they get. Only one will actually happen. The one
that does not happen is the **counterfactual**, the outcome from the world that
did not occur. A person's own **individual treatment effect** is simply Y1 minus
Y0, how much the mentor moved *their* belonging.

You will work in one concrete **mentoring-program world** all notebook long. Each
person is a first-year. The outcome is a belonging score from 0 to 100. The
treatment is being paired with a peer mentor. Here is the trick only a simulation
allows: you will build a world where you can see **both** Y0 and Y1 for **every**
person at once. Reality never lets you do that. That is exactly why you start in
simulation, to look at the thing reality hides, name it, then watch it vanish.

> **A question that often comes up here:** *"Is the mentor's effect the same for
> everyone?"* In reality, almost never. A mentor might lift one person a lot and
> barely move another. The teaching world you are about to build sets that
> variation aside and gives everyone the same nudge, because a constant effect
> makes a crisp answer key you can check estimators against later. Simple first,
> rich later, on purpose.

### 🛠️ Run the Study: build a world you can see completely

Run the cell below. It defines the mentoring-program world and simulates 100
people, giving you both potential outcomes and each person's individual treatment
effect. This is the god's-eye view that only a simulation offers. Read the table
before you read the next markdown cell.

```python
# The mentoring-program world. Everything downstream derives from these
# three dials, so edit only here if you want to change the world.
N = 100
EFFECT = 2.0        # the constant nudge a mentor gives, in belonging points
BASE_NOISE = 2.0    # person-to-person wobble in baseline belonging

def make_world(n=N, effect=EFFECT, noise=BASE_NOISE, rng=rng):
    """Each person's belonging (0-100) without a mentor (Y0) and with one
    (Y1 = Y0 + effect). A simulation, so BOTH columns are visible."""
    Y0 = rng.normal(50, 10, n) + rng.normal(0, noise, n)
    return pd.DataFrame({"Y0": Y0.round(1), "Y1": (Y0 + effect).round(1)})

world = make_world()
world["individual_effect"] = (world["Y1"] - world["Y0"]).round(1)

print("✓ Simulated a 100-person mentoring-program world (both worlds visible).")
print("\nFirst 8 people — the god's-eye view:")
print(world.head(8).to_string())
print(f"\nIn THIS world every person's individual effect is the same: "
      f"{world['individual_effect'].iloc[0]} points — by deliberate construction.")
```

### Reality's move: one world per person

Now do the thing reality does. Each person either gets a mentor or does not, so
for each person **one** potential outcome becomes the number you observe and the
other silently vanishes into the world that did not happen. Run the next cell to
flip a coin for each person and keep only the outcome that really occurred.

The result is the table every real study is stuck with: a column of observed
outcomes and a ghost column where the counterfactual used to be. The individual
effect you just computed becomes uncomputable, because no person still holds both
numbers.

```python
# Reality's move: assign each person to mentor (Z=1) or not (Z=0) by a coin,
# then observe ONLY the potential outcome that matches their assignment.
world["Z"] = rng.integers(0, 2, len(world))                      # the coin flip
world["observed"] = np.where(world["Z"] == 1, world["Y1"], world["Y0"])

real_study = world[["Z", "observed"]].copy()
real_study["Y0_seen"] = np.where(world["Z"] == 0, world["Y0"], np.nan)
real_study["Y1_seen"] = np.where(world["Z"] == 1, world["Y1"], np.nan)
real_study["individual_effect"] = "?"   # can never be filled in

print("What a real study is stuck with (first 8 people):")
print(real_study.head(8).to_string())
print(f"\nMentored: {int(world['Z'].sum())} people   "
      f"Not mentored: {int((world['Z'] == 0).sum())} people")
print("\nThe individual_effect column is now unknowable for every single person.")
```

That missing column is not a data-collection failure a bigger budget could fix.
It is a logical wall: a person cannot both have and not have a mentor this
semester. This is the **fundamental problem of causal inference**: for any unit,
at most one potential outcome is ever observable. Every causal method later in
the course is, at heart, a disciplined way to reason about that missing column
without pretending you can see it.

## 3. Drawing the Model: DAGs

**Guiding question:** *how do you put a belief about what causes what on paper,
clearly enough that the world could prove you wrong?*

So far your model lives in prose. The standard way to draw it is a **DAG**, short
for directed acyclic graph. Do not let the name scare you. A DAG is a drawn story
about what could affect what, and you need no graph theory to read one.

- A **node** is a circle standing for one variable. Example: a circle labeled
  "peer mentor."
- An **arrow** (also called an edge) is a claim that one variable *could* directly
  influence another. Example: an arrow from "peer mentor" to "belonging" claims a
  mentor could move belonging.

The moment you draw the arrows, you commit to a story the world could contradict.
You also expose the variables that quietly threaten a causal claim. Suppose
first-years who were **already more involved on campus** are both more likely to
go seek a mentor *and* more likely to feel they belong anyway. Then prior
involvement is a **confounder**: a common cause of both the treatment and the
outcome. Example: an arrow from "prior involvement" into *both* "peer mentor" and
"belonging." A plain mentored-versus-unmentored comparison would then partly
measure involvement, not the mentor. That confounding arrow is the exact one the
two teams in the opening puzzle disagreed about.

> - [ ] Read the estimand sentence next to your question. Does it name the same
>   quantity and the same units you meant, or did it quietly swap in an easier one?

```python
# A model of the mentoring world, drawn from an edge list. Each arrow is a
# belief: "this variable could directly affect that one." Edit the edges to
# change the model; positions are keyed to the same node names.
edges = [
    ("prior\ninvolvement", "peer\nmentor"),   # already-involved first-years seek mentors
    ("prior\ninvolvement", "belonging"),        # ...and tend to feel they belong anyway
    ("peer\nmentor",       "belonging"),        # the causal arrow we actually care about
]
pos = {
    "prior\ninvolvement": (1.5, 2.0),
    "peer\nmentor":       (0.0, 0.0),
    "belonging":           (3.0, 0.0),
}
dag = nx.DiGraph(edges)

fig, ax = plt.subplots(figsize=(8, 5))
nx.draw_networkx_nodes(dag, pos, node_color="#DCE6F1", edgecolors="#4C72B0",
                       linewidths=1.6, node_size=6000, ax=ax)
nx.draw_networkx_labels(dag, pos, font_size=10, ax=ax)
nx.draw_networkx_edges(dag, pos, arrowstyle="-|>", arrowsize=22, edge_color="#555555",
                       width=1.7, node_size=6000, ax=ax)
ax.set_title("A three-node model: prior involvement confounds mentor → belonging")
ax.axis("off")
plt.tight_layout()
plt.show()

print("Nodes:", [n.replace(chr(10), ' ') for n in dag.nodes()])
print("Arrows (each is a claim about the real world):")
for a, b in dag.edges():
    print(f"  {a.replace(chr(10), ' ')}  ->  {b.replace(chr(10), ' ')}")
```

**Reading the diagram.** Prior involvement sits on top with an arrow into both
"peer mentor" and "belonging." That double arrow is what makes it a confounder.
Follow the danger: if involved first-years are the ones who seek mentors, then the
mentored group is also the more-involved group, and involved people feel they
belong anyway. A raw comparison credits the mentor for belonging that involvement
would have delivered on its own.

Here is the resolution of the opening puzzle. In our simulation, a **coin flip**
decided who got a mentor, so prior involvement could not steer assignment. That
coin flip erased the top-left arrow. Team B's lottery world is the coin-flip
world, and a plain comparison is trustworthy there. Team A's world keeps the
arrow live, and a plain comparison is contaminated. The evidence that would settle
which world is true: check whether involvement actually predicts who got a mentor.
If assignment was a real lottery, it does not. If people self-selected, it does.

*In plain terms, the picture says the top variable poisons a naive comparison,
and random assignment is the move that erases the poisoned arrow.*

## 4. Naming the Inquiry: The Genie Test

**Guiding question:** *what exact number do you want, and could a genie who knew
everything hand it to you without any data at all?*

The **inquiry** is the I in MIDA: the one exact quantity you want from your model.
Written as a formula it is called an **estimand**: the target number your study is
trying to estimate, named before any data arrive. Example: "the average of Y1
minus Y0 across all first-years," which is the average effect of a mentor.

A quick check keeps an inquiry honest. The **genie test**: imagine a genie who can
see your whole model, every potential outcome for every unit, with no missing
column. If the genie could compute your number on the spot, your inquiry is well
defined. If even the genie would ask "which number do you mean?", your inquiry is
still vague. Example: "does mentoring help?" fails the test, because help how much
and for whom? "The average of Y1 minus Y0 over all first-years" passes.

Two kinds of inquiry cover most of this course, and you tell them apart by reading
the formula.

- **Descriptive inquiry**: a summary of the world as it is, using outcomes you
  could in principle observe. Example: "the share of first-years who feel they
  belong," which is just an average of one column.
- **Causal inquiry**: a comparison of two potential-outcome worlds for the same
  units. Example: "the average of Y1 minus Y0," which subtracts a world that did
  not happen. Any formula with both Y1 and Y0 in it is causal, because it reaches
  for the counterfactual.

> **A question that often comes up here:** *"Why name the number before I have any
> data? Isn't that backwards?"* It is the opposite of backwards. If you pick the
> quantity after seeing results, you can quietly slide toward whichever number
> looks best, and no one can tell the design ever promised something else. Naming
> the estimand first is what lets a reader hold you to it.

## 5. Declare the Design, Then Diagnose It

**Guiding question:** *once your four MIDA parts are written down, how do you find
out whether they actually work before you spend a semester collecting data?*

> *"Do not bring me a design you have only argued for. Bring me one you have
> already run a thousand times in simulation, so you can tell me how often it lands
> on the truth and how often it embarrasses you. I trust a diagnosed design. I do
> not trust a confident one."*
> — a **skeptical methodologist** on your committee, before you collect any data

To **declare** a design is to write all four MIDA parts as something a computer can
run: a model that generates fake data, an inquiry it can compute on that fake data,
a data strategy that samples and assigns, and an answer strategy that produces an
estimate. Once declared, you can **diagnose** it: run the whole design many times
on simulated worlds and watch how its estimate behaves. Three numbers summarize
that behavior, and together they say what the design can promise.

- **Bias**: the average distance between your estimate and the truth, across many
  runs. A design is **unbiased** when its estimates center on the true answer.
  Example: if the true effect is 2 points and your estimates average 2.0, bias is
  about zero. Bias is a systematic tilt, and it does not shrink when you collect
  more data.
- **Variance**: how much the estimate wobbles from one run to the next, measured as
  the spread of the estimates. Example: estimates that scatter from -6 to +10 have
  high variance; ones that hug 1.8 to 2.2 have low variance. More data shrinks
  variance.
- **Power**: how often the design would actually detect the effect, meaning
  conclude "there is something here" rather than "nothing." Example: power of 0.9
  means nine runs in ten would find the real effect. Low power means a real effect
  usually slips through as a shrug.

Here is the payoff. Bias asks *are we aimed at the truth?* Variance asks *how
steady is our aim?* Power asks *will we actually hit the target often enough to
bother?* The next cells declare the mentoring design and diagnose all three.

### 🛠️ Run the Study: diagnose the declared design

Run the cell below. It declares the mentoring design, then simulates it 2000
times: each run builds a fresh world, assigns mentors by a coin flip, and computes
the difference in average belonging. It reports bias, variance, and power, and
plots every run's estimate. Read the spread before you read the next markdown cell.

```python
# Declare the mentoring design, then diagnose it by simulation.
def diagnose(n_per_arm, assignment="randomized", effect=EFFECT, reps=2000, seed=SEED):
    # Run a declared design `reps` times; return every run's estimate plus its
    # bias, variance (SD of the estimates), and power (share of runs that detect).
    rng_d = np.random.default_rng(seed)          # own stream, reproducible on its own
    n = n_per_arm * 2
    ests = np.empty(reps)
    detected = np.zeros(reps, dtype=bool)
    for r in range(reps):
        prior = rng_d.normal(0, 1, n)                   # prior campus involvement
        Y0 = 50 + 6 * prior + rng_d.normal(0, 8, n)     # belonging without a mentor
        Y1 = Y0 + effect                                # with a mentor (constant nudge)
        if assignment == "randomized":
            Z = np.zeros(n, dtype=int)
            Z[rng_d.permutation(n)[:n // 2]] = 1        # a coin flip: exactly half mentored
        else:                                           # confounded: involved -> mentor
            Z = (rng_d.random(n) < 1 / (1 + np.exp(-1.5 * prior))).astype(int)
        Y = np.where(Z == 1, Y1, Y0)                    # observe only the assigned world
        est = Y[Z == 1].mean() - Y[Z == 0].mean()       # answer strategy: difference in means
        se = np.sqrt(Y[Z == 1].var(ddof=1) / (Z == 1).sum()
                     + Y[Z == 0].var(ddof=1) / (Z == 0).sum())
        ests[r] = est
        detected[r] = abs(est) / se > 1.96              # would this run 'find' an effect?
    return ests, ests.mean() - effect, ests.std(ddof=1), detected.mean()

TRUE_ATE = EFFECT
ests_small, bias_s, var_s, power_s = diagnose(n_per_arm=10)
print("Declared design diagnosed over 2000 runs (10 first-years per group):")
print(f"  true effect (the answer key)       : {TRUE_ATE:.2f} belonging points")
print(f"  BIAS  (avg estimate minus truth)   : {bias_s:+.3f}    near 0 means on target")
print(f"  VARIANCE (spread of estimates, SD) : {var_s:.3f}    how much the answer wobbles")
print(f"  POWER (share of runs that detect)  : {power_s:.1%}    how often it finds the effect")

fig, ax = plt.subplots()
ax.hist(ests_small, bins=30, color="#4C72B0", edgecolor="white")
ax.axvline(TRUE_ATE, color="black", linestyle="--", linewidth=2, label=f"true effect = {TRUE_ATE:.0f}")
ax.axvline(0, color="gray", linestyle=":", linewidth=1.5, label="no effect (0)")
ax.set_xlabel("Estimated mentor effect, one value per run (belonging points)")
ax.set_ylabel("Number of runs")
ax.set_title("Diagnosis of the small design: centered on the truth, but a very wide spread")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Self-check: the small design is unbiased yet badly underpowered.
assert abs(bias_s) < 0.5, "a randomized design should be ~unbiased"
assert power_s < 0.15, "the small design should be badly underpowered"
assert 4.0 < var_s < 5.0, "estimate spread drifted — is the seed 464?"
print(f"✓ Self-check passed: bias is {bias_s:+.2f} (on target), yet the design catches the "
      f"real effect only {power_s:.0%} of the time.\n  Aimed at the truth, far too wide to "
      f"trust any single run.")
```

**Reading the diagnosis.** Look at what the three numbers say together. The
estimates center almost exactly on 2, so the design is **unbiased**: its aim is
true. Yet the spread is enormous, roughly plus or minus 9 points around the truth,
so any single run could easily land at -5 or +8. That wide spread is **variance**,
and it is why **power** comes out below one in ten. The estimate is so wobbly that
most runs cannot rule out zero, even though the effect is really there. The design
is honest and useless at the same time. It points at the truth on average and
almost never gets close enough on any one try to prove it.

*Figure note: the histogram shows 2000 estimates from the small design; the black
dashed line is the true effect and the gray dotted line is zero, and the pile
straddles both, which is exactly what low power looks like.*

## 6. Redesign: The One Change That Reads the Diagnosis

**Guiding question:** *given what the diagnosis just told you, which single change
buys the most, and which problem can no amount of data fix?*

**Redesign** means changing one part of MIDA in response to what the diagnosis
showed, then diagnosing again. The small design's problem was named for you: low
power driven by high variance. Variance shrinks with more data, so the obvious
redesign is more units. The next cell keeps everything else fixed and raises the
sample to 600 first-years per group.

It also runs a second, different design as a warning. Imagine the coin flip is
gone and mentors are taken up by the already-involved first-years, exactly the
confounding arrow from Lecture 1. This design also has 600 per group, so its
estimates are just as steady. Watch where they steady around.

> **A question that often comes up here:** *"If more data fixes the small design,
> why not just collect more data for everything?"* Because more data only cures
> variance, never bias. A confounded design with a huge sample gives you a very
> precise estimate of the wrong number. Power climbs, the spread shrinks, and
> every tight run agrees on an answer that is off. Size buys steadiness, not truth,
> and only a design change, here restoring random assignment, can move bias.

**Before you run, predict:** the small design caught the effect under one time in
ten. After raising the sample to 600 per group, what power would you expect, and
what will happen to the confounded design's center? Write a quick guess, then run
the cell below.

```python
# Redesign 1: more units (600 per group), everything else unchanged.
ests_big, bias_b, var_b, power_b = diagnose(n_per_arm=600)
# A different design: same big sample, but CONFOUNDED (involved first-years self-select in).
ests_conf, bias_c, var_c, power_c = diagnose(n_per_arm=600, assignment="confounded")

print("Redesign 1 -- more units (600 per group), still randomized:")
print(f"  bias {bias_b:+.3f}    variance(SD) {var_b:.3f}    power {power_b:.1%}")
print("\nA confounded design -- big sample, but mentors taken up by the already-involved:")
print(f"  bias {bias_c:+.3f}    variance(SD) {var_c:.3f}    power {power_c:.1%}")

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(ests_small, bins=30, color="#4C72B0", alpha=0.55, edgecolor="white",
        label="small randomized (10 per group)")
ax.hist(ests_big, bins=30, color="#55A868", alpha=0.85, edgecolor="white",
        label="big randomized (600 per group)")
ax.hist(ests_conf, bins=30, color="#DD8452", alpha=0.85, edgecolor="white",
        label="big but confounded (600 per group)")
ax.axvline(TRUE_ATE, color="black", linestyle="--", linewidth=2,
           label=f"true effect = {TRUE_ATE:.0f}")
ax.axvline(0, color="gray", linestyle=":", linewidth=1.3)
ax.set_xlabel("Estimated mentor effect (belonging points)")
ax.set_ylabel("Number of runs")
ax.set_title("Redesign reads the diagnosis: more units narrows the spread; confounding shifts the center")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Self-check: more units rescues power without adding bias; confounding does not.
assert abs(bias_b) < 0.3 and power_b > 0.85, "more units should rescue power with no new bias"
assert bias_c > 4.0 and power_c > 0.95, "the confounded design should be badly biased yet 'sure'"
print(f"✓ Self-check passed: more units lifted power from {power_s:.0%} to {power_b:.0%} and shrank "
      f"the spread from {var_s:.2f} to {var_b:.2f}, bias still near 0.\n  The confounded design is "
      f"off by {bias_c:+.1f} points and 'detects' it {power_c:.0%} of the time: confident and wrong.")
```

**Reading the redesign.** More units did exactly what the diagnosis predicted.
Power jumped from under one in ten to more than nine in ten, and the spread of
estimates collapsed toward the truth, all while bias stayed near zero. That is
variance being cured by sample size. The confounded design tells the cautionary
half. Its estimates are just as tight, but they cluster around roughly 8 instead
of the true 2, off by about six belonging points, and it "detects" that wrong
answer on essentially every run. It is confident and wrong. The single change that
rescues the first design is more units; the single change that would rescue the
second is not more data at all, but restoring the coin flip.

*Figure note: three piles of 2000 estimates each; the two randomized designs (blue
and green) sit over the true dashed line while the confounded pile (orange) sits
well to its right, narrow but centered on the wrong number.*

### 🧭 What Your Diagnosed Design Can Claim: Identify vs Associate

**Guiding question:** *when is it honest to say your treatment "causes" the
outcome, and when must you stop at "is associated with" it?*

The diagnosis just drew a bright line between the two randomized designs and the
confounded one. That line has a name, and your M3 charter is graded on drawing it.

- **Identify a cause**: your design pins the outcome gap on the treatment and rules
  out the other explanations, so you can defend the word *causes*. Example: the coin
  flip erased the prior-involvement arrow, so in the randomized design the
  mentored-minus-unmentored gap can be credited to the mentor.
- **Association**: two things move together, but your design cannot rule out a
  common cause, so you can only say they are *related*. Example: in the confounded
  design the mentored group scores higher, yet prior involvement lifted both who got
  a mentor and who felt they belonged, so that gap is an association, not a proven
  effect.
- **Silent scope change**: quietly upgrading your wording from *associated with* to
  *causes* without the design that earns it. Example: reporting "mentoring raises
  belonging" from the confounded study, whose honest claim stops at "mentoring is
  associated with higher belonging."

Here is the rule for your charter's claim boundary. A design that **identifies** the
effect, through random assignment or a strong argument that nothing else explains
the gap, licenses causal words. A design that only shows an **association** does
not, no matter how large the sample or how tight the spread. Naming which one you
hold, and refusing the silent scope change, is exactly what the two boundary
sentences of your M3 declaration have to do.

> **A question that often comes up here:** *"My project is descriptive, so isn't
> this beside the point?"* No. Every compass position draws a boundary. A
> description that only reaches the people in your data cannot silently speak for a
> whole population, and a prediction that fit last year cannot silently promise the
> next. *Identify* and *associate* are the causal version of a discipline every
> position needs: claim exactly what your design earns, and not one step more.

**association**, and calling that association a cause is the **silent scope change**
your claim boundary has to refuse.

> **"Rigor is a diagnosed property, not a vibe. A design earns a claim only when its
> declared parts align and its diagnosis says, in numbers, what the answer can and
> cannot support. A design you never diagnosed makes no promises, no matter how
> rigorous it sounds."**
