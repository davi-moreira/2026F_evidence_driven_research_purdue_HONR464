## 1. Why This Matters

> *"I can tell within one paragraph whether an author knows what question they
> are asking. If they don't, nothing downstream can save the paper. Not the
> data, not the methods, not the writing."*
> — a **skeptical peer** you will one day be grateful for

Everything in research is downstream of the question. A vague question makes
every later choice arbitrary: any data will do, any method will do, and any
answer will do, because nothing was pinned down enough to be wrong. A sharp
question does the opposite. It tells you what evidence you need, which compass
position fits, and what would count as being mistaken. That is why this
notebook comes before any tool does.

The path has four stops, and you will walk it with your own project:

**curiosity** *(I wonder about…)* → **topic** *(the studyable territory)* →
**problem** *(the specific unknown worth solving, anchored in a gap)* →
**question** *(the sentence evidence can answer)*.

Lecture 1 of this topic walks the first half of the path, from curiosity to
topic to gap. Lecture 2 finishes the job, from problem to answerable
question, and ends with your M01 pitch assembled.

> **A question that often comes up here:** *"My curiosity feels too big for
> this. I care about something huge, like inequality or mental health. Do I
> have to shrink it?"* You narrow the QUESTION, not the caring. The huge thing
> stays as your motivation; the introduction of every great study names one.
> The question is the part small enough to answer this semester. Answering it
> is how you earn the right to say something about the huge thing.

## 3. Gaps: What Is Actually Missing From What We Know?

> *"Reviewer 2 here. The authors claim no one has studied X. A ten-minute
> search found four papers that studied X. What the authors MEAN is that no
> one has studied X in this population with this measure. That would be
> interesting, and they never say it."*
> — a **journal reviewer**, doing you a favor the hard way

A research **problem** is not "my topic is interesting." It is this: *the
field does not yet know one specific thing, and that absence matters.* The
specific missing thing is the **gap**. Gaps come in recognizable kinds, and
learning the kinds turns "I guess nobody's done this?" into a claim you can
actually check and defend:

| Gap type | What is missing | Sounds like |
|---|---|---|
| **Evidence gap** | The question is basically unstudied | "No study has examined…" |
| **Population gap** | Studied, but not for these people or contexts | "…but not among first-generation undergraduates" |
| **Measurement gap** | Studied, but with measures that miss the concept | "…using self-reports only, never behavior" |
| **Method gap** | Studied, but with designs too weak for the claim made | "…only one-time-snapshot (cross-sectional) correlations, so nothing causal" |
| **Contradiction gap** | Studied, and the findings disagree | "…with results split roughly down the middle" |

One discipline above all: state your gap as a **bounded claim**, a claim with
edges you can verify. "Based on the sources I examined, X has not been
established for Y." Never write the absolute "nothing is known." The bounded
version is honest, checkable, and survives Reviewer 2. The absolute version
is one search away from embarrassment.

> **A question that often comes up here:** *"What if I find out my gap has
> already been filled?"* Then you just saved yourself a semester of
> rediscovering something. And you almost certainly found a NEW gap in the
> process, because every study that fills a gap opens others: a different
> population, a contradicting result, a measure you can improve on. Gaps
> don't run out; unexamined ones do.

## 4. The Answerable Question

**Guiding question:** *what turns a sharpened topic into a question evidence
can actually answer, and be wrong about?*

You have a topic and a gap. Together they define your research **problem**:
the specific unknown your project will tackle, and why that absence matters.
The final step of the funnel is the sentence that will govern your whole
semester: a question **evidence can answer**. Four marks distinguish it, and
together they form the answerability rubric:

1. **Units.** WHO or WHAT is the question about? People, teams, counties,
   posts, games.
2. **Outcome.** WHAT observable thing varies? A score, a rate, a count, a
   category. Something that can be measured, even imperfectly.
3. **Comparison.** Compared to WHAT? Between groups, over time, or against a
   benchmark. Even purely descriptive questions compare against "how much or
   how many."
4. **Scope.** WHERE and WHEN does the question live? A population, a place, a
   period. Edges you could defend.

And one boundary to respect: an **empirical** question asks what IS, which is
evidence territory. A **normative** question asks what OUGHT to be, which is
values territory. "Is remote work good?" is normative, and evidence cannot
answer it. But evidence can answer the empirical questions hiding inside it,
such as "does remote work reduce first-year mentoring interactions at large
firms?", and those answers can INFORM the values debate. Research earns its
keep by answering the inside questions honestly.

Watch one repair, end to end:

> ✗ *"Does social media hurt mental health?"* No units (whose?), a vague
> outcome ("hurt" how, measured how?), no comparison, no scope.
>
> ✓ *"Do U.S. college undergraduates who spend more than three hours daily on
> short-video apps report more anxiety symptoms (on the GAD-7, a standard
> seven-question anxiety screening) than lighter users, in their first
> year?"* Units ✓ outcome ✓ comparison ✓ scope ✓. Not yet perfect, but now it
> can be WRONG, which means it can be answered.
