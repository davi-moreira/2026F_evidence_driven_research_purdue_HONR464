## 1. Why This Matters

> *"An author once cited a paper of mine to support a claim my paper flatly
> contradicts. They never opened it. An AI summary told them what it said, and
> they believed the summary. That single unretrieved citation is why I now check
> every reference in a manuscript against the actual source."*
> — a **journal reviewer**, on the fastest way to lose their trust

Your gap from nb01 only counts if it is real. "Real" means checked against the
literature that actually exists, not the literature an AI assistant describes.
AI tools are extraordinary at finding leads. They are also unreliable narrators
about those leads. They sometimes invent authors, attach real names to papers
that were never written, and describe journals nobody can find. They do this
most confidently exactly where you know least. **In plain terms:** a citation
you cannot open is not a weak source. It is not a source at all.

So every claim in your project passes through one loop:

**Ask** (use any tool to surface leads) → **Retrieve** (open the actual source
yourself) → **Verify** (confirm it exists and says what was claimed) →
**Document** (log it in your ledger, with where you found it).

> **A question that often comes up here:** *"If I have to verify everything
> myself, what is the AI even for?"* For breadth and speed you could not manage
> alone. It reads the shape of a field in seconds and hands you names to chase,
> and that is genuinely valuable. What it cannot do is vouch for anything. Treat
> its output as a map drawn by someone who has never visited the territory:
> priceless for pointing you somewhere, worthless as proof you arrived.

## 2. Where Claims Live: the Source Hierarchy

**Guiding question:** *when two sources disagree, which one stands on stronger
evidence, and how do you find the real study behind a secondhand claim?*

Start with four terms you will use for the rest of the semester.

- **Peer review** is the check a study passes before a journal publishes it:
  independent experts read the manuscript, demand fixes, and can reject it. It
  filters out a lot of sloppiness, though not all error.
- A **venue** is the outlet where a work appears: a specific journal, a
  university press, a conference. Real venues can be looked up. *Science*, the
  journal that published Tversky & Kahneman (1974), has a searchable archive
  going back decades.
- A **primary source** is the original study that reports the evidence
  firsthand. A **secondary source** talks *about* that study: a textbook, a news
  article, a review. A news story on a vaccine trial is secondary; the trial
  report it summarizes is primary.

Not all sources carry the same weight. When two disagree, the question is never
"which sounds better" but "which stands on stronger evidence." A rough
hierarchy, strongest backing first:

- **Peer-reviewed empirical studies.** A claim tied to data that independent
  experts reviewed before publication.
- **Systematic reviews and meta-analyses.** Many studies pooled into one
  answer; strong *when the pooled studies were sound*.
- **Reports and preprints** (papers shared before peer review). Often
  excellent, but nobody has checked them yet, so read the methods yourself.
- **Popular writing**: books, journalism, explainers. Great for leads and
  framing. Chase the primary source underneath before you cite the claim.

The move that separates a researcher from a repeater is **snowballing**: once
you have one solid source, walk its reference list backward to see what it
built on, and its citations forward to see who built on it. Two good sources,
snowballed, become ten.

> **A question that often comes up here:** *"A blog post states a striking
> statistic with no citation. Can I use it?"* Not as evidence, yet. You can use
> it as a lead: search the exact statistic, find the study it came from, open
> that study, and cite *that*, if it says what the blog claimed. If the trail
> dead-ends, the statistic dead-ends with it. No trail, no claim.

## 3. From Ledger to Claim Map: Drawing What a Field Argues

> *"Show me the argument of your literature, not a pile of summaries. Which
> sources agree, which fight, and where exactly is the hole you say you're
> filling? If you can't draw it, you haven't read it yet."*
> — a **thesis advisor**, sending a first draft back

A reading list is not a literature review. A pile of sources tells your reader
nothing about the *shape* of what is known. A **claim map** does. It is a
diagram of a literature: each verified source becomes one claim, lines connect
claims that agree, contradict, or qualify one another, and one marked spot shows
where no source reaches. That spot is your gap. Three moves build it:

1. **Extract the central claim.** One sentence per source: what does it
   actually assert? Not its topic; its claim.
2. **Type the evidence.** Which compass position backs the claim: description,
   generalization, prediction, or causal? A causal claim resting on descriptive
   evidence is a crack in the map worth marking.
3. **Draw the edges, place the gap.** Connect claims that **agree**,
   **contradict**, or **qualify** one another. Then put your gap where no
   source reaches, and check it is a hole rather than somewhere you have not
   read.

One honesty rule governs the whole picture. A map of claims is **not proof**.
It shows what the field *asserts*, not what is *true*. And an unread source is
not on the map at all.

> **A question that often comes up here:** *"How many sources do I need before
> I can honestly draw a map and place a gap?"* Fewer than you fear, but every
> one retrieved. A map of four or five well-chosen, verified sources that
> genuinely agree and disagree beats twenty citations skimmed from an AI
> summary. The map's honesty comes from the retrieval behind each node, not
> from the node count. Start with what you have verified, and let snowballing
> grow it.

### 🛠️ Hands-On: build a claim map

The cell below builds a claim map from two small tables: one of claims, one of
relationships. The literature here is a compact worked example, four simplified
claims of the kind a real social-media-and-trust literature contains. Run it and
read the picture. Then notice the design: you build your *own* map later by
swapping your ledger into the two tables. The graph draws itself.

> 💡 **Gemini Prompt:** "Here is a Python cell that builds a claim map with
> networkx from two tables, one of claims and one of relationships: [paste the
> next cell]. Explain how the `edges` table decides which nodes get connected,
> and tell me what I would change to add a fifth source that agrees with claim
> C1."
>
> **After running, verify (the responsible-AI habit):**
> - [ ] Check Gemini's account of the edges against the graph the cell actually
>   draws. Do the arrows it describes match the ones you see on screen?
> - [ ] Try its suggested edit on your own copy and re-run. The redrawn graph,
>   not Gemini's description of it, is the proof the new edge landed where you
>   meant.
> - [ ] Log this use in your AI ledger: tool, task, verification.

```python
# A worked claim map. To build your OWN, replace the two tables with your ledger.
import networkx as nx
from matplotlib.lines import Line2D

claims = pd.DataFrame({
    "id":       ["C1", "C2", "C3", "C4", "GAP"],
    "kind":     ["source", "source", "source", "source", "gap"],
    "evidence": ["generalization", "generalization", "description", "generalization", "—"],
    "label":    ["C1: heavy use links\nto lower trust",
                 "C2: the link is\nsmall and mixed",
                 "C3: trust fell\nbefore social media",
                 "C4: strong local ties\nbuffer the link",
                 "GAP: not shown for\nfirst-gen undergrads"],
})
edges = pd.DataFrame({
    "src":      ["C2", "C3", "C4", "C4"],
    "dst":      ["C1", "C1", "C1", "GAP"],
    "relation": ["contradicts", "contradicts", "qualifies", "opens"],
})

print("Claims, each typed by the kind of evidence behind it:")
readable = claims.assign(label=claims["label"].str.replace("\n", " ", regex=False))
print(readable[["id", "evidence", "label"]].to_string(index=False))

G = nx.DiGraph()
for _, r in claims.iterrows():
    G.add_node(r["id"], kind=r["kind"])
for _, e in edges.iterrows():
    G.add_edge(e["src"], e["dst"], relation=e["relation"])

pos = nx.spring_layout(G, seed=SEED, k=1.6)
ecolor = {"agrees": "#2a9d8f", "contradicts": "#e76f51",
          "qualifies": "#8a8a8a", "opens": "#6a4c93"}
node_colors = ["#f4a259" if G.nodes[n]["kind"] == "gap" else "#8ec7d2"
               for n in G.nodes]
edge_colors = [ecolor[G.edges[e]["relation"]] for e in G.edges]
labels = dict(zip(claims["id"], claims["label"]))

fig, ax = plt.subplots(figsize=(9, 6))
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=4200, ax=ax)
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2.2,
                       arrowsize=18, node_size=4200, ax=ax)
nx.draw_networkx_labels(G, pos, labels, font_size=8, ax=ax)
ax.set_title("A claim map: what the sources assert, how they relate, where the gap sits")
ax.axis("off")
handles = [Line2D([0], [0], color=c, lw=3, label=r) for r, c in ecolor.items()]
handles.append(Line2D([0], [0], marker="o", color="w", label="gap (unfilled)",
                      markerfacecolor="#f4a259", markersize=12))
ax.legend(handles=handles, loc="lower left", fontsize=8, frameon=False)
plt.tight_layout()
plt.show()
```

```python
# Self-check — a claim map needs exactly one placed gap and its relations intact.
assert G.number_of_nodes() == 5 and G.number_of_edges() == 4, "map shape changed unexpectedly"
assert (claims["kind"] == "gap").sum() == 1, "a claim map needs exactly one gap node placed"
assert set(edges["relation"]) <= set(ecolor), "an edge relation has no color — add it to the legend"
print("✓ Self-check passed: 5 claims, 4 relations, exactly one gap placed.")
```
