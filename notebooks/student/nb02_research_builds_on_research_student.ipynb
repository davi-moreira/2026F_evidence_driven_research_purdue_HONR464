## 1. Research Is a Cumulative Conversation

**Guiding question:** *how does a field actually accumulate what it knows, and where
does a genuinely new question fit into that conversation?*

> *"Show me the conversation your work joins, not a pile of summaries. Which studies
> agree, which fight, and where exactly is the hole you say you are filling? If you
> cannot place your question in that conversation, you have not read it yet."*
> — a **thesis advisor**, sending a first draft back

A field is a **cumulative conversation**: a running argument where each study
answers, extends, or challenges the ones before it. **In plain terms:** nobody
starts from zero. Your job is not to invent an island. It is to find the
conversation your question belongs to and add one honest sentence to it.

Five moves add to that conversation. Learn the names, because your M2 map will sort
your sources by which move each one makes.

- **Replication**: running the same study again to see if the result holds. Example:
  a second lab repeats a memory experiment with new people and checks whether the
  effect shows up again.
- **Extension**: taking a known result somewhere new. Example: a finding about voters
  in one city is tested on voters in another.
- **Synthesis**: pulling many studies into one picture. Example: a review pools
  twenty small trials to report what they say together.
- **Comparison**: setting two findings side by side to see where they agree or
  clash. Example: two studies measure the same thing and report different sizes.
- **Boundary conditions**: mapping where a result holds and where it stops. Example:
  a training effect that works for beginners but fades for experts.

> **A question that often comes up here:** *"A blog post states a striking statistic
> with no citation. Can I use it?"* Not as evidence yet. You can use it as a lead:
> search the exact statistic, find the study it came from, open that study, and cite
> *that*, if it says what the blog claimed. If the trail dead-ends, the statistic
> dead-ends with it. No trail, no claim.

---

Every source you keep passes through one loop. Learn its four steps under one
name, because it is the habit the whole course is built on.

- **Lead**: a candidate source you have not verified yet. Example: a citation Gemini
  hands you, or a title in another paper's reference list.
- **Retrieval-verification loop**: **Ask** (use any tool to surface leads),
  **Retrieve** (open the actual source yourself), **Verify** (confirm it exists and
  says what was claimed), **Document** (log it, with where you found it). A lead
  stays a lead until retrieval and verification promote it to a source.

Two named failures are what the loop exists to catch. You will meet both again in
the AI Error Taxonomy.

- **Citation hallucination**: an AI inventing a source that does not exist, presented
  with full confidence. Example: a real-sounding author pair, a recent year, and a
  real-looking journal, for a paper nobody ever wrote. This is the taxonomy's
  *confident fabrication*.
- **Mischaracterized source**: a real source cited for a claim it does not actually
  make. Example: a paper that reports a correlation, cited as if it proved a cause.
- **Claim-evidence-source alignment**: the check that the claim you attribute to a
  source is the claim the source actually makes, on the evidence it actually has. A
  source can be real and still be mischaracterized, so this check is separate from
  "does it exist."

> **A question that often comes up here:** *"The AI handed me a perfect-sounding
> study I cannot find anywhere. What are the possibilities?"* Three. It **fabricated**
> the study whole, and there is nothing to find. It **garbled** a real source, right
> idea but wrong author, year, or journal, so search the claim and the likely real
> authors instead of the mangled title. Or it is real but **paywalled or obscure**,
> reachable through the library portal or by emailing the author. Your next step is
> the same in all three: the lead stays out of your verified column until retrieval
> succeeds.

## 2. Citation Chains: How Two Sources Become Ten

**Guiding question:** *once you have one solid source, how do you let it lead you to
the rest of the conversation, and how do you know when you have looked hard enough?*

The move that separates a researcher from a repeater is **snowballing**: once you
hold one verified source, you walk its links in two directions to find more.

- **Backward citation search**: read the source's own **reference list** to find the
  older work it built on. Example: a 2020 paper's bibliography points you to the
  1990s study that started the idea.
- **Forward citation search**: find the newer works that **cite** the source. Example:
  Google Scholar's "Cited by" link shows who has used the paper since. Two good
  sources, snowballed both ways, become ten.

This matters for the puzzle you opened with. A **genuine gap** is a question your
verified sources actually leave unanswered, confirmed by looking. An **unsuccessful
search** is a question you *stopped* looking for. They feel identical from your desk,
and only one is a real contribution. The test is retrieval: you have earned the word
"gap" only after real sources, followed forward and backward, still stay silent on it.

> **A question that often comes up here:** *"How many sources do I need before I can
> honestly say I found a gap?"* Fewer than you fear, but every one retrieved. Four or
> five well-chosen, verified sources that genuinely agree and disagree beat twenty
> citations skimmed from an AI summary. The honesty comes from the retrieval behind
> each source, not from the count. Start with what you verified, and let snowballing
> grow it.

```python
# Draw a citation chain. An arrow A -> B reads "A cites B".
# (networkx is imported once in the setup cell.)
C = nx.DiGraph()
pos = {
    "Older ref\nR1":     (0.0, 1.4),
    "Older ref\nR2":     (0.0, 0.6),
    "Seed source\nS":    (2.0, 1.0),
    "Citing work\nF1":   (4.0, 1.4),
    "Citing work\nF2":   (4.0, 0.6),
}
edges = [
    ("Seed source\nS", "Older ref\nR1"),    # backward: S's own reference list
    ("Seed source\nS", "Older ref\nR2"),
    ("Citing work\nF1", "Seed source\nS"),   # forward: newer work that cites S
    ("Citing work\nF2", "Seed source\nS"),
]
C.add_edges_from(edges)

fig, ax = plt.subplots(figsize=(10, 4.4))
nx.draw_networkx_nodes(C, pos, node_color="#DCE6F1", edgecolors="#4C72B0",
                       linewidths=1.5, node_size=5000, ax=ax)
nx.draw_networkx_labels(C, pos, font_size=9, ax=ax)
nx.draw_networkx_edges(C, pos, arrowstyle="-|>", arrowsize=18, edge_color="#555555",
                       width=1.6, node_size=5000, ax=ax)
ax.text(1.0, 2.05, "backward search:\nfollow the reference list",
        fontsize=8, color="#333333", ha="center")
ax.text(3.0, 2.05, "forward search:\nfind who cites it",
        fontsize=8, color="#333333", ha="center")
ax.set_ylim(0.2, 2.3)
ax.set_title("A citation chain: an arrow A -> B means 'A cites B'")
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ Citation chain drawn.")
print("  Backward search = follow S's own reference list to older work it built on.")
print("  Forward search  = find newer work that cites S.")
print("  Two verified seed sources, snowballed both ways, become ten.")
```

*In plain terms, the picture says every verified source is a hub. Its reference list
points backward to what it built on, and its citing works point forward to what came
after. You grow a real evidence base by walking those arrows, not by trusting a list
you never opened.*

```python
# Two engines that grow an evidence base. Watch which one you can trust.
rng_sim = np.random.default_rng(SEED)  # reproducible on its own (seed 464)

# Engine 1 — an AI channel proposes a big, tidy-looking pile of candidate citations.
n_candidates = 15
# Each candidate faces the retrieval-verification loop. Two checks kill leads:
exists   = rng_sim.random(n_candidates) < 0.60   # check 1: retrievable at all
supports = rng_sim.random(n_candidates) < 0.75   # check 2: actually supports the claim
retrievable = int(exists.sum())
verified_from_ai = int((exists & supports).sum())  # a source only if it clears BOTH

print("Engine 1 — the AI candidate pile through the retrieval-verification loop:")
print(f"  candidates proposed        : {n_candidates}")
print(f"  cleared check 1 (retrieved): {retrievable}")
print(f"  verified (retrieved + supports): {verified_from_ai}")
print(f"  -> a tidy pile of {n_candidates} collapsed to {verified_from_ai} verified sources")
print()

# Engine 2 — citation chains snowball from the sources you already TRUST.
# Each verified source exposes real references (backward) and citing works (forward).
base = verified_from_ai
trajectory = [base]
for _ in range(3):
    new = int(rng_sim.binomial(base, 0.6))  # a modest, real branching factor
    base += new
    trajectory.append(base)

print("Engine 2 — citation-chain snowball from the verified seeds:")
print(f"  verified base per round : {trajectory}")
print(f"  -> {trajectory[0]} trusted seeds grew to {trajectory[-1]} retrievable works")

fig, axes = plt.subplots(1, 2, figsize=(11, 4.2))
stages = ["candidates\nproposed", "cleared\ncheck 1", "verified\nsources"]
vals = [n_candidates, retrievable, verified_from_ai]
bars = axes[0].bar(stages, vals, color=["#B0B0B0", "#4C72B0", "#55A868"],
                   edgecolor="white")
for b, v in zip(bars, vals):
    axes[0].text(b.get_x() + b.get_width() / 2, v + 0.2, str(v),
                 ha="center", fontsize=11)
axes[0].set_ylabel("Number of citations")
axes[0].set_title("Engine 1: the AI pile collapses under verification")
axes[0].set_ylim(0, n_candidates + 2)

axes[1].plot(range(len(trajectory)), trajectory, marker="o", color="#6A4C93",
             linewidth=2)
for i, v in enumerate(trajectory):
    axes[1].text(i, v + 0.4, str(v), ha="center", fontsize=10)
axes[1].set_xlabel("Snowball round")
axes[1].set_ylabel("Verified sources in base")
axes[1].set_title("Engine 2: chains from real seeds grow the base")
axes[1].set_xticks(range(len(trajectory)))
plt.tight_layout()
plt.show()
```

```python
# Self-check: the loop drops leads at each check; chains grow the trusted base.
assert n_candidates == 15, "candidate count changed unexpectedly"
assert verified_from_ai < retrievable <= n_candidates, "each check should drop some leads"
assert trajectory[-1] > trajectory[0], "snowballing should grow the verified base"
print(f"✓ Self-check passed: {n_candidates} AI candidates collapsed to "
      f"{verified_from_ai} verified;\n  citation chains grew {trajectory[0]} seeds to "
      f"{trajectory[-1]} retrievable works across {len(trajectory) - 1} rounds.")
```

## 3. From Verified Sources to an Evidence Map

**Guiding question:** *how do you turn a stack of verified sources into a picture of
what the field argues, with your gap placed honestly where no source reaches?*

> *"Show me the argument of your literature, not a reading list. A pile of summaries
> tells me nothing. Which sources agree, which fight, and where is the hole you say
> you are filling? If you cannot draw it, you have not read it yet."*
> — a **journal reviewer**, on the first draft they will ever accept

A reading list is not a literature review. An **evidence map** is a diagram of a
literature: each verified source becomes one **claim** node, lines connect claims
that agree or contradict or qualify one another, and one marked spot shows where no
source reaches. That spot is your gap. Three terms carry the map, and your M2 rubric
grades all three.

- **Claim**: what a source actually asserts, in one sentence. Not its topic, its
  claim. Example: "heavy social-media use is linked to lower trust" is a claim; "social
  media and trust" is only a topic.
- **Agreement and contradiction**: how two claims relate. They **agree** when they
  point the same way, **contradict** when they clash, and **qualify** when one narrows
  the other. Example: a study finding no link *contradicts* one finding a strong link.
- **Contribution**: the honest sentence your map earns, naming where the field is
  silent and what your project would add. Example: "none of my verified sources looks
  at first-year commuters, which is my gap."

One honesty rule governs the whole picture. A map of claims is **not proof**. It shows
what the field *asserts*, not what is *true*, and an unread source is not on the map
at all.

> **A question that often comes up here:** *"Two of my verified sources flatly
> contradict each other. Does one have to be wrong?"* Usually not. Contradiction is
> data, not a verdict. Ask what would make them *both* right: different populations,
> measures, time windows, or methods. That question is where new gaps are born, so
> log the contradiction on your map instead of hiding it.

### 🛠️ Hands-On: build an evidence map

The cell below builds an evidence map from two small tables: one of claims, one of
relationships. The literature here is a compact worked example, four simplified claims
of the kind a real social-media-and-trust literature contains. Run it and read the
picture. To build your *own* map later, swap your verified sources into the two
tables. The graph draws itself.

```python
# A worked evidence map. To build your OWN, replace the two tables with your sources.
from matplotlib.lines import Line2D

claims = pd.DataFrame({
    "id":       ["C1", "C2", "C3", "C4", "GAP"],
    "kind":     ["source", "source", "source", "source", "gap"],
    "evidence": ["generalization", "generalization", "description", "generalization", "-"],
    "label":    ["C1: heavy use links\nto lower trust",
                 "C2: the link is\nsmall and mixed",
                 "C3: trust fell\nbefore social media",
                 "C4: strong local ties\nbuffer the link",
                 "GAP: not shown for\nfirst-year commuters"],
})
edges = pd.DataFrame({
    "src":      ["C2", "C3", "C2", "C4", "C4"],
    "dst":      ["C1", "C1", "C3", "C1", "GAP"],
    "relation": ["contradicts", "contradicts", "agrees", "qualifies", "opens"],
})

print("Claims, each typed by the kind of evidence behind it:")
readable = claims.assign(label=claims["label"].str.replace("\n", " ", regex=False))
print(readable[["id", "evidence", "label"]].to_string(index=False))

G = nx.DiGraph()
for _, r in claims.iterrows():
    G.add_node(r["id"], kind=r["kind"])
for _, e in edges.iterrows():
    G.add_edge(e["src"], e["dst"], relation=e["relation"])

pos = nx.spring_layout(G, seed=SEED, k=1.7)
# Each relation gets BOTH a color and a line style, so the map never relies on color alone.
style = {"agrees":      ("#2A9D8F", "solid"),
         "contradicts": ("#E76F51", "dashed"),
         "qualifies":   ("#8A8A8A", "dotted"),
         "opens":       ("#6A4C93", "dashdot")}
source_nodes = [n for n in G.nodes if G.nodes[n]["kind"] == "source"]
gap_nodes    = [n for n in G.nodes if G.nodes[n]["kind"] == "gap"]
labels = dict(zip(claims["id"], claims["label"]))

fig, ax = plt.subplots(figsize=(9, 6))
# Sources are circles; the gap is a square, so shape (not just color) marks it.
nx.draw_networkx_nodes(G, pos, nodelist=source_nodes, node_shape="o",
                       node_color="#8EC7D2", node_size=4200, ax=ax)
nx.draw_networkx_nodes(G, pos, nodelist=gap_nodes, node_shape="s",
                       node_color="#F4A259", node_size=4600, ax=ax)
for rel, (color, ls) in style.items():
    rel_edges = [(u, v) for u, v in G.edges if G.edges[u, v]["relation"] == rel]
    nx.draw_networkx_edges(G, pos, edgelist=rel_edges, edge_color=color, style=ls,
                           width=2.2, arrowsize=18, node_size=4200, ax=ax)
nx.draw_networkx_labels(G, pos, labels, font_size=8, ax=ax)
ax.set_title("An evidence map: what the sources assert, how they relate, where the gap sits")
ax.axis("off")
handles = [Line2D([0], [0], color=c, lw=2.5, ls=ls, label=rel)
           for rel, (c, ls) in style.items()]
handles += [Line2D([0], [0], marker="o", color="w", label="verified source",
                   markerfacecolor="#8EC7D2", markersize=12),
            Line2D([0], [0], marker="s", color="w", label="gap (unfilled)",
                   markerfacecolor="#F4A259", markersize=12)]
ax.legend(handles=handles, loc="lower left", fontsize=8, frameon=False)
plt.tight_layout()
plt.show()
```

```python
# Self-check: an evidence map needs exactly one placed gap and its relations intact.
assert G.number_of_nodes() == 5 and G.number_of_edges() == 5, "map shape changed unexpectedly"
assert (claims["kind"] == "gap").sum() == 1, "an evidence map needs exactly one gap node placed"
assert set(edges["relation"]) <= set(style), "an edge relation has no style — add it to the legend"
print("✓ Self-check passed: 5 claims, 5 relations, exactly one gap placed.")
print("  Contradictions (C2, C3 vs C1) are logged, not hidden; C4 qualifies C1 and opens the gap.")
```

*In plain terms, the picture says the field does not agree with itself: two sources
push against the "heavy use lowers trust" claim, one qualifies it, and none reaches the
first-year commuters your project is about. That unreached square is the gap, and it is
honest only because every circle around it is a source you retrieved.*

Read the map as an argument, not a scoreboard. The contradictions between C1, C2, and
C3 are the most useful thing on it: they tell you the question is live, not settled,
