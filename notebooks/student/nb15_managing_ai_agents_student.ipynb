## 1. Decompose the Chapter Into Scoped Roles

**Guiding question:** *how do you split one big writing task into jobs small enough
that a single AI role can own each one and be checked on it?*

> *"Bring me a chapter, not a pile of AI output. I want to know which part each tool
> touched, what you told it to do, and how you checked it. If you cannot tell me
> that, you did not write the chapter. The tools did, and neither of us can defend
> it."*
> — your **thesis advisor**, on the draft you will defend this week

The move that makes a team of roles work is **task decomposition**: breaking one job
into smaller subtasks, each with a clear owner. Example: instead of "improve my
chapter," you split it into check the design, check the citations, tighten the prose.
Each piece is now checkable on its own.

Three ideas make decomposition pay off, and each is a term your M15 portfolio uses.

- **Scoped task**: a job narrow enough to state in one sentence and check in one
  pass. Example: "flag every claim in section 2 that reaches past my sample."
  A scoped task has a clear pass or fail. "Make it better" does not.
- **Agent specialization**: giving one role one lens instead of asking a generalist
  for everything. Example: a skeptic briefed only to attack your weakest claim
  finds sharper holes than a role told to "review the whole thing." A narrow brief
  earns a sharper answer.
- **Context allocation**: deciding how much of your material each role actually
  sees. Example: your citation-checker needs the reference list, not your
  half-finished discussion section. More context is not always better. Extra
  material can bury the one thing you wanted checked.

> **A question that often comes up here:** *"If specializing a role makes it sharper,
> why not split my chapter into twenty tiny roles?"* Because every split costs you a
> hand-off you have to manage and verify. Past a point, you spend more effort wiring
> and checking roles than the sharper answers are worth. Decompose until each subtask
> is checkable, then stop. The goal is a workflow you can supervise, not the most
> roles possible.

Two more terms name what turns a pile of roles into a workflow you can trust.

- **Independent review**: a second role checks the first role's work without sharing
  its assumptions. Example: the role that drafts your summary should not also be the
  role that decides the summary is accurate. The checker has to be able to disagree.
- **Worker-critic structure**: pairing a role that produces (the worker) with a
  role that attacks what it produced (the critic). Example: a worker drafts your
  limitations paragraph, then a critic hunts for the limitation it left out. The
  critic's whole job is to find what the worker missed.

The worker-critic pair is the backbone of this week. A worker alone gives you a
draft that is confident about its own quality, which is exactly the blind spot you
cannot see from the inside. A critic that never wrote the draft can. But the critic
only helps if it is genuinely independent, and Lecture 2 shows what happens when it
is not.

Now wire the roles into an order.

## 2. Wire the Workflow: Parallel, Sequential, and the Critical Path

**Guiding question:** *which roles can safely work at the same time, and which must
wait for another role's output before they can start?*

Two words decide the wiring, and getting them wrong either wastes effort or lets one
early error poison everything downstream.

- **Parallel**: two roles run at the same time because neither needs the other's
  output. Example: a reviewer checking clarity and a citation-checker checking
  sources can both read the same draft at once. They are independent lenses on the
  same input.
- **Sequential**: one role must wait for another to finish, because it needs that
  output as its input. Example: a critic cannot attack a draft that the worker has
  not written yet. This is a **dependency**, a required order between two subtasks.

The **critical path** is the longest chain of dependencies in your workflow: the
roles that must happen one after another, with no shortcut. Example: worker drafts,
then critic attacks, then you integrate. That chain sets how many rounds your
workflow takes at minimum, because nothing can collapse a step that truly depends on
the one before it.

You will build a small worker-critic workflow as a **directed graph**, a picture of
arrows where each arrow means "this feeds that." Then you will read its critical path
off the code, not off a guess.

```python
# Build the worker-critic workflow as a directed graph.
# Nodes: five AI roles (worker, three critics, editor) plus YOU, the integrator.
# (networkx is imported once in the setup cell.)
G = nx.DiGraph()
parallel_edges = [
    ("Worker\ndraft", "Reviewer"),
    ("Worker\ndraft", "Methodologist"),
    ("Worker\ndraft", "Skeptic"),
    ("Reviewer", "You\nintegrate"),
    ("Methodologist", "You\nintegrate"),
    ("Skeptic", "You\nintegrate"),
    ("You\nintegrate", "Editor"),
]
G.add_edges_from(parallel_edges)

generations = list(nx.topological_generations(G))
depth = nx.dag_longest_path_length(G)          # longest dependency chain, in arrows
width = max(len(layer) for layer in generations)  # most roles that can run at once

print("Worker-critic workflow (three critics in PARALLEL):")
print(f"  roles + you (nodes)     : {G.number_of_nodes()}")
print(f"  dependencies (arrows)   : {G.number_of_edges()}")
print(f"  critical-path depth     : {depth} arrows  ({len(generations)} rounds)")
print(f"  maximum parallel width  : {width} roles at once")
print(f"  the critical path       : {' -> '.join(p.replace(chr(10), ' ') for p in nx.dag_longest_path(G))}")

# The SEQUENTIAL variant: the skeptic now waits for the reviewer's notes.
H = nx.DiGraph()
H.add_edges_from([
    ("Worker\ndraft", "Reviewer"),
    ("Worker\ndraft", "Methodologist"),
    ("Reviewer", "Skeptic"),               # skeptic now depends on the reviewer
    ("Skeptic", "You\nintegrate"),
    ("Methodologist", "You\nintegrate"),
    ("You\nintegrate", "Editor"),
])
gens_seq = list(nx.topological_generations(H))
print("\nIf the skeptic instead waits for the reviewer (one critic made SEQUENTIAL):")
print(f"  critical-path depth     : {nx.dag_longest_path_length(H)} arrows  ({len(gens_seq)} rounds)")
print(f"  maximum parallel width  : {max(len(l) for l in gens_seq)} roles at once")

# Draw the parallel workflow.
pos = {
    "Worker\ndraft":    (0.0, 1.0),
    "Reviewer":          (1.6, 1.9),
    "Methodologist":     (1.6, 1.0),
    "Skeptic":           (1.6, 0.1),
    "You\nintegrate":   (3.2, 1.0),
    "Editor":            (4.8, 1.0),
}
colors = {
    "Worker\ndraft": "#4C72B0", "Reviewer": "#8AA9D6", "Methodologist": "#8AA9D6",
    "Skeptic": "#8AA9D6", "You\nintegrate": "#55A868", "Editor": "#4C72B0",
}
fig, ax = plt.subplots(figsize=(10, 4.5))
nx.draw_networkx_nodes(G, pos, node_color=[colors[n] for n in G.nodes()],
                       edgecolors="#333333", linewidths=1.5, node_size=4600, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=8.5, ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=16, edge_color="#777777",
                       width=1.5, node_size=4600, ax=ax)
ax.set_title("A worker-critic workflow: three critics in parallel, then you integrate")
ax.axis("off")
plt.tight_layout()
plt.show()

print("\n✓ Workflow built. The green node is YOU: the integrator no role can replace.")
```

```python
# Self-check: the parallel workflow has depth 3 and width 3; sequencing a critic costs a round.
assert G.number_of_nodes() == 6, "expected 5 AI roles + you"
assert depth == 3, "parallel critical-path depth should be 3 arrows"
assert width == 3, "three critics should run in parallel (width 3)"
assert nx.dag_longest_path_length(H) == 4, "sequencing the skeptic should deepen the path to 4"
print(f"✓ Self-check passed: parallel depth {depth} (width {width}); the sequential "
      f"variant deepens to {nx.dag_longest_path_length(H)} arrows.\n"
      f"  Parallel buys you speed and three independent lenses at once; sequencing buys "
      f"the skeptic the reviewer's context, at the cost of one more round.")
```

## 3. Conflicting Outputs: Real Disagreement, Correlated Error, False Consensus

**Guiding question:** *when two AI roles disagree, and when they all agree, what are
you actually looking at?*

> *"Do not tell me your reviewers agreed. Tell me whether they could have agreed for
> the same wrong reason. Agreement is the easiest thing in the world to manufacture,
> and the most dangerous to trust."*
> — a **skeptical peer** at your practice defense

**Reconciling conflicting outputs** is the work of deciding what to believe when your
roles do not match. The first move is not to pick a winner. It is to diagnose *why*
they differ, because there are three very different cases and they call for opposite
responses.

- **Real disagreement**: the roles genuinely read the evidence differently. Example:
  your reviewer calls a sentence "clear and strong" while your methodologist flags
  that same sentence as reaching past your data. This is *useful*. It points you
  straight at the claim you need to check.
- **Correlated error**: the roles *agree*, but on the same wrong assumption, so the
  agreement proves nothing. Example: three reviewers built from the same base model,
  given the same prompt, all miss the same flaw because they share the blind spot.
  Ask: could they be wrong the *same* way? If yes, their consensus is one signal, not
  three.
- **False consensus**: the roles agree only because your prompt framed the task so
  they had to. Example: you ask "confirm my chapter is ready," and every role obliges.
  You led all the witnesses. Reframe the ask neutrally and run it again.

> **A question that often comes up here:** *"Conflict feels like the problem. Isn't
> agreement what I want?"* It is the reverse. Conflict announces itself and tells you
> where to look. Correlated error and false consensus feel like confirmation while
> quietly agreeing on something false. The trap to fear most is agreement, not
> conflict. When roles agree on something that matters, your job is to ask whether
> they were ever independent before you trust it.

## 4. Does Agreement Mean Confirmation? A Seeded Test

**Guiding question:** *if you add more AI reviewers, does your chance of catching a
real flaw actually improve, or does it stall?*

Here is a clean way to see the trap. Imagine every chapter you send out has exactly
one hidden flaw. You run a team of reviewers, and the team "misses" whenever *every*
reviewer says the chapter looks fine. That unanimous-but-wrong verdict is a **false
consensus**: the team agreed, and the team was wrong.

You will compare two teams of the same size. In the **independent** team, each
reviewer catches the flaw on its own, uncorrelated with the others. In the
**correlated** team, the reviewers share a blind spot: some of the time, the flaw
falls in a region they are *all* blind to, so every one of them misses it together.
Same number of reviewers. The only difference is whether their errors are
independent.

```python
# Two four-reviewer teams on 5,000 flawed chapters. Fresh seed so this is
# reproducible on its own, regardless of what ran before (seed 464).
rng_sim = np.random.default_rng(SEED)

n_chapters = 5000
k = 4        # reviewers per team
detect = 0.5 # each reviewer catches the hidden flaw about half the time, on its own
rho = 0.40   # the correlated team shares a blind spot on ~40% of chapters

# Independent team: each reviewer catches the flaw on its own. A MISS = all four fail.
caught_indep = rng_sim.random((n_chapters, k)) < detect
miss_indep = ~caught_indep.any(axis=1)
rate_indep = miss_indep.mean()

# Correlated team: on a shared-blind-spot chapter, EVERY reviewer misses together;
# otherwise each catches on its own, just like the independent team.
blind = rng_sim.random(n_chapters) < rho
caught_corr = (rng_sim.random((n_chapters, k)) < detect) & (~blind[:, None])
miss_corr = ~caught_corr.any(axis=1)
rate_corr = miss_corr.mean()

print("False consensus = all four reviewers say 'fine' while a real flaw is present.")
print(f"  Independent team miss rate : {rate_indep:.1%}   ({miss_indep.sum()} of {n_chapters})")
print(f"  Correlated team miss rate  : {rate_corr:.1%}   ({miss_corr.sum()} of {n_chapters})")
print(f"  The correlated team is unanimous-and-wrong about {rate_corr/rate_indep:.1f}x as often.")

fig, ax = plt.subplots(figsize=(7, 5))
# Independent gets the calm blue; correlated gets the warning orange (never the truth hue).
ax.bar(["Independent\nreviewers", "Correlated\nreviewers"],
       [rate_indep, rate_corr], color=["#4C72B0", "#DD8452"], edgecolor="white")
ax.set_ylim(0, 0.55)
ax.set_ylabel("Share of flawed chapters the team MISSES")
ax.set_title("Same four reviewers: shared blind spot manufactures false consensus")
for i, r in enumerate([rate_indep, rate_corr]):
    ax.text(i, r + 0.012, f"{r:.1%}", ha="center", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# Self-check: the correlated team misses real flaws several times more often.
assert 0.04 < rate_indep < 0.09, "independent miss rate drifted — is the seed 464?"
assert 0.40 < rate_corr < 0.47, "correlated miss rate drifted — is the seed 464?"
assert rate_corr > 5 * rate_indep, "the shared blind spot should dominate"
print(f"✓ Self-check passed: independent misses {rate_indep:.1%} of flawed chapters, "
      f"correlated misses {rate_corr:.1%}.\n  Four reviewers either way, about "
      f"{rate_corr/rate_indep:.0f}x the false-confidence rate. Agreement is only worth "
      f"the independence behind it.")
```

The independent team is unanimous-and-wrong on about **6%** of flawed chapters. The
correlated team, the same four reviewers, is unanimous-and-wrong on about **43%**,
roughly seven times as often. Nothing changed except whether the reviewers could miss
the flaw *together*. When they share a blind spot, four "looks fine" verdicts are not
four confirmations. They are close to one. This is why asking the same model twice,
or four roles built the same way, can feel like a chorus of agreement while every
voice is repeating the same mistake.

```python
# Sweep the team size. Fresh seed so the sweep is reproducible on its own (seed 464).
rng_sweep = np.random.default_rng(SEED)

team_sizes = [2, 4, 8, 16]
indep_rates, corr_rates = [], []
for kk in team_sizes:
    caught_i = rng_sweep.random((n_chapters, kk)) < detect
    indep_rates.append((~caught_i.any(axis=1)).mean())
    bl = rng_sweep.random(n_chapters) < rho
    caught_c = (rng_sweep.random((n_chapters, kk)) < detect) & (~bl[:, None])
    corr_rates.append((~caught_c.any(axis=1)).mean())

print("Miss rate as the team grows:")
print(f"  {'reviewers':>10} | {'independent':>12} | {'correlated':>11}")
for kk, ri, rc in zip(team_sizes, indep_rates, corr_rates):
    print(f"  {kk:>10} | {ri:>11.1%} | {rc:>10.1%}")
print(f"\n  The correlated floor sits near the shared-blind-spot rate ({rho:.0%}): "
      f"adding roles cannot push below it.")

fig, ax = plt.subplots()
ax.plot(team_sizes, indep_rates, marker="o", color="#4C72B0", label="independent reviewers")
ax.plot(team_sizes, corr_rates, marker="s", color="#DD8452", label="correlated reviewers")
ax.axhline(rho, color="#8C8C8C", linestyle=":", linewidth=1.3, label=f"shared blind-spot rate ({rho:.0%})")
ax.set_xlabel("Number of reviewers on the team")
ax.set_ylabel("Share of flawed chapters missed")
ax.set_title("More reviewers help only if their errors are independent")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Self-check: independence buys you everything; correlation floors out near rho.
assert indep_rates[0] > indep_rates[-1], "independent miss rate should fall as the team grows"
assert indep_rates[-1] < 0.01, "16 independent reviewers should almost never all miss"
assert corr_rates[-1] > 0.30, "the correlated team should stay stuck near the shared blind-spot floor"
print(f"✓ Self-check passed: independent misses fall from {indep_rates[0]:.1%} to "
      f"{indep_rates[-1]:.1%} across 2 to 16 reviewers;\n  the correlated team barely "
      f"moves, from {corr_rates[0]:.1%} to {corr_rates[-1]:.1%}. Count is not independence.")
```

*The two lines say it plainly: the independent team's misses drop toward zero as it
grows, while the correlated team flattens against a floor set by its shared blind
spot. Piling on more correlated reviewers cannot buy what independence buys.*

## 5. Escalation, Human Override, and Stopping Rules

**Guiding question:** *at what moment do you stop letting AI drive and take the
decision back into your own hands, and how do you know when you are done?*

Some decisions are never the tools' to make. The habit that protects them is
**escalation to human judgment**: the moment an AI output would *settle* rather than
*inform* a decision that is yours, you stop and decide it yourself. Example: a role
proposes calling your result "causal." That would settle your claim boundary, so it
escalates to you. The five triggers that force an escalation are listed in
[`ai_resources/escalation_protocol.md`](https://github.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/blob/main/ai_resources/escalation_protocol.md);
any one is enough.

What you do at an escalation is a **human override**: you take the pen back and make
the call in your own words, backed by your own verification, even when the roles
agreed on something else. Example: three roles agree your causal wording is fine, and
you overrule all three because your design has no identification argument. A
decision that just counted the roles is not a decision. It is a vote, and votes among
correlated models are meaningless.

You also need a **stopping rule**: a written statement of what "done" looks like,
about *your* confidence, not the roles' agreement. Example: "I stop when I can defend
each surviving claim to a hostile reviewer with a verification I ran myself." A bad
stopping rule is "when the roles all agree," because that can happen while everyone is
wrong, as your simulation just showed.

Finally, some calls are **decisions that should never be automated** no matter how
good the tools get: which research problem to pursue, the claim boundary, an ethical
judgment, and the public defense. These stay human by design.

> **A question that often comes up here:** *"If I keep overriding the AI, why use it
> at all?"* Because the tools are excellent at the parts you *can* delegate: drafting,
> surfacing leads, listing threats, tightening prose. Escalation does not mean
> distrusting everything. It means keeping a human at the few decisions that carry
> your name. A semester-long ledger with zero escalations is a warning sign, not a
> badge. It usually means AI was never used on anything that mattered, or a boundary
> was crossed without anyone noticing.

## 6. Staging the Evidence Defense

**Guiding question:** *how do you stand behind your claims out loud, at their
boundaries, under questions you cannot rehearse for?*

The course ends where it began, with a claim you would put your name on, now defended
in public. The **Evidence Defense** is a ten-minute defense followed by a five-minute
cross-examination, in which you defend not just *what* you claim but *why* your
compass position, your data, and your checks license it. The full format lives in
[`project/final_dossier/evidence_defense_protocol.md`](https://github.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/blob/main/project/final_dossier/evidence_defense_protocol.md).
Your ten minutes have three movements, in order:

- **Claims (about three minutes)**: your headline claim and the exact edge it stops
  at. Example: "In my sample I find X; I do not claim it holds for the population."
- **Choices (about four minutes)**: why this compass position, this data strategy,
  this operationalization, and the road you did not take.
- **Verification (about three minutes)**: how you checked, the sensitivity that
  survived, the sources confirmed real. Verification is the spine of the defense, not
  an afterthought.

Two more terms decide whether the defense goes well. **Concede gracefully** means
that where your ledger says a claim reaches too far, you say so precisely and name the
smaller true claim, instead of digging in. Conceding the right point on time is
scored as strength, not weakness. And the **forbidden moves** are three ways to lose
the room: inventing a new claim under pressure to escape a hard question, un-conceding
a point by insistence rather than evidence, and blaming the AI tool. "The AI gave me
that" is never a defense. You verified, or you did not.

All of this feeds your **AI-management portfolio**: the record of how you ran your AI
team this semester, what you automated, where you escalated, and which decisions you
kept human. It is read partly for whether you recognized those boundaries and took the
pen back when you should have.

> **A question that often comes up here:** *"What if I get a question I cannot
> answer?"* Answer from your ledger if the row exists, and if it does not, concede the
> point and name the smaller claim you *can* defend. That is a far stronger move than
> inventing an answer, which is a forbidden move precisely because the room can tell.
> A defense is not a performance of certainty. It is an honest account of what your
> evidence holds and where it stops.
