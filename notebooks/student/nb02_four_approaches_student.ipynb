## 1. Why This Matters

> *"Two people can ask what sounds like the same question and need completely
> different evidence to answer it. The first thing I do with any draft is figure
> out which of four questions the author is actually asking — because that
> decides everything they are allowed to conclude."*
> — a **thesis advisor**, on the first thing they check

You already built a question with units, an outcome, a comparison, and a scope.
That question is answerable — but *what kind* of answer does it want? That is not
a detail. It decides which evidence counts, which method fits, and, above all,
what your final poster is **permitted to claim**. Get the kind wrong and every
later choice is aimed at the wrong target.

There are exactly four kinds, and this whole course is built to teach you to hear
the difference the way a musician hears a wrong note. Here is the map you will
carry for the rest of your life:

| Approach | The question it asks | What it is really after |
|---|---|---|
| **Description** | What is here, in the data I have? | a **pattern**, in the world **as-is** |
| **Statistical inference** | What holds in the wider population, and how sure can I be? | a **pattern**, reaching **beyond** the sample |
| **Predictive modeling** | Can I forecast cases I have not seen? | a **pattern**, aimed at **new cases** |
| **Causal reasoning** | What would change if someone intervened? | a **process**, in a world you **change** |

> **A question that often comes up here:** *"Isn't one of these just the 'best'
> one — the most scientific?"* No — and believing that is how good researchers
> overreach. Each approach answers a different question honestly; the failure is
> never using description, it is using description and then *talking like you did
> causation*. The skill is not climbing a ladder. It is knowing which rung you
> are standing on.

## 2. The 2×2 Map — and the Left Column: Description and Inference

**Guiding question:** *what kind of answer does your question want — a pattern
or a process, in the world as-is or a world you reach into?*

Two yes/no questions place any research question on the map.

**Question 1 — pattern or process?** Are you asking *what is there / what goes
with what* (a pattern), or *what makes it happen* (a process — a mechanism you
could intervene on)?

**Question 2 — the world as-is, or a world you reach into?** Are you talking
about the world exactly as you observed it, or reaching past it — forward to new
cases, or sideways to a world where something was changed?

|  | **World as-is** (observe, don't change) | **Reach past it** (new cases / a change) |
|---|---|---|
| **Pattern** (what's there; no mechanism needed) | **Description** (this sample) → **Inference** (the population) | **Prediction** (forecast the next cases, vs a baseline) |
| **Process** (the mechanism — what produces what) | ⚠️ *the trap*: a cause **asserted** from pattern data, never tested | **Causal reasoning** (what a change would do; needs an identification argument) |

The single sharpest line on this map is the bottom-right cell: a real causal
claim lives there, and it is the hardest to earn. The most common failure in all
of research is sliding a top-row (pattern) finding into that bottom-left ⚠️ cell —
*asserting* a cause because two things move together. Today's left column is the
gentler half: **description** and **inference**, both about a pattern in the
world as-is. Friday takes the right column.

Two design levers decide whether a study can legitimately make the map's two
hardest crossings — the leap to a population and the leap to a cause. It is
worth meeting them now, even though earning them is October's work:

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/random_assignment_sampling.png" width="70%"/></center>

*Random **sampling** is what lets a finding generalize from a sample to a wider population, and random **assignment** is what lets a comparison support a causal claim. (From Professor Moreira's QM 67000 Business Analytics slides.)*

> **A question that often comes up here:** *"What is the difference between
> description and inference if both are just patterns?"* **Reach.** Description
> speaks only about the cases in front of you — the rows in your file. Inference
> uses those cases to say something about a wider population they were sampled
> from, and it must attach uncertainty to that leap. "In this survey, 53%…" is
> description. "In the country, about 53%, give or take a few points" is
> inference — a bigger claim that needs a defensible sample to stand on.

## 3. The Right Column I — Prediction: Forecasting Without Understanding

> *"A model on my desk forecasts next term's course demand almost perfectly. My
> staff wanted to use it to argue that our advising emails *cause* enrollment.
> Those are two completely different claims, and only one of them is in the
> model."*
> — a **policy stakeholder** (a university registrar), catching the slip early

Prediction asks a narrow, honest question: *for cases I have not seen yet, what is
my best guess of the outcome — and is that guess better than a dumb baseline?*
Notice what it does **not** ask: it does not ask *why*. A forecast can be
excellent and completely ignorant of mechanism. That is not a weakness — it is the
deal. You trade understanding for reach, and in return you get a number for cases
that have not happened.

Two ideas make a forecast honest:

- **A baseline.** "My model is good" means nothing alone. Good *compared to what?*
  The honest yardstick is a stupid rule — "predict the average every time." Beat
  that, and your forecast is earning its keep; fail to, and your fancy model is
  decoration.
- **New cases.** A forecast is a claim about data you did not use to build it.
  (Doing that split rigorously — and the ways it goes wrong — is nb12; today is
  the intuition.)

> **A question that often comes up here:** *"If a forecast works, doesn't that
> prove the thing it uses is a cause?"* No — and this is the costliest confusion
> in modern data work. Ice-cream sales forecast drowning deaths beautifully; ice
> cream drowns no one (summer heat drives both). A forecast rides on patterns,
> and patterns include every innocent bystander that happens to move with the
> outcome. "Predicts well" and "would change the outcome if we intervened" are
> different sentences, answered by different evidence.

```python
# A tiny forecasting demo — the whole past is 10 club events.
rsvp   = np.array([12, 30, 8, 22, 40, 17, 25, 5, 33, 19])
# Attendance TRACKS rsvps plus noise. We did not decree this rule; we observe it.
showed = np.clip((0.6 * rsvp + rng.normal(0, 3, size=10)).round(), 0, None).astype(int)

# Baseline: predict the overall average attendance for EVERY event.
baseline = np.full(10, showed.mean())
# Model: one number learned from the pattern — attendance per RSVP — times rsvp.
per_rsvp = showed.sum() / rsvp.sum()
model = per_rsvp * rsvp

mae_baseline = np.abs(showed - baseline).mean()
mae_model    = np.abs(showed - model).mean()

print("Attendance per RSVP (learned):", round(per_rsvp, 3))
print("Average error — baseline (guess the mean):", round(mae_baseline, 2))
print("Average error — model (use the RSVPs)    :", round(mae_model, 2))
print()
for new_rsvp in [28, 45]:
    print(f"Forecast for a NEW event with {new_rsvp} RSVPs:",
          round(per_rsvp * new_rsvp, 1), "expected to show up")
```

```python
# Self-check — the model should beat the baseline, and the seed fixes the numbers.
assert mae_model < mae_baseline, "model did not beat baseline — check the setup"
assert round(mae_baseline, 2) == 5.28 and round(mae_model, 2) == 2.91, "numbers drifted — is SEED = 464?"
print("✓ Self-check passed: model error 2.91 vs baseline 5.28 — the RSVP count carries real signal.")
```

## 4. The Right Column II — Causation: The Question About a Changed World

> *"Every causal claim is secretly a claim about a world that does not exist —
> the world where this person got the treatment *and* the world where the same
> person didn't. You never see both. So before I believe 'X changes Y,' I need to
> hear the argument for how you stood in for the world you couldn't observe."*
> — a **skeptical peer** reviewing your design

A causal question asks what would happen **if someone intervened** — if the class
moved to 10 a.m., if the scholarship doubled, if the app hid the like-counts.
Answering it means comparing the world as it is to a world that was *changed*, and
the deep problem is that you only ever get to see one of them. That is why causal
claims need an **identification argument**: a stated reason your comparison stands
in for the missing world (random assignment is the cleanest such reason — the
whole point of October).

Here is the same rough topic as four different questions, one per cell of the map,
so you can feel the borders:

- **Description:** *In this survey, what share of heavy short-video users report
  anxiety symptoms?* (a pattern, here)
- **Inference:** *Among all U.S. undergraduates, what share of heavy users are
  anxious, give or take?* (the pattern, out there)
- **Prediction:** *Given a new undergraduate's screen-time log, can we forecast
  their anxiety score?* (the pattern, for a new case)
- **Causal:** *If a heavy user cut their screen time in half, would their anxiety
  fall?* (a changed world — needs an identification argument)

Four sentences, four kinds of evidence, four different claims you are allowed to
make. The words "if… would" in the last one are the fingerprint of causation.
