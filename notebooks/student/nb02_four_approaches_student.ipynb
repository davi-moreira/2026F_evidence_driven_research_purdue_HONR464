## 1. Why This Matters

> *"Two people can ask what sounds like the same question and need completely
> different evidence to answer it. The first thing I do with any draft is work
> out what kind of question the author is asking, and how far the answer has to
> travel — because those two things decide everything they are allowed to
> conclude."*
> — a **thesis advisor**, on the first thing they check

You already built a question with units, an outcome, a comparison, and a scope.
That question is answerable — but *what kind* of answer does it want, and *for
which cases* must that answer hold? Neither is a detail. Together they decide
which evidence counts, which design fits, and, above all, what your final poster
is **permitted to claim**. Get them wrong and every later choice is aimed at the
wrong target.

Your course book asks exactly two questions of every inquiry, and this whole
course is built to teach you to hear them the way a musician hears a wrong note.
**Kind:** is the question *descriptive* (what the world is or was) or *causal*
(what an intervention would change)? **Reach:** must the answer hold for *the
data at hand*, for *a population beyond the data*, or for *cases not yet seen*?
The answers place your question at one of four named positions — the compass you
will carry for the rest of your life:

| Position (kind · reach) | The question it asks | What it is really after |
|---|---|---|
| **Description** (descriptive · the data at hand) | What is here, in the data I have? | a pattern, in **these** cases |
| **Generalization** (descriptive · a population) | What holds in the wider population, and how sure can I be? | the pattern, **beyond** the sample |
| **Prediction** (descriptive · unseen cases) | Can I forecast cases I have not seen? | the pattern, aimed at **new** cases |
| **Causal reasoning** (causal) | What would change if someone intervened? | a **changed world**, not just a pattern |

> **A question that often comes up here:** *"Isn't one of these just the 'best'
> one — the most scientific?"* No — and believing that is how good researchers
> overreach. Each position answers a different question honestly; the failure is
> never using description, it is using description and then *talking like you did
> causation*. The skill is not climbing a ladder. It is knowing which position
> you are standing on.

## 2. The Compass — and Today's Half: the Descriptive Kind at Two Reaches

**Guiding question:** *what kind of answer does your question want — descriptive
or causal — and how far must that answer reach?*

Two questions place any research question on the compass.

**Question 1 — KIND: descriptive or causal?** Are you asking *what the world is
or was* (descriptive — no imagined intervention needed), or *what would change if
someone intervened* (causal — the answer requires comparing this world to a
counterfactual one)?

**Question 2 — REACH: for which cases must the answer hold?** The data at hand —
the rows actually in your file? A population beyond the data, which your cases
somehow stand in for? Or cases not yet seen — the future events, the next
applicants, the readers you haven't met?

|  | **the data at hand** | **a population beyond it** | **cases not yet seen** |
|---|---|---|---|
| **Descriptive kind** | **Description** | **Generalization** | **Prediction** |
| **Causal kind** | **Causal reasoning** — at any reach, but only under an identification argument. ⚠️ *The trap*: causal talk with no such argument — a pattern narrated as a cause. | | |

Here is the discipline hiding in that little table, and it is the deepest idea in
the course: **every step away from "descriptive · data at hand" must be paid
for with a design element.** Reaching a population is *bought* with a sampling
strategy and honest uncertainty. Reaching unseen cases is *bought* with a
baseline and an evaluation the model never peeked at. Crossing from descriptive
to causal is *bought* with random assignment or a defended identification
argument. Every overclaim you will ever catch — in the news, in a journal, in
your own draft — is a compass crossing somebody made without paying. Two of
those purchase mechanisms deserve a face early, even though earning them is
October's work:

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/random_assignment_sampling.png" width="70%"/></center>

*Random **sampling** buys reach — it lets a finding generalize from a sample to a wider population. Random **assignment** buys the causal kind — it lets a comparison support a causal claim. (From Professor Moreira's QM 67000 Business Analytics slides.)*

And whichever position your question takes, the rest of the semester will make
you say it in the book's four-part grammar — a **M**odel of the world, an
**I**nquiry (the exact quantity you seek), a **D**ata strategy, and an **A**nswer
strategy:

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/rdss_fig_2_1.png" width="70%"/></center>

*MIDA — every declared design pairs the theoretical side (M, I) with the empirical side (D, A). The compass fixes what I is and how far it reaches; MIDA is how you declare and defend it, starting in nb04. (Figure from the replication materials of Blair, Coppock & Humphreys (2023), Research Design in the Social Sciences — MIT-licensed archive; the book is free at book.declaredesign.org.)*

Today is the gentler half of the compass: the **descriptive kind** at its first
two reaches — **description** and **generalization**. Friday takes prediction
and the causal kind.

> **A question that often comes up here:** *"What is the difference between
> description and generalization if both are descriptive?"* **Reach — nothing
> else.** Description speaks only about the cases in front of you — the rows in
> your file. Generalization uses those cases to say something about a wider
> population they were sampled from, and it must attach uncertainty to that
> leap. "In this survey, 53%…" is description. "In the country, about 53%, give
> or take a few points" is generalization — a bigger claim that needs a
> defensible sample to stand on.

## 3. Prediction — the Descriptive Kind Reaching Cases Not Yet Seen

> *"A model on my desk forecasts next term's course demand almost perfectly. My
> staff wanted to use it to argue that our advising emails *cause* enrollment.
> Those are two completely different claims, and only one of them is in the
> model."*
> — a **policy stakeholder** (a university registrar), catching the slip early

Prediction asks a narrow, honest question: *for cases I have not seen yet, what is
my best guess of the outcome — and is that guess better than a dumb baseline?*
Notice what it does **not** ask: it does not ask *why*. A forecast can be
excellent and completely ignorant of mechanism. That is not a weakness — it is the
deal. You trade understanding for reach, and in return you get a number for cases
that have not happened.

Two ideas make a forecast honest:

- **A baseline.** "My model is good" means nothing alone. Good *compared to what?*
  The honest yardstick is a stupid rule — "predict the average every time." Beat
  that, and your forecast is earning its keep; fail to, and your fancy model is
  decoration.
- **New cases.** A forecast is a claim about data you did not use to build it.
  (Doing that split rigorously — and the ways it goes wrong — is nb12; today is
  the intuition.)

> **A question that often comes up here:** *"If a forecast works, doesn't that
> prove the thing it uses is a cause?"* No — and this is the costliest confusion
> in modern data work. Ice-cream sales forecast drowning deaths beautifully; ice
> cream drowns no one (summer heat drives both). A forecast rides on patterns,
> and patterns include every innocent bystander that happens to move with the
> outcome. "Predicts well" and "would change the outcome if we intervened" are
> different sentences, answered by different evidence.

```python
# A tiny forecasting demo — the whole past is 10 club events.
rsvp   = np.array([12, 30, 8, 22, 40, 17, 25, 5, 33, 19])
# Attendance TRACKS rsvps plus noise. We did not decree this rule; we observe it.
showed = np.clip((0.6 * rsvp + rng.normal(0, 3, size=10)).round(), 0, None).astype(int)

# Baseline: predict the overall average attendance for EVERY event.
baseline = np.full(10, showed.mean())
# Model: one number learned from the pattern — attendance per RSVP — times rsvp.
per_rsvp = showed.sum() / rsvp.sum()
model = per_rsvp * rsvp

mae_baseline = np.abs(showed - baseline).mean()
mae_model    = np.abs(showed - model).mean()

print("Attendance per RSVP (learned):", round(per_rsvp, 3))
print("Average error — baseline (guess the mean):", round(mae_baseline, 2))
print("Average error — model (use the RSVPs)    :", round(mae_model, 2))
print()
for new_rsvp in [28, 45]:
    print(f"Forecast for a NEW event with {new_rsvp} RSVPs:",
          round(per_rsvp * new_rsvp, 1), "expected to show up")
```

```python
# Self-check — the model should beat the baseline, and the seed fixes the numbers.
assert mae_model < mae_baseline, "model did not beat baseline — check the setup"
assert round(mae_baseline, 2) == 5.28 and round(mae_model, 2) == 2.91, "numbers drifted — is SEED = 464?"
print("✓ Self-check passed: model error 2.91 vs baseline 5.28 — the RSVP count carries real signal.")
```

## 4. The Causal Kind — The Question About a Changed World

> *"Every causal claim is secretly a claim about a world that does not exist —
> the world where this person got the treatment *and* the world where the same
> person didn't. You never see both. So before I believe 'X changes Y,' I need to
> hear the argument for how you stood in for the world you couldn't observe."*
> — a **skeptical peer** reviewing your design

A causal question asks what would happen **if someone intervened** — if the class
moved to 10 a.m., if the scholarship doubled, if the app hid the like-counts.
Answering it means comparing the world as it is to a world that was *changed*, and
the deep problem is that you only ever get to see one of them. That is why causal
claims need an **identification argument**: a stated reason your comparison stands
in for the missing world (random assignment is the cleanest such reason — the
whole point of October).

Here is the same rough topic as four different questions, one per compass
position, so you can feel the borders:

- **Description** (descriptive · data at hand): *In this survey, what share of
  heavy short-video users report anxiety symptoms?* (the pattern, here)
- **Generalization** (descriptive · a population): *Among all U.S.
  undergraduates, what share of heavy users are anxious, give or take?* (the
  pattern, out there)
- **Prediction** (descriptive · unseen cases): *Given a new undergraduate's
  screen-time log, can we forecast their anxiety score?* (the pattern, for a new
  case)
- **Causal** (the causal kind): *If a heavy user cut their screen time in half,
  would their anxiety fall?* (a changed world — needs an identification
  argument)

Four sentences, four kinds of evidence, four different claims you are allowed to
make. The words "if… would" in the last one are the fingerprint of the causal
kind.
