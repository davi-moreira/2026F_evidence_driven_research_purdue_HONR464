## 1. Why This Matters

> *"Two people can ask what sounds like the same question and need completely
> different evidence to answer it. The first thing I do with any draft is work
> out what kind of question it asks, and how far the answer has to travel.
> Those two things decide everything the author is allowed to conclude."*
> — a **thesis advisor**, on the first thing they check

In nb01 you built a question with units, an outcome, a comparison, and a scope.
That question is answerable. But two things about it are still undecided. What
type of answer does it want? And for which cases must that answer be true?
Neither is a detail. Together they decide which evidence counts, which design
fits, and, above all, what your final poster is permitted to claim. Get them
wrong and every later choice aims at the wrong target.

Your course book asks exactly two questions of every inquiry, and this notebook
trains you to hear them the way a musician hears a wrong note. The two answers
place your question at one of four named spots on the course's signature
instrument, the **inquiry compass**:

| Position | The question it asks | What it is really after |
|---|---|---|
| **Description** | What is here, in the data I have? | a pattern, in **these** cases |
| **Generalization** | What holds in the wider population, and how sure can I be? | the pattern, **beyond** the sample |
| **Prediction** | Can I forecast cases I have not seen? | the pattern, aimed at **new** cases |
| **Causal reasoning** | What would change if someone intervened? | a **changed world**, not just a pattern |

> **A question that often comes up here:** *"Isn't one of these just the 'best'
> one, the most scientific?"* No, and believing that is how good researchers
> overreach. Each position answers a different question honestly. The failure
> is never using description; the failure is using description and then talking
> like you did causation. The skill is not climbing a ladder. It is knowing
> which position you are standing on.

## 2. The Compass: Two Questions, Four Positions

**Guiding question:** *what kind of answer does your question want, and how far
must that answer reach?*

Two words do all the work in this notebook, so let's pin them down before
anything else.

**Kind** is the first question you ask of any research question: what type of
answer does it want? A **descriptive** question asks what the world is or was:
how many, how much, how often, who believes what. "How many first-years at
Purdue skip breakfast?" is descriptive. A **causal** question asks what would
change if someone intervened. "Would a free breakfast program raise first-year
attendance?" is causal. The giveaway is the hidden "if... would": a causal
question compares the world as it is to a changed version of it, and you never
get to watch both.

**Reach** is the second question: for which cases must the answer hold? Three
distances are possible. **The data at hand** means only the cases actually in
front of you. If you ask who in your group chat prefers morning classes, the
answer only has to be true for the chat. **A population beyond the data** means
a wider group your cases stand in for. A **population** is simply every case
your claim wants to speak about, like all Purdue first-years when you surveyed
only forty of them. **Cases not yet seen** means cases with no row in your data
yet: next semester's enrollment, tomorrow's applicants, the roommate you have
not met.

Cross the two answers and you get the compass:

|  | **the data at hand** | **a population beyond it** | **cases not yet seen** |
|---|---|---|---|
| **Descriptive kind** | **Description** | **Generalization** | **Prediction** |
| **Causal kind** | **Causal reasoning**: at any reach, and only ever under an **identification argument**, a stated reason your comparison can stand in for the changed world you never observed (say, a lottery deciding who got a scholarship) | | |

A **position**, in this course's vocabulary, is one of those four named spots:
the place where your question's kind and reach meet. Say your position out loud
for your own question and you have made your first real design decision.

> **A question that often comes up here:** *"What is the difference between
> description and generalization if both are descriptive?"* Reach, nothing
> else. Description speaks only about the cases in front of you, the rows in
> your file. Generalization uses those cases to say something about a wider
> population they were drawn from, and it must attach uncertainty to that leap.
> "In this survey, 53% trust the police" is description. "In the country, about
> 53%, give or take a few points" is generalization: a bigger claim that needs
> a defensible sample to stand on.

### The Three Crossings, and What Each One Costs

Here is the deepest idea in the course, hiding in that little table. Every step
away from "descriptive, data at hand" is a **crossing**: a move that makes your
claim travel farther than your raw evidence. And every crossing must be paid
for with a **license**: a design element that earns the bigger claim, the way a
driver's license proves you did the training before you took the highway.

| The crossing | The license that pays for it | The violation if you skip it |
|---|---|---|
| data at hand → a population | a **sampling strategy** (a deliberate plan for how cases enter your data, so they can stand in for the wider group) plus honest uncertainty | the *silent upgrade*: "these 200 responses" quietly becomes "Americans" |
| data at hand → unseen cases | a **baseline** (a deliberately simple guess, like "predict the average every time") plus a test on cases the forecast never used | grading a forecast on the very data it memorized |
| descriptive → causal | random assignment, or a defended identification argument | narrating a pattern as a cause |

Every overclaim you will ever catch, in the news, in a journal, or in your own
draft, is one of these three crossings made without paying. Two of the licenses
deserve a face early, even though earning them comes later in the course.
**Random sampling** means chance decides which cases enter your data. **Random
assignment** means chance decides which cases receive a treatment. They sound
alike and they buy completely different things:

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/random_assignment_sampling.png" width="70%"/></center>

*Random **sampling** buys reach: it lets a finding travel from a sample to a wider population. Random **assignment** buys the causal kind: it lets a comparison support a causal claim. (From Professor Moreira's QM 67000 Business Analytics slides.)*

And whichever position your question takes, the rest of the semester will teach
you to state it in the book's four-part grammar: a **M**odel of the world, an
**I**nquiry (the exact quantity you seek), a **D**ata strategy, and an
**A**nswer strategy.

<center><img src="https://raw.githubusercontent.com/davi-moreira/2026F_evidence_driven_research_purdue_HONR464/main/notebooks/figures/rdss_fig_2_1.png" width="70%"/></center>

*MIDA: every declared design pairs the theoretical side (M, I) with the empirical side (D, A). The compass names your inquiry; MIDA, starting in nb04, is how you declare and defend it. (Figure from the replication materials of Blair, Coppock & Humphreys (2023), Research Design in the Social Sciences, an MIT-licensed archive; the book is free at book.declaredesign.org.)*

Enough map. Time to stand somewhere. The gentlest position is home base,
description, and the best way to feel its edge is to make one honest claim and
then watch how quickly you are tempted past it.

## 4. Feeling the Borders: One Topic, Asked Four Ways

> *"A model on my desk forecasts next term's course demand almost perfectly. My
> staff wanted to use it to argue that our advising emails cause enrollment.
> Those are two completely different claims, and only one of them is in the
> model."*
> — a **policy stakeholder** (a university registrar), catching the slip early

Take one rough topic, short-video use and anxiety, and ask it from every
position on the compass:

- **Description** (descriptive · the data at hand): *In this survey, what share
  of heavy short-video users report anxiety symptoms?* The pattern, here.
- **Generalization** (descriptive · a population): *Among all U.S.
  undergraduates, what share of heavy users are anxious, give or take?* The
  pattern, out there.
- **Prediction** (descriptive · unseen cases): *Given a new undergraduate's
  screen-time log, can we forecast their anxiety score?* The pattern, for a
  case with no row in the data yet.
- **Causal reasoning** (the causal kind): *If a heavy user cut their screen
  time in half, would their anxiety fall?* A changed world, and it needs an
  identification argument.

Four sentences, four kinds of evidence, four different claims you are allowed
to make. The words "if... would" in the last one are the fingerprint of the
causal kind: the question is about a world that was changed, and no pattern in
the unchanged world settles it by itself.

The prediction position deserves one more beat, because it carries the
costliest confusion in modern data work. Prediction asks a narrow, honest
question: for cases I have not seen, what is my best guess, and does that guess
beat a baseline? Notice what it does not ask. It does not ask *why*. A forecast
can be excellent and completely ignorant of mechanism. That is the deal: you
trade understanding for reach, and you get a number for cases that have not
happened yet.

> **A question that often comes up here:** *"If a forecast works, doesn't that
> prove the thing it uses is a cause?"* No. Ice-cream sales forecast drowning
> deaths beautifully, and ice cream drowns no one; summer heat drives both. A
> forecast rides on patterns, and patterns include every innocent bystander
> that happens to move with the outcome. "Predicts well" and "would change the
> outcome if we intervened" are different sentences, answered by different
> evidence.

### 🛠️ Hands-On (optional): a tiny forecast, judged against a baseline

The next cell holds ten past events of a campus club: how many people RSVP'd,
and how many actually showed up. It builds two forecasts of attendance. The
**baseline** guesses the overall average every time. The **model** uses the
RSVP count. Each is scored by its average miss (how far off it lands, on
average; smaller is better). Then the model forecasts two brand-new events the
club has not held yet.

Before you run it, make two quick calls. Will the RSVP model beat the
average-guessing baseline? And if it does, would that justify blasting out more
RSVP invites in order to raise turnout?

> 💡 **Gemini Prompt:** "Here is a small forecasting cell from my research
> course: [paste the next cell]. Explain what the 'baseline' is predicting
> versus what the 'model' is predicting, and tell me which one you expect to
> have the smaller average error and why."
>
> **After running, verify (the responsible-AI habit):**
> - [ ] Check Gemini's guess against the two error numbers the cell prints. Did
>   the model really beat the baseline, and by roughly the margin it expected?
> - [ ] If it says the model "understands" why people attend, push back: the
>   cell only learns attendance-per-RSVP. Confirm nothing in the output
>   supports a causal reading.
> - [ ] Log this use in your AI ledger: tool, task, verification.

```python
# A tiny forecasting demo — the whole past is 10 club events.
rsvp   = np.array([12, 30, 8, 22, 40, 17, 25, 5, 33, 19])
# Attendance TRACKS rsvps plus noise. We did not decree this rule; we observe it.
showed = np.clip((0.6 * rsvp + rng.normal(0, 3, size=10)).round(), 0, None).astype(int)

# Baseline: predict the overall average attendance for EVERY event.
baseline = np.full(10, showed.mean())
# Model: one number learned from the pattern — attendance per RSVP — times rsvp.
per_rsvp = showed.sum() / rsvp.sum()
model = per_rsvp * rsvp

mae_baseline = np.abs(showed - baseline).mean()
mae_model    = np.abs(showed - model).mean()

print("Attendance per RSVP (learned):", round(per_rsvp, 3))
print("Average error — baseline (guess the mean):", round(mae_baseline, 2))
print("Average error — model (use the RSVPs)    :", round(mae_model, 2))
print()
for new_rsvp in [28, 45]:
    print(f"Forecast for a NEW event with {new_rsvp} RSVPs:",
          round(per_rsvp * new_rsvp, 1), "expected to show up")
```

```python
# Self-check — the model should beat the baseline, and the seed fixes the numbers.
assert mae_model < mae_baseline, "model did not beat baseline — check the setup"
assert round(mae_baseline, 2) == 5.28 and round(mae_model, 2) == 2.91, "numbers drifted — is SEED = 464?"
print("✓ Self-check passed: model error 2.91 vs baseline 5.28 — the RSVP count carries real signal.")
```
