## 1. Why This Matters: the Talking Is the Poster Now

**Guiding question:** *a stranger gives you thirty seconds, then maybe three
minutes. What do you say in each, and how do you keep every version honest?*

> *"I stop at a poster and the presenter has about a minute before I decide
> whether to stay. If those seconds are a confident, honest tour of one claim, I
> lean in. If they are an apology or a wall of data, I nod and drift. The talking
> is the poster now. The paper on the board is just the visual aid."*
> — a **skeptical peer** you will meet many times at the Expo

Your poster is done. Now you learn to stand beside it and speak it. What carries
a poster session is a **pitch**: a short, prepared spoken version of your project,
built to fit a fixed slice of time. The 30 seconds you give a visitor who pauses
is a pitch. So is the guided tour you give the visitor who stays.

This topic runs across two lectures. Lecture 1 builds the three pitch layers and
puts them in your mouth through delivery rounds. Lecture 2 teaches you to speak
your study's limits out loud and to answer the hardest fair question without
bluffing. The iron rule across both is **compression without inflation**: a
**compression** is a shorter version of a claim that keeps its boundary, and an
**inflation** is a version that quietly makes the claim bigger. Example: "in my
sample, longer commutes went with lower attendance" compresses to "in my survey,
longer commutes tracked with skipping more class," but inflates the moment it
becomes "longer commutes make people skip class."

> **A question that often comes up here:** *"I'm not a performer. Won't a scripted
> pitch sound fake?"* A script is not a performance. It is preparation. You are
> not memorizing lines to recite. You are deciding, calmly and in advance, exactly
> what you will say when a stranger is staring and your pulse is up. The boundary
> sentence especially: you rehearse it until it is automatic, because nerves erase
> precisely the careful wording you most need.

## 2. The Three Pitch Layers, and the One Rule That Governs Them

**Guiding question:** *how do the 30-second, 90-second, and full pitches differ in
job, and where does the claim boundary live when you are speaking, not writing?*

A poster session asks for three spoken versions of the same project, sized to how
much time a visitor gives you. All three are the first component of milestone
**M11**, your presentation package. They share one skeleton, the **pitch
architecture**: the fixed set of beats every good pitch moves through. Its five
beats are **hook, claim, evidence, boundary, invitation**.

- **Hook**: one sentence a stranger outside your field cares about. Example: "every
  morning thousands of commuters trade sleep for a drive."
- **Claim**: your headline, said plainly and inside its boundary. Example: "in my
  survey, longer commutes went with lower attendance."
- **Evidence**: the finding that earns the claim, in one spoken sentence per block.
- **Boundary sentence**: one sentence naming what the claim does NOT show. Example:
  "this is an association in one sample, so it cannot say the drive caused it."
  Mark this sentence and rehearse it until it is reflex. It is the one pressure
  will try to drop.
- **Invitation**: a reason to look closer. Example: "the surprising part is on the
  left, want the walk?"

The three layers are the same architecture at three lengths:

- **The 30-second hook**: hook, claim, boundary, invitation. Its whole job is to
  earn the next two minutes. No evidence blocks yet.
- **The 90-second walk**: the hook plus two or three evidence blocks and the honest
  bound. It moves the visitor's eye along the poster's read path.
- **The full pitch**: the three-to-five-minute version for the visitor who wants
  the whole tour, evidence in your ranked order, ending on uncertainty and an
  invitation.

The longer layers **add** blocks to the same spine. They never restate a bigger
claim. If your full pitch says something larger than your 30-second hook, the
extra time inflated the claim instead of supporting it.

Two more moves keep the spoken version honest, and both are graded in M11.

- **Audience adaptation**: changing which words you use for who is listening, while
  keeping the claim identical. Example: you lead a fellow biology major with the
  mechanism, and a visitor from the art school with the stakes, but the headline
  and its boundary do not change between them.
- **Jargon**: a field-specific word a stranger cannot decode without training.
  Example: "heteroskedasticity" is jargon at a poster. Explaining your method
  *without* jargon means naming what you did in plain words, so the visitor follows
  the reasoning rather than nodding at a term they cannot climb.

The single failure that this whole week hunts is the **excited-mouth upgrade**:
the moment your spoken verb outruns the written one and the boundary moves while
nobody is watching the words. You wrote "is associated with"; the adrenaline says
"causes." You wrote "in my sample of 812"; the mouth says "in everyone." Your
boundary sentence exists to hold that line out loud.

> **A question that often comes up here:** *"How is compressing different from
> inflating? Both change the claim."* Compression drops detail while keeping the
> boundary. "In my sample of 812 commuters, longer commutes go with lower
> attendance" compresses to "in my survey, longer commutes track with skipping
> more class." Inflation moves the boundary: "longer commutes make people skip
> class." Same length, different sin. One lost a number. The other lost the truth.

---

```python
# Draw the pitch architecture: the five beats, and how the three layers nest.
# (networkx is imported once in the setup cell.)
G = nx.DiGraph()
pos = {
    "Hook":        (0.0, 0.0),
    "Claim":       (1.3, 0.0),
    "Evidence 1":  (2.6, 0.6),
    "Evidence 2":  (2.6, -0.6),
    "Boundary\nsentence": (3.9, 0.0),
    "Invitation":  (5.2, 0.0),
}
edges = [
    ("Hook", "Claim"),
    ("Claim", "Evidence 1"),
    ("Claim", "Evidence 2"),
    ("Evidence 1", "Boundary\nsentence"),
    ("Evidence 2", "Boundary\nsentence"),
    ("Claim", "Boundary\nsentence"),        # the 30-second skip: straight to the bound
    ("Boundary\nsentence", "Invitation"),
]
G.add_edges_from(edges)

# The boundary node is drawn with a heavy dark outline (not a color code) so it
# reads as the load-bearing beat even in grayscale.
node_colors = ["#DCE6F1"] * len(pos)
outline = ["#4C72B0"] * len(pos)
widths = [1.5] * len(pos)
nodes = list(pos.keys())
bi = nodes.index("Boundary\nsentence")
outline[bi] = "#222222"
widths[bi] = 3.5

fig, ax = plt.subplots(figsize=(11, 4.5))
nx.draw_networkx_nodes(G, pos, node_color=node_colors, edgecolors=outline,
                       linewidths=widths, node_size=4600, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=9, ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle="-|>", arrowsize=16, edge_color="#666666",
                       width=1.4, node_size=4600, ax=ax)
ax.set_title("The pitch architecture: one spine, three lengths")
ax.axis("off")
# Annotate which layer uses which beats.
ax.text(0.0, -1.4, "30-second hook: Hook → Claim → Boundary → Invitation (skips evidence)",
        fontsize=9, color="#333333")
ax.text(0.0, -1.7, "90-second walk & full pitch: the same spine, with evidence blocks added",
        fontsize=9, color="#333333")
plt.tight_layout()
plt.show()

print("✓ Pitch architecture drawn.")
print("  The heavy-outlined 'Boundary sentence' beat is the one you rehearse to reflex.")
print("  Longer layers ADD evidence to this spine; they never enlarge the claim.")
```

**Reading the diagram.** Every pitch walks the same path, left to right: hook,
claim, then either straight to the boundary (the 30-second version) or through the
evidence blocks first (the longer versions), and out on the invitation. The
heavy-outlined boundary beat is the one that keeps the claim honest, and it is the
one you cannot skip at any length.

*In plain terms, the picture says all three pitches are the same skeleton. The
short one is the skeleton; the long ones hang evidence on it without growing the
head.*

```python
# Pacing calculator — how long will a script actually take to say?
RATE_WPM = 130   # a clear, unrushed pace; under pressure people speed up, so budget slow

def words_that_fit(seconds, wpm=RATE_WPM):
    # Seconds times words-per-second gives the word budget for that slice of time.
    return int(seconds * wpm / 60)

def duration_seconds(script_text, wpm=RATE_WPM):
    # Words in the script, divided by words-per-second, gives seconds aloud.
    return len(script_text.split()) / (wpm / 60.0)

budget_30 = words_that_fit(30)
budget_90 = words_that_fit(90)
budget_full = words_that_fit(240)   # the full pitch: budget the 4-minute end of 3-to-5
print(f"At {RATE_WPM} words per minute, an honest budget is:")
print(f"  30-second hook : about {budget_30} words")
print(f"  90-second walk : about {budget_90} words")
print(f"  full pitch     : about {budget_full} words  (at 4 minutes)")

# A sample 30-second hook: hook -> claim -> boundary sentence -> invitation.
my_hook = (
    "Every morning thousands of Purdue commuters trade sleep for a long drive. "
    "In my survey of 812 of them, longer commutes went with lower reported class "
    "attendance. I want to be precise. This is an association in one campus sample, "
    "not proof the drive itself causes the absences. Come see the pattern, and "
    "tell me what you think could explain it."
)
words = len(my_hook.split())
secs = duration_seconds(my_hook)
print(f"\nSample hook: {words} words -> {secs:.0f} seconds (budget 30s, so it fits with room).")
```

```python
# Self-check: the pacing math is an identity, and the sample hook fits its 30 seconds.
assert words_that_fit(60) == 130, "one minute must be 130 words at 130 wpm"
assert budget_30 == 65 and budget_90 == 195, "word budgets drifted — check RATE_WPM"
assert secs <= 30, "sample hook overruns 30 seconds — cut words, never speed up"
print(f"✓ Self-check passed: 30s holds {budget_30} words, 90s holds {budget_90}; "
      f"the {words}-word sample hook lands at {secs:.0f}s, inside the 30-second budget.")
```

## 3. Uncertainty and Limitations, Spoken Like a Scientist

**Guiding question:** *can you say what your study cannot show without either
bluffing or apologizing your finding away?*

> *"When a presenter tells me, unprompted, exactly what their study can't show, I
> trust everything else they said more. When they insist it's airtight, I start
> looking for the crack myself. Knowing where your claim stops is the most expert
> thing you can say."*
> — a **policy stakeholder** who has heard a thousand pitches

Here is the counterintuitive heart of scientific communication: stating your
limits makes you more believable, not less. Two terms carry the move.

- **Uncertainty statement**: a sentence saying how much your number could wobble,
  and why. Example: "with 812 people, my estimate could easily be a few points
  higher or lower."
- **Limitation**: a true sentence about what your design cannot establish. Example:
  "a one-time survey cannot show that commuting causes the absences."

The move that delivers both out loud is the **ULN move**: uncertainty, then
limitation, then next step, three sentences in order with a straight spine.

- **U, uncertainty:** how much wiggle sits around your estimate, and why.
- **L, limitation:** what your design cannot establish, named plainly. This is the
  boundary itself.
- **N, next step:** the design that *would* settle it. This is the study you would
  run next.

The **next step** is what turns a limitation from an apology into a research
agenda. It says: I know exactly what I would do next, which is the mark of someone
in command of their evidence.

> **A question that often comes up here:** *"If I lead with what my study can't do,
> won't people just dismiss it?"* They dismiss the drowner, not the namer. "This is
> an association in one campus sample, it can't establish cause, and a panel study
> could" is three seconds of pure competence. It tells the visitor you are the
> person in the room who most precisely knows the finding's shape.

---

### The two ways to fail the move: hedging versus precision

The ULN move has two failure modes, and both misstate the evidence.

- **Hedging**: vague self-protection that blurs where the claim stops. Example:
  "sort of," "probably wrong," "I guess." Hedging protects you and tells the
  listener nothing.
- **Precision**: naming the exact boundary. Example: "this is an association in one
  campus sample." Precision informs the listener.

The worst hedging failure has its own name, the **apology spiral**: burying a real,
defensible finding under self-erasure. Example: "honestly this is probably all
wrong, my sample's tiny, I'm not a statistician." The antidote is physical and
verbal. Plant your feet, drop "sorry" and "just" and "I know this is bad," and end
on the next step, not the flaw. You are reporting a boundary, not confessing a crime.

Partners score each ULN delivery on the **strength-not-apology scale**, a 1-to-5
anchored rating where 1 is the apology spiral and 5 is expertise. Aim for 4 or 5.
The figure below draws the scale so you can hear where a delivery lands.

### 🛠️ Hands-On: the apology-spiral detector

Run the cell. It scans three sample deliveries for self-erasing words and reports a

> '[paste your ULN]'. My detector flags these self-erasing words: sorry, just,
> probably, i guess, i'm no, not sure, sort of, kind of, maybe, possibly, all
> wrong, can't really. First, PREDICT exactly how many of those my delivery
> contains, as a number. Then score it 1 to 5 on a strength-not-apology scale and

>
> **After running, verify (counters *confident fabrication*: a predicted count is a
> guess until the detector runs, and a fluent score can invent flags you never said):**
> - [ ] Paste your delivery into `my_delivery`, rerun the cell, and compare its
>   printed flag count to Gemini's predicted number. If they differ, the cell is right.
> - [ ] Check every word Gemini "flagged" against your actual sentence. Reject any
>   flag for a word you did not say.
> - [ ] Keep the score only if it matches what you hear: a delivery that ends on a
>   next step should not be scored a 1 or 2.
> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# Apology-spiral detector — counts self-erasing words and checks for a next step.
APOLOGY_MARKERS = ["sorry", "just", "probably", "i guess", "i'm no", "not sure",
                   "sort of", "kind of", "maybe", "possibly", "all wrong", "can't really"]
NEXT_STEP_MARKERS = ["next step", "the study that could", "the design that could",
                     "would settle", "how you'd find out", "how one would find out"]

def audit_delivery(text):
    low = text.lower()
    flags = sum(low.count(m) for m in APOLOGY_MARKERS)
    ends_on_next_step = any(m in low for m in NEXT_STEP_MARKERS)
    return flags, ends_on_next_step

deliveries = {
    "A (apology spiral)": (
        "Honestly this is probably all wrong, it's just 812 people from one campus, "
        "and I'm no statistician, so I guess you can't really conclude anything."),
    "B (hedge-heavy)": (
        "It sort of seems like maybe there might be a kind of weak association, "
        "possibly, but I'm not sure it really holds up."),
    "C (clean ULN move)": (
        "The estimate carries a wide interval because it is one campus sample, so I "
        "read it as a real pattern, not a precise number. Because it is an "
        "association, it cannot show the drive causes absences. The study that could "
        "settle it is a panel that follows people as their commutes change."),
}

# Paste your own delivery here to test it, then rerun.
my_delivery = ""

print(f"{'Delivery':<22}{'apology flags':>14}{'ends on next step?':>22}")
for name, text in deliveries.items():
    flags, nxt = audit_delivery(text)
    print(f"{name:<22}{flags:>14}{('yes' if nxt else 'no'):>22}")
if my_delivery.strip():
    f, n = audit_delivery(my_delivery)
    print(f"{'MINE':<22}{f:>14}{('yes' if n else 'no'):>22}")
```

```python
# Self-check: the spiral and hedge deliveries trip flags; the clean ULN trips none
# AND is the only one that ends on a next step.
fa, na = audit_delivery(deliveries["A (apology spiral)"])
fb, nb_ = audit_delivery(deliveries["B (hedge-heavy)"])
fc, nc = audit_delivery(deliveries["C (clean ULN move)"])
assert fa >= 5 and fb >= 4, "sample apology/hedge deliveries should trip several flags"
assert fc == 0, "the clean ULN delivery must trip zero apology flags"
assert nc and not na and not nb_, "only the clean ULN delivery ends on a next step"
print(f"✓ Self-check passed: A flags {fa}, B flags {fb}, C flags {fc}; only C ends on a next step.")
print("  The detector catches self-erasure. It CANNOT see whether C's boundary is the RIGHT one.")
```

### The causal question aimed at a noncausal project

Now the puzzle's payoff. A visitor asks **"so does it CAUSE it?"** about your
association-only finding. The honest reply is a compressed ULN move, two sentences,
and it does both jobs at once: it refuses the causal word, and it hands the visitor
something to chew on so they stay interested.

The shape is: **name the boundary, then offer the design that would answer their

curiosity as a real research question, not a mistake).

> stranger has just committed the excited-mouth upgrade in public, and any careful
> visitor now trusts nothing else on the board.

## 4. The Hardest Fair Question, Prepared in Advance

**Guiding question:** *what does an honest "I don't know" contain, and how do you
decide, on the spot, whether to defend your claim or concede a point?*

At the Expo a stranger will ask you the hardest fair question about your project,
and how you answer decides whether they leave trusting you. The instrument that
turns that from an ambush into a strength is the **question bank**: a short written
list of the hardest fair questions a visitor could ask, each with a prepared
answer. Think of it as a study guide you write for an exam only you will take.
Yours needs five, one of each recurring type. This is the second component of M11.

- **Method**: probes whether your compass position can bear the claim. Example:
  "why is a description enough here, didn't you need a comparison group?"
- **Alternative explanation**: names a rival cause of your pattern. Example:
  "couldn't selection into your sample produce this on its own?"
- **So-what**: asks why the finding matters. Example: "suppose you're right, what
  changes because of it?"
- **Generalization**: asks how far the claim travels. Example: "does this hold
  beyond the people in your sample?"
- **AI-use**: probes the integrity of your process. Example: "you used an AI tool
  to find sources, how did you confirm they exist?" Your answer here points at your
  ledger and your verification step, Ask then Verify then Document, performed.

Two more tools finish the toolkit.

- **The honest-don't-know formula**: when you genuinely don't know, say so *with
  substance*: "I don't know, but here is what I DO know that bears on it, and here
  is how one would find out." Three parts: the honest admission, the adjacent thing
  you can stand on, and the method that would resolve it. A prepared don't-know
  reads as confidence. **Bluffing**, inventing an answer you do not have, is the one
  move that can destroy a presenter's credibility on the spot.
- **Defend vs concede**: the judgment under every answer. You **defend** when your
  claim ledger backs the claim, pointing at the row. You **concede** when the
  questioner names a real limit: "that's a fair hit, it moves my claim from X to Y."
  A graceful concession raises your credibility; a dug-in defense of an
  indefensible point sinks it. The test is simple. Does my ledger back this? If yes,
  defend. If no, concede and state the smaller claim that survives.

> **A question that often comes up here:** *"Isn't saying 'I don't know' at my own
> poster a disaster?"* Only as a bare shrug. In the full form it is the opposite.
> "I don't know whether it holds for graduate peers, my frame was undergraduates,
> what I *do* know is the pattern was stable across three undergraduate years, and
> to find out you'd re-run it on a graduate sample" reads as someone who knows
> exactly where their knowledge stops. That precision is the most credible thing at
> the poster.

```python
# Draw the five hard-question types as spokes around your claim.
# (networkx is imported once in the setup cell.)
Q = nx.DiGraph()
center = "YOUR CLAIM\n+ its boundary"
types = ["Method", "Alternative\nexplanation", "So-what", "Generalization", "AI-use"]
for t in types:
    Q.add_edge(t, center)

import numpy as _np
angles = _np.linspace(0.5 * _np.pi, 2.5 * _np.pi, len(types), endpoint=False)
qpos = {center: (0.0, 0.0)}
for t, a in zip(types, angles):
    qpos[t] = (_np.cos(a), _np.sin(a))

fig, ax = plt.subplots(figsize=(8.5, 6))
nx.draw_networkx_nodes(Q, qpos, nodelist=[center], node_color="#DCE6F1",
                       edgecolors="#222222", linewidths=3.0, node_size=6000, ax=ax)
nx.draw_networkx_nodes(Q, qpos, nodelist=types, node_color="#EAEAF2",
                       edgecolors="#4C72B0", linewidths=1.5, node_size=4200, ax=ax)
nx.draw_networkx_labels(Q, qpos, font_size=9, ax=ax)
nx.draw_networkx_edges(Q, qpos, arrowstyle="-|>", arrowsize=15, edge_color="#777777",
                       width=1.3, node_size=4200, ax=ax)
ax.set_title("Every fair hard question aims at one of five spokes")
ax.axis("off")
plt.tight_layout()
plt.show()

print("✓ Question-type map drawn.")
print("  Write ONE hard, fair version of each spoke for your own poster.")
print("  At least one prepared answer should be an honest don't-know in the full form.")
```

**Reading the diagram.** Every fair hard question a stranger can ask aims at one of
these five spokes, all pointed at the same target: your claim and its boundary. If
you have written and answered one hard version of each spoke, you have covered the
space of fair questions, and a surprise on Expo day is far more likely to be a
variant you can adapt than a bolt from nowhere.

*In plain terms, the picture says the hard questions are not infinite. They come in
five flavors, and you can prepare one of each in advance.*
