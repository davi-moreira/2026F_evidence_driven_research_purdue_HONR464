```python
# M0 baseline scaffold — edit the two lists below; everything else derives from
# them. Single source of truth: CURIOSITIES and CITATION_CATCH. Rename or reword
# freely; the summary recomputes from whatever you put here. Nothing downstream
# hardcodes your text, so an edit above cannot fall out of sync below.
from collections import Counter

# EDIT ME — your three genuine curiosities, one short label each. Open three
# DIFFERENT doors, not one door painted three colors.
CURIOSITIES = [
    "EDIT ME — curiosity 1 (a few words)",
    "EDIT ME — curiosity 2 (a different door)",
    "EDIT ME — curiosity 3 (a third door)",
]

# EDIT ME — your Lecture 1 citation catch. One verdict per source you checked:
# "retrieved" (you opened the real paper), "distorted" (a detail is wrong), or
# "not_found" (nothing turned up). Add or remove rows to match what you did.
CITATION_CATCH = ["retrieved", "retrieved", "not_found", "distorted"]

# --- derived summary: nothing below needs editing ---
verdicts = Counter(v.lower().strip() for v in CITATION_CATCH)
n_checked = sum(verdicts.values())
n_confirmed = verdicts.get("retrieved", 0)

# A reproducible pitch order, so the curiosity you pitch first is decided the same
# way on every run (seed 464). Everything here reads from CURIOSITIES only.
pitch_order = list(rng.permutation(len(CURIOSITIES)))
first_pitch = CURIOSITIES[pitch_order[0]]

print(f"Curiosities opened : {len(CURIOSITIES)}")
print(f"Sources checked    : {n_checked}  "
      f"(retrieved {n_confirmed}, distorted {verdicts.get('distorted', 0)}, "
      f"not found {verdicts.get('not_found', 0)})")
print(f"Reproducible pitch order (seed {SEED}): {[i + 1 for i in pitch_order]}")
print(f"→ a fair way to pick your 30-second pitch: {first_pitch}")

assert len(CURIOSITIES) == 3, "M0 asks for exactly three curiosities — open three doors."
assert n_checked >= 1, "Log at least one citation-catch verdict from Lecture 1."
print("\n✓ Baseline scaffold ran — replace every EDIT ME, then re-run.")
```
