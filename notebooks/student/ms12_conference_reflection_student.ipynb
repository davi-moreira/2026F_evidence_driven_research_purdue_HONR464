## 🎯 Definition of Done

You are done with this sprint when your submission contains all six parts of the
package, built from **your own** Expo afternoon:

1. **The coded question tally.** Every question you were asked, in a real table,
   coded by **type** (method, alternative-explanation, so-what, generalization,
   AI-use, other), **target** (claims vs communication), and **recurrence** (how
   many times).
2. **The pattern you found.** One or two sentences read straight off the counts:
   which type recurred most, whether the heavier target was claims or
   communication, and the one claims question that recurred enough to be a finding.
3. **The honest reflection paragraph.** What the Expo changed in how you see your
   project, stated with its uncertainty and its limits — not "it went well."
4. **The poster-criticism portfolio.** Three structured, kind-but-critical readings
   of peers' posters, each naming an **axis** (claims or communication), pointing at
   the evidence on the poster, and phrased so it attacks the claim, not the person.
5. **The self-evaluation.** One line on what your presentation did well and one on
   what you would do differently, judged against your counts, not the applause.
6. **AI Research Ledger rows.** One row for every AI-assisted step, each with a
   named verification method. A missing ledger returns the submission.

Plus the **dossier update line** naming where each part now lives. Remember the two
halves: the **Expo presentation** is graded live and terminal (present your poster,
evaluate at least three peers), and the written package is due at the **Friday
reflection studio**. The dated version of both lives in the Brightspace brief.

> **A question that often comes up here:** *"Most of my feedback was positive.
> Doesn't that mean my project is solid?"* It means you communicated well, which is
> worth a great deal and worth almost nothing to the truth of your claim. A warm
> room is a fact about your communication. Only the coded, counted questions tell
> you what your evidence still owes.

> - [ ] For every re-code it proposes, decide yourself whether the question doubted

> - [ ] Log this use in your AI Research Ledger: task, tool, decision, verification.

```python
# Coded-tally scaffold — replace the EDIT ME rows with YOUR OWN Expo questions.
# The DataFrame `tally` is the single source of truth: every summary below is
# recomputed from it, so there is no separate count to fall out of sync. Keep the
# four columns; use only the type codes in TYPE_CODES and the two target values.

# The valid vocabulary, defined once. Editing a row to an off-list code trips the
# check below instead of silently mis-tallying.
TYPE_CODES   = ["method", "alternative-explanation", "so-what",
                "generalization", "AI-use", "other"]
TARGET_CODES = ["claims", "communication"]

# EDIT ME — replace these placeholder rows with the questions you actually logged.
tally = pd.DataFrame([
    {"question": "Would this hold on a different campus or sample?",      "type": "generalization",          "target": "claims",        "count": 3},
    {"question": "Could something you didn't measure explain the gap?",   "type": "alternative-explanation", "target": "claims",        "count": 2},
    {"question": "I couldn't tell which line was your comparison group.", "type": "other",                   "target": "communication", "count": 4},
    {"question": "How exactly did you estimate that number?",             "type": "method",                  "target": "claims",        "count": 2},
    {"question": "Did you use AI in the analysis, and did you check it?", "type": "AI-use",                  "target": "claims",        "count": 1},
])

# Edit-safe checks: catch a typo'd code before it corrupts the counts.
assert set(tally["type"]).issubset(TYPE_CODES),     f"unknown type code — use only {TYPE_CODES}"
assert set(tally["target"]).issubset(TARGET_CODES), f"unknown target — use only {TARGET_CODES}"

# Everything below is DERIVED from `tally` (the single source of truth).
total     = int(tally["count"].sum())
by_type   = tally.groupby("type")["count"].sum().sort_values(ascending=False)
by_target = tally.groupby("target")["count"].sum().sort_values(ascending=False)
loudest   = tally.loc[tally["count"].idxmax()]

print(f"✓ Coded {len(tally)} distinct questions, {total} asked in total")
print()
print("By TYPE (weighted by recurrence):")
print(by_type.to_string())
print()
print("By TARGET (claims vs communication):")
print(by_target.to_string())
print()
print(f"Most-repeated single question ({int(loudest['count'])}x, target={loudest['target']}):")
print(f'  "{loudest["question"]}"')

# Why capture in the moment: memory resamples your afternoon. Expand to one row per
# asked question, then keep a random subset (seeded) to imitate partial recall.
instances = tally.loc[tally.index.repeat(tally["count"])].reset_index(drop=True)
true_claims_share = (instances["target"] == "claims").mean()
print()
print(f"Full sheet: claims share = {true_claims_share:.2f} "
      f"({int((instances['target']=='claims').sum())} of {len(instances)})")
n_recalled = max(3, round(len(instances) * 0.4))  # you recall only part of the afternoon
for i in range(3):
    idx = rng.choice(len(instances), size=n_recalled, replace=False)
    share = (instances.iloc[idx]["target"] == "claims").mean()
    print(f"  Memory-only recall #{i + 1} ({n_recalled} of {len(instances)} "
          f"questions): claims share = {share:.2f}")
print("  A partial recall wobbles from draw to draw; the full sheet does not.")
```

### 🗡️ Red-Team Exchange

The studio's 29–41 minute block. Pair with one partner and run two rounds of attack.
Bring in an AI reviewer as a third voice using the **Poster Critic** role card
(`ai_resources/agent_role_cards.md`; full briefing in
`genai_studio/roles/poster_critic.md`). The Poster Critic flags untraceable numbers
and claim-boundary overreach in a pasted poster claim, which is exactly the read you
need for both rounds.

**Round 1: attack your reflection's reading (about 6 minutes).**
Your partner reads your pattern sentence and your reflection, then checks whether the
pattern is read off your **counts** or off your **mood**. They challenge one claim
you cannot trace to a row: "show me the count that says the recurring doubt was
about your claims, not your figure." You defend or concede, then rewrite so every
sentence points at a number in your tally.

**Round 2: attack one portfolio entry (about 6 minutes).**
Take one criticism from your poster-criticism portfolio. Your partner, and the
Poster Critic on the peer's pasted claim and evidence, test two things: is the gap
you named **really on the poster**, and did you tag the right **axis** (claims vs
communication)? A gap neither your partner nor you can point to on the poster is one
you imagined; cut it. A criticism that names no axis is a vibe; fix it.

**What to log.** Write down the single strongest hit from each round and how you
changed the package in response. An attack that changed nothing still gets a line
saying why it did not land. That record is part of what you submit.
